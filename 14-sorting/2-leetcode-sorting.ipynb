## 1. LeetCode 179 – Largest Number

**Problem:** Given a list of non-negative integers `nums`, arrange them such that they form the largest number and return it as a string.

**Example:**
- Input: `nums = [3, 30, 34, 5, 9]`
- Output: `"9534330"`

**Key Insight:** Compare two numbers a, b by checking whether str(a)+str(b) > str(b)+str(a).

**Approaches:**
| Approach | TC | SC |
|----------|----|----|
| Brute Force (all permutations) | O(n! × n) | O(n) |
| Optimal (custom sort) | O(n log n) | O(n) |


### Dry Run — nums = [3, 30, 34, 5, 9]
| Step | Action | Result |
|------|--------|--------|
| 1 | Convert to string | ["3","30","34","5","9"] |
| 2 | Multiply strings (x*10) | ["3333..","3030..","3434..","5555..","9999.."] |
| 3 | Sort descending | ["9","5","34","3","30"] |
| 4 | Join strings | "9534330" |
| 5 | Check leading zero | Not needed |
| 6 | Final output | "9534330" |

👉 x * 10 means: repeat the number to make it longer so we can compare it properly.

Example:

"3"  → "3333333333"  
"30" → "3030303030"  

Now see:    
👉 "3333..." > "3030..."   
👉 so 3 comes before 30   

````python

from typing import List
class Solution:
    def largestNumber(self, nums:List[int])-> str:
        nums_str = [str(num) for num in nums] #["3","30","34","5","9"]
        nums_str.sort(key=lambda x: x*10, reverse =True)
        #multiply =>(x*10) | ["3333..","3030..","3434..","5555..","9999.."]
        # after sorting=> ["9","5","34","3","30"]
        result = ''.join(nums_str)
        return result if result[0] !='0' else "0"
# Run
sol = Solution()
print(sol.largestNumber([3, 30, 34, 5,9]))
````

## 2. LeetCode 274 – H-Index

**Problem:** Given an array `citations` where `citations[i]` is the number of citations a researcher's i-th paper has received, return the researcher's **h-index**.

H-index = largest value `h` such that at least `h` papers have `>= h` citations.

**Example:**
- Input: `citations = [3, 0, 6, 1, 5]` → Output: `3`
  - 3 papers (cited 3, 5, 6) have >= 3 citations ✅

**Approaches:**
| Approach | TC | SC |
|----------|----|----|
| Brute Force (check all h) | O(n²) | O(1) |
| Optimal (sort + scan) | O(n log n) | O(1) |


### Dry Run — citations = [3, 0, 6, 1, 5]

**Step 1:** Sort descending: `[6, 5, 3, 1, 0]`

**Step 2:** For each index i (0-based), check if `citations[i] >= i+1`

| i | citations[i] | h candidate (i+1) | citations[i] >= i+1? |
|---|-------------|-------------------|----------------------|
| 0 | 6           | 1                 | ✅ Yes → h = 1        |
| 1 | 5           | 2                 | ✅ Yes → h = 2        |
| 2 | 3           | 3                 | ✅ Yes → h = 3        |
| 3 | 1           | 4                 | ❌ No → stop          |

**Result:** h = 3

**Why this works:** After sorting descending, the i-th paper (0-indexed) is one of the top (i+1) papers. If it has >= (i+1) citations, then at least (i+1) papers have >= (i+1) citations.


| Step | We are checking how many papers | Papers considered | Minimum required citations | Actual citations of last paper | Condition | h |
|------|-------------------------------|------------------|----------------------------|-------------------------------|----------|---|
| 1 | Check 1 paper | [6] | Each should have ≥ 1 | 6 | 6 ≥ 1 ✅ | h = 1 |
| 2 | Check 2 papers | [6, 5] | Each should have ≥ 2 | 5 | 5 ≥ 2 ✅ | h = 2 |
| 3 | Check 3 papers | [6, 5, 3] | Each should have ≥ 3 | 3 | 3 ≥ 3 ✅ | h = 3 |
| 4 | Check 4 papers | [6, 5, 3, 1] | Each should have ≥ 4 | 1 | 1 ≥ 4 ❌ | STOP |

h = count of “good papers”

👉 h matlab: kitne papers ache perform kar rahe hain (minimum utne hi citations ke saath)

````python
from typing import List
class Solution:
    def hIndex(self,citations :List[int])-> int:
        citations.sort(reverse=True)
        h = 0
        for i, c in enumerate(citations):
            if c >= i + 1:  #6>= 0+1, #5>=1+1 #3>= 2+1

                h = i + 1# h =0+1 =1, #h =1+1 =2, # 2+1=3
                
            else:
                break
        return h  
# Run
sol = Solution()
print(sol.hIndex([3, 0, 6,1, 5]))              
````

|--|--|
| Time Complexity | O(n log n) — sorting |
| Space Complexity | O(1) — in-place sort |


## 3. LeetCode 148 – Sort List

**Problem:** Sort a linked list in O(n log n) time and O(1) space.

**Example:**
- Input: `4 → 2 → 1 → 3`
- Output: `1 → 2 → 3 → 4`

**Approaches:**
| Approach | TC | SC |
|----------|----|----|
| Convert to array, sort, rebuild | O(n log n) | O(n) |
| Merge Sort (bottom-up) | O(n log n) | O(1) |
| Merge Sort (top-down, recursive) | O(n log n) | O(log n) |

**Optimal:** Merge Sort on linked list (top-down recursive)


### Dry Run — [4, 2, 1, 3]

**Top-Down Merge Sort:**

```
Split:   [4,2,1,3]
         [4,2]  [1,3]
         [4][2] [1][3]

Merge:   [2,4]  [1,3]
Merge:   [1,2,3,4]
```

| Step | Action | List State |
|------|--------|------------|
| 1 | Find middle of [4→2→1→3] → mid=2 | Left=[4→2], Right=[1→3] |
| 2 | Find middle of [4→2] → mid=4 | Left=[4], Right=[2] |
| 3 | Merge [4] and [2] | [2→4] |
| 4 | Find middle of [1→3] → mid=1 | Left=[1], Right=[3] |
| 5 | Merge [1] and [3] | [1→3] |
| 6 | Merge [2→4] and [1→3] | [1→2→3→4] |

**Merge step (2→4 with 1→3):**

| Compare | Pick | Remaining |
|---------|------|-----------|
| 2 vs 1 | 1 | [2→4] vs [3] |
| 2 vs 3 | 2 | [4] vs [3] |
| 4 vs 3 | 3 | [4] vs [] |
| - | 4 (append rest) | done |


````python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        # Base case: 0 or 1 node
        if not head or not head.next:
            return head

        # Step 1: Find middle using slow/fast pointers
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        # Step 2: Split into two halves
        mid = slow.next
        slow.next = None   # Cut the list

        # Step 3: Recursively sort both halves
        left = self.sortList(head)
        right = self.sortList(mid)

        # Step 4: Merge the two sorted halves
        return self.merge(left, right)

    def merge(self, l1, l2):
        dummy = ListNode(0)
        curr = dummy
        while l1 and l2:
            if l1.val <= l2.val:
                curr.next = l1
                l1 = l1.next
            else:
                curr.next = l2
                l2 = l2.next
            curr = curr.next
        curr.next = l1 or l2   # Append remaining
        return dummy.next

# Helper to build and print list
def build(vals):
    dummy = ListNode(0)
    curr = dummy
    for v in vals:
        curr.next = ListNode(v)
        curr = curr.next
    return dummy.next

def to_list(node):
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result

sol = Solution()
head = build([4, 2, 1, 3])
print(to_list(sol.sortList(head)))   # [1, 2, 3, 4]
````

|--|--|
| Time Complexity | O(n log n) — merge sort |
| Space Complexity | O(log n) — recursive call stack |


## 4. LeetCode 969 – Pancake Sorting

**Problem:** Given an array `arr`, sort it using pancake flips. A pancake flip of size k reverses the first k elements. Return the sequence of k values for each flip.

**Example:**
- Input: `arr = [3, 2, 4, 1]`
- Output: `[4, 2, 4, 3]` (one valid answer)

**Strategy (Selection Sort style):**
1. Find max in unsorted portion
2. If not already at front, flip to front (flip max_idx+1)
3. Flip to correct final position (flip current_size)
4. Repeat for decreasing size

**Approaches:**
| Approach | TC | SC |
|----------|----|----|
| Brute Force (bubble sort with flips) | O(n²) | O(n) |
| Optimal (selection sort with flips) | O(n²) | O(n) |


### Dry Run — arr = [3, 2, 4, 1]

**Iteration 1 (size=4):** Find max=4 at index 2

| Step | Action | arr | k values |
|------|--------|-----|----------|
| 1 | max=4 at idx=2 | [3,2,4,1] | [] |
| 2 | Flip(3): bring 4 to front | [4,2,3,1] | [3] |
| 3 | Flip(4): send 4 to end | [1,3,2,4] | [3,4] |

**Iteration 2 (size=3):** Find max=3 at index 1

| Step | Action | arr | k values |
|------|--------|-----|----------|
| 1 | max=3 at idx=1 | [1,3,2,4] | [3,4] |
| 2 | Flip(2): bring 3 to front | [3,1,2,4] | [3,4,2] |
| 3 | Flip(3): send 3 to position 2 | [2,1,3,4] | [3,4,2,3] |

**Iteration 3 (size=2):** Find max=2 at index 0

| Step | Action | arr | k values |
|------|--------|-----|----------|
| 1 | max=2 at idx=0 | [2,1,3,4] | [3,4,2,3] |
| 2 | Already at front, skip flip | | |
| 3 | Flip(2): send 2 to position 1 | [1,2,3,4] | [3,4,2,3,2] |

**Result:** arr = [1,2,3,4] ✅


````python
from typing import List

class Solution:
    def pancakeSort(self, arr: List[int]) -> List[int]:
        result = []
        n = len(arr)

        for size in range(n, 1, -1):
            # Find index of max element in arr[0..size-1]
            max_idx = arr.index(max(arr[:size]))

            if max_idx == size - 1:
                continue   # Already in correct position

            # Flip max to the front (if not already there)
            if max_idx != 0:
                result.append(max_idx + 1)
                arr[:max_idx + 1] = arr[:max_idx + 1][::-1]

            # Flip max to its correct final position
            result.append(size)
            arr[:size] = arr[:size][::-1]

        return result

# Test
sol = Solution()
arr = [3, 2, 4, 1]
print(sol.pancakeSort(arr))   # [3, 4, 2, 3, 2] or another valid sequence
print(arr)                     # [1, 2, 3, 4] (sorted)

arr2 = [1, 2, 3]
print(sol.pancakeSort(arr2))   # [] (already sorted)
````

### Complexity — LeetCode 969

| | Value |
|--|--|
| Time Complexity | O(n²) — n iterations, each with O(n) flip |
| Space Complexity | O(n) — result array |


## 5. LeetCode 253 – Meeting Rooms II

**Problem:** Given an array of meeting time intervals `intervals[i] = [start, end]`, return the **minimum number of conference rooms** required.

**Example:**
- Input: `intervals = [[0,30],[5,10],[15,20]]`
- Output: `2`

**Key Insight:** Sort by start time. Use a min-heap of end times. If the next meeting starts after the earliest ending meeting, reuse that room. Otherwise, add a new room.

**Approaches:**
| Approach | TC | SC |
|----------|----|----|
| Brute Force (check all pairs) | O(n²) | O(n) |
| Optimal (sort + min-heap) | O(n log n) | O(n) |
| Two-pointer (separate start/end arrays) | O(n log n) | O(n) |


### Dry Run — intervals = [[0,30],[5,10],[15,20]]

**Step 1:** Sort by start time: `[[0,30],[5,10],[15,20]]` (already sorted)

**Step 2:** Process with min-heap (tracks end times of active meetings):

| Step | Meeting | Heap before | Action | Heap after | Rooms |
|------|---------|------------|--------|------------|-------|
| 1 | [0,30] | [] | Add new room (heap empty) | [30] | 1 |
| 2 | [5,10] | [30] | 5 < 30, add new room | [10,30] | 2 |
| 3 | [15,20] | [10,30] | 15 >= 10, reuse room (pop 10, push 20) | [20,30] | 2 |

**Final answer: 2 rooms** ✅

**Two-Pointer approach (alternative):**

| starts sorted | ends sorted | i | j | rooms | max_rooms |
|--------------|------------|---|---|-------|-----------|
| [0,5,15] | [10,20,30] | 0 | 0 | 1 | 1 |
| | | 1 | 0 | 2 | 2 |
| | | 2 | 1 | 2 | 2 |


````python
from typing import List
import heapq

class Solution:
    def minMeetingRooms(self, intervals: List[List[int]]) -> int:
        if not intervals:
            return 0

        # Sort by start time
        intervals.sort(key=lambda x: x[0])

        # Min-heap stores end times of active meetings
        heap = []

        for start, end in intervals:
            # If earliest ending meeting finishes before this one starts, reuse room
            if heap and heap[0] <= start:
                heapq.heappop(heap)   # Free that room

            # Add this meeting's end time (new room or reused room)
            heapq.heappush(heap, end)

        # Number of rooms in use = size of heap
        return len(heap)

# Test
sol = Solution()
print(sol.minMeetingRooms([[0,30],[5,10],[15,20]]))   # 2
print(sol.minMeetingRooms([[7,10],[2,4]]))             # 1
print(sol.minMeetingRooms([[1,5],[2,6],[3,7]]))        # 3
````

| | Value |
|--|--|
| Time Complexity | O(n log n) — sorting + heap operations |
| Space Complexity | O(n) — heap stores up to n end times |
