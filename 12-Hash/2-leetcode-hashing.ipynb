#  - 1 #383 â€“ Ransom Note
Given two strings ransomNote and magazine, check if you can construct ransomNote using letters from magazine.
Each letter in magazine can only be used once.

ðŸ§© Approach (Hashing / Counter)  
- Count frequency of each character in magazine.
- For each character in ransomNote, check if it exists in magazine with enough count.
- If any character is missing or exhausted, return False.
- Otherwise, return True.

magazine has 'a':2, 'b':1  
ransomNote needs 'a':2  
Enough letters â†’ âœ… True

| Step |  ch | freq[ch] before | Action  | freq after    |
| ---: | :-: | :-------------: | ------- | ------------- |
|    1 |  a  |        2        | Use `a` | {'a':1,'b':1} |
|    2 |  a  |        1        | Use `a` | {'a':0,'b':1} |


```python
from collections import Counter

class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        # {'a': 2, 'b': 1}
        freq = Counter(magazine) 
        for ch in ransomNote: #ch=a, ch=a
            if freq[ch] <=0:
                return False
            freq[ch] -= 1 
        return True 
ransomNote = "aa"      
magazine ="aba" 
print(Solution().canConstruct(ransomNote, magazine))
```

| i | nums[i] | right BEFORE | result[i] BEFORE | result[i] AFTER | right AFTER |
| - | ------- | ------------ | ---------------- | --------------- | ----------- |
| 3 | 4       | 1            | 6                | 6 Ã— 1 = 6       | 1 Ã— 4 = 4   |
| 2 | 3       | 4            | 2                | 2 Ã— 4 = 8       | 4 Ã— 3 = 12  |
| 1 | 2       | 12           | 1                | 1 Ã— 12 = 12     | 12 Ã— 2 = 24 |
| 0 | 1       | 24           | 1                | 1 Ã— 24 = 24     | 24 Ã— 1 = 24 |


```python
class Solution:
    def productExceptSelf(self, nums):
        n = len(nums)
        result = [1] * n

        left = 1
        for i in range(n):
            result[i] = left
            left *= nums[i]

        right = 1
        for i in range(n - 1, -1, -1):
            result[i] *= right
            right *= nums[i]

        return result
```

```python
class Solution:
    def productExceptSelf(self, nums):
        n = len(nums)
        hashmap = {}
        total_product = 1

        # Step 1: Store values in hash map and compute total product
        for i in range(n):
            hashmap[i] = nums[i]
            total_product *= nums[i]

        # Step 2: Build result
        result = [0] * n #[0,0]
        for i in range(n):
            result[i] = total_product // hashmap[i]

        return result
# Correct usage
print(Solution().productExceptSelf([2, 3]))
```

Check:  
res[0] = 6 â†’ product of all except nums[0] = 2Ã—3 = 6 âœ…  
res[1] = 3 â†’ product of all except nums[1] = 1Ã—3 = 3 âœ…  
res[2] = 2 â†’ product of all except nums[2] = 1Ã—2 = 2 âœ…

```python
class Solution:
    def productExceptSelf(self, nums):
        n = len(nums) #3
        res = [1] * n  # Result array

        # Step 1: Compute prefix product (left side)
        left = 1
        for i in range(n):
            res[i] = left
            left *= nums[i]

        # Step 2: Compute suffix product (right side) and multiply
        right = 1
        for i in range(n-1, -1, -1):
            res[i] *= right
            right *= nums[i]

        return res

print(Solution().productExceptSelf([1, 2, 3]))
    
```

# 3. LeetCode 451 â€” Sort Characters By Frequency
- Count frequency of each character using a hash map
- Sort characters by their frequency (highest first)
- Build the answer string
Input:  "tree"   
Counts: {'t':1, 'r':1, 'e':2}  
Output: "eetr" (or "eert")


s = "tree"  
freq = {'t':1, 'r':1, 'e':2}  
sorted freq = [('e',2), ('t',1), ('r',1)]  
result = "ee" + "t" + "r" = "eetr"  


```python
from collections import Counter

class Solution:
    def frequencySort(self, s: str) -> str:
        # 1. Count frequency using hash map
        #freq = {'t': 1, 'r': 1, 'e': 2}
        freq = Counter(s)

        # 2. Sort characters by frequency (descending)
        # sorted_chars = [('e', 2), ('t', 1), ('r', 1)]
        freq_items = sorted(freq.items(), key = lambda x:x[1], reverse=True)
        
        # 3. Build result string
        ## result = ['ee', 't', 'r']
        result = ""
        for ch, count in freq_items:
            result += ch * count        
        #return 'eetr'  # or 'eert'
        return result
print(Solution().frequencySort("tree"))    
```

## 4 #387. Find the First Non-Repeating Character

Input:  s = "leetcode"

Counter:  {'l':1, 'e':3, 't':1, 'c':1, 'o':1, 'd':1}

Loop:  i=0, char='l', count=1 â†’ return 0

Output: 0 âœ…

```python
from collections import Counter
class Solution:
    def firstUniqChar(self, s: str) -> int:
        count = Counter(s)
        for i, char in enumerate(s):
            if count[char] == 1:
                return i
        return -1          
print(Solution().firstUniqChar("leetcode"))    
```

## 5 #3541. Find Most Frequent Vowel and Consonant

s = "succes" (note: one s, one u, two c, one e, one s â†’ total letters = 6)  
ðŸ”¹ Character Frequency (Counter)  
freq = {'s': 2, 'u': 1, 'c': 2, 'e': 1}  
vowels = {'a','e','i','o','u'}

| Step | ch | count | ch in vowels? | max_vowel BEFORE | max_vowel AFTER | max_consonant BEFORE | max_consonant AFTER | Explanation |
|-----|----|-------|---------------|------------------|-----------------|----------------------|---------------------|-------------|
| 0 | - | - | - | 0 | 0 | 0 | 0 | Initialization |
| 1 | s | 2 | No | 0 | 0 | 0 | 2 | 's' is consonant â†’ update max_consonant |
| 2 | u | 1 | Yes | 0 | 1 | 2 | 2 | 'u' is vowel â†’ update max_vowel |
| 3 | c | 2 | No | 1 | 1 | 2 | 2 | 'c' is consonant â†’ no change |
| 4 | e | 1 | Yes | 1 | 1 | 2 | 2 | 'e' is vowel â†’ no change |
| Final | - | - | - | 1 | 1 | 2 | 2 | Return 1 + 2 = **3** |


```python
from collections import Counter

class Solution:
    def maxFreqSum(self, s: str) -> int:
        freq = Counter(s)
        vowels = set("aeiou") 
        max_vowel = 0
        max_consonant = 0
        for ch, count in freq.items():
            if ch in vowels:
                max_vowel = max(max_vowel, count)
            else:
                max_consonant = max(max_consonant, count)    
        
        
        return max_vowel + max_consonant
print(Solution().maxFreqSum("succes"))        
```
