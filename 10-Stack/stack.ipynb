# Stack — Complete Basics

A **Stack** is a linear data structure that follows **LIFO** — Last In, First Out.

The last element inserted is the first one removed.

Think of a pile of plates:
- You put a new plate on **top**
- You remove from the **top**
- To reach the bottom plate, remove all plates above it first

```
        ┌───┐   <- TOP (last inserted, first removed)
        │ 3 │
        ├───┤
        │ 2 │
        ├───┤
        │ 1 │
        └───┘   <- BOTTOM (first inserted, last removed)
```

---
## Key Stack Operations

| Operation | Description | Time Complexity |
|---|---|---|
| `push(val)` | Add element to top | O(1) |
| `pop()` | Remove and return top element | O(1) |
| `peek()` | See top element without removing | O(1) |
| `isEmpty()` | Check if stack is empty | O(1) |
| `size()` | Number of elements | O(1) |

---
## 1. Stack Using Python List (Simple)

Python's built-in list is perfect for a stack:
- `append()` = push (add to end = top)
- `pop()` = pop (remove from end = top)
- `list[-1]` = peek (see last element)

````python
class Stack:
    def __init__(self):
        self.list = []

    def __str__(self):
        # Print from top to bottom
        values = [str(x) for x in reversed(self.list)]
        return '\n'.join(values)

    def isEmpty(self):
        return self.list == []

    def push(self, value):
        self.list.append(value)
        return "inserted"

    def pop(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.list.pop()

    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.list[-1]

    def size(self):
        return len(self.list)

    def delete(self):
        self.list = []


# Test
s = Stack()
s.push(1)
s.push(2)
s.push(3)
s.push(4)

print("Stack (top to bottom):")
print(s)
print("\nPeek:", s.peek())
print("Size:", s.size())
print("Pop:", s.pop())
print("After pop:", s.peek())
print("isEmpty:", s.isEmpty())
s.delete()
print("After delete, isEmpty:", s.isEmpty())
````

## 2. Stack With Max Size (Bounded Stack)

Sometimes you need to limit how many elements a stack can hold.
Add a `maxsize` check in `push()`.

````python
class BoundedStack:
    def __init__(self, maxsize):
        self.maxsize = maxsize
        self.list = []

    def __str__(self):
        values = [str(x) for x in reversed(self.list)]
        return '\n'.join(values)

    def isEmpty(self):
        return self.list == []

    def isFull(self):
        return len(self.list) == self.maxsize

    def push(self, value):
        if self.isFull():
            return "Stack is Full"
        self.list.append(value)
        return "inserted"

    def pop(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.list.pop()

    def peek(self):
        if self.isEmpty():
            return "Stack is Empty"
        return self.list[-1]

    def delete(self):
        self.list = []


# Test
s = BoundedStack(3)    # max 3 elements
print(s.push(10))      # inserted
print(s.push(20))      # inserted
print(s.push(30))      # inserted
print(s.push(40))      # Stack is Full
print("isFull:", s.isFull())
print("Peek:", s.peek())
print("Pop:", s.pop())
print("isFull after pop:", s.isFull())
print(s.push(99))      # inserted now
````

## 3. Stack Using Linked List (Node + top pointer)

Instead of a Python list, use nodes connected by pointers.

```
Push 1: top -> [1|None]
Push 2: top -> [2|next] -> [1|None]
Push 3: top -> [3|next] -> [2|next] -> [1|None]
Pop   : top -> [2|next] -> [1|None]   (3 removed)
```

**Advantage:** No fixed size, no wasted memory. Size tracked with `length` counter → O(1).

````python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


class LinkedStack:
    def __init__(self):
        self.top = None      # points to top node
        self.length = 0      # O(1) size check

    def __str__(self):
        current = self.top
        values = []
        while current:
            values.append(str(current.value))
            current = current.next
        return ' -> '.join(values)   # e.g. "3 -> 2 -> 1"

    def __len__(self):
        return self.length

    def isEmpty(self):
        return self.top is None

    def push(self, value):
        new_node = Node(value)
        new_node.next = self.top   # new node points to current top
        self.top = new_node        # new node becomes the top
        self.length += 1

    def pop(self):
        if self.isEmpty():
            return "Stack is empty"
        popped = self.top
        self.top = self.top.next   # move top down
        self.length -= 1
        return popped.value

    def peek(self):
        if self.isEmpty():
            return "Stack is empty"
        return self.top.value

    def clear(self):
        self.top = None
        self.length = 0


# Test
s = LinkedStack()
s.push(1)
s.push(2)
s.push(3)
print("Stack:", s)             # 3 -> 2 -> 1
print("Peek:", s.peek())       # 3
print("Pop:", s.pop())         # 3
print("Stack:", s)             # 2 -> 1
print("Length:", len(s))       # 2
````

## 4. Multi-Stack (Multiple Stacks in One Array)

Store 3 stacks inside a **single array** by dividing it into equal sections.

```
stacksize = 3, numstacks = 3  →  array of size 9

Index:   0   1   2   |  3   4   5  |  6   7   8
         ← Stack 0 → |  ← Stack 1→ | ← Stack 2 →
```

**Why?** Memory efficient when you need multiple stacks but want to avoid separate allocations.

````python
class MultiStack:
    def __init__(self, stacksize):
        self.numstacks = 3
        self.array = [0] * (stacksize * self.numstacks)  # single array for all 3 stacks
        self.sizes = [0] * self.numstacks                # how many elements in each stack
        self.stacksize = stacksize

    def isFull(self, stacknum):
        return self.sizes[stacknum] == self.stacksize

    def isEmpty(self, stacknum):
        return self.sizes[stacknum] == 0

    def indexOfTop(self, stacknum):
        offset = stacknum * self.stacksize           # where this stack starts
        return offset + self.sizes[stacknum] - 1    # current top index

    def push(self, item, stacknum):
        if self.isFull(stacknum):
            return f"Stack {stacknum} is Full"
        self.sizes[stacknum] += 1
        self.array[self.indexOfTop(stacknum)] = item
        return "inserted"

    def pop(self, stacknum):
        if self.isEmpty(stacknum):
            return f"Stack {stacknum} is Empty"
        value = self.array[self.indexOfTop(stacknum)]
        self.array[self.indexOfTop(stacknum)] = 0
        self.sizes[stacknum] -= 1
        return value

    def peek(self, stacknum):
        if self.isEmpty(stacknum):
            return f"Stack {stacknum} is Empty"
        return self.array[self.indexOfTop(stacknum)]


# Test
ms = MultiStack(3)   # 3 stacks, each can hold 3 elements

ms.push(10, 0)    # Stack 0
ms.push(11, 0)
ms.push(20, 1)    # Stack 1
ms.push(30, 2)    # Stack 2
ms.push(31, 2)

print("Peek Stack 0:", ms.peek(0))       # 11
print("Peek Stack 1:", ms.peek(1))       # 20
print("Peek Stack 2:", ms.peek(2))       # 31
print("Pop Stack 0:", ms.pop(0))         # 11
print("Peek Stack 0 now:", ms.peek(0))   # 10
print("Array:", ms.array)               # see full internal array
````

## 5. Python Built-in Stack (Use This in Interviews)

You don't need to build a class for LeetCode. Just use a plain list:

````python
stack = []

# push
stack.append(1)
stack.append(2)
stack.append(3)
print("stack:", stack)          # [1, 2, 3]

# peek
print("peek:", stack[-1])       # 3

# pop
print("pop:", stack.pop())      # 3
print("stack:", stack)          # [1, 2]

# isEmpty
print("empty:", len(stack) == 0)   # False
````
