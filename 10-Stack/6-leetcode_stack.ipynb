# 155. Min Stack

["MinStack","push","push","push","getMin","pop","top","getMin"] 

[[],[-2],[0],[-3],[],[],[],[]]


| Step | Operation    | Value | `stack` (main) | `min_stack` (minimums) | Return / Explanation         |
| ---: | ------------ | ----- | -------------- | ---------------------- | ---------------------------- |
|    0 | `MinStack()` | ‚Äì     | `[]`           | `[]`                   | Initialize both stacks empty |
|    1 | `push(-2)`   | -2    | `[-2]`         | `[-2]`                 | First element ‚Üí min = -2     |
|    2 | `push(0)`    | 0     | `[-2, 0]`      | `[-2, -2]`             | Min stays -2 ‚Üí push same min |
|    3 | `push(-3)`   | -3    | `[-2, 0, -3]`  | `[-2, -2, -3]`         | New min found ‚Üí -3           |
|    4 | `getMin()`   | ‚Äì     | `[-2, 0, -3]`  | `[-2, -2, -3]`         | **Return -3**                |
|    5 | `pop()`      | ‚Äì     | `[-2, 0]`      | `[-2, -2]`             | Top removed from both stacks |
|    6 | `top()`      | ‚Äì     | `[-2, 0]`      | `[-2, -2]`             | **Return 0**                 |
|    7 | `getMin()`   | ‚Äì     | `[-2, 0]`      | `[-2, -2]`             | **Return -2**                |


[null, null, null, null, -3, null, 0, -2]


```python
class MinStack:
    def __init__(self):
        # Main stack to store values
        self.stack = [] #-2,0,-3
        # Extra stack to store current minimum at each level
        self.min_stack = [] #-2,-2,-3

    def push(self, val: int) -> None:
        self.stack.append(val)
        # Agar min_stack empty hai ya naya value <= current min hai
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)
        else:    
            # Push the same minimum again (to keep lengths same)
            self.min_stack.append(self.min_stack[-1])

    def pop(self) -> None:
        self.stack.pop()
        self.min_stack.pop()
    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.min_stack[-1]
# Example usage of MinStack
minStack = MinStack()

minStack.push(-2)   
minStack.push(0)    
minStack.push(-3)   

print(minStack.getMin())  # Output: -3  (minimum in stack)
minStack.pop()            # removes -3
print(minStack.top())     # Output: 0   (top of stack now)
print(minStack.getMin())  # Output: -2  (new minimum)    
```

# 1172. Dinner Plate Stacks

Operations:  
["DinnerPlates","push","push","push","push","popAtStack","pop","push"]

Capacity = 2  
push(1), push(2), push(3), push(4), popAtStack(0), pop(), push(5)


```python
import heapq

class DinnerPlates:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.stacks = []
        self.available = []   # min-heap

    def push(self, val: int) -> None:
        # clean invalid indices
        while self.available and (
            self.available[0] >= len(self.stacks) or
            len(self.stacks[self.available[0]]) == self.capacity
        ):
            heapq.heappop(self.available)

        if self.available:
            idx = heapq.heappop(self.available)
            self.stacks[idx].append(val)
            if len(self.stacks[idx]) < self.capacity:
                heapq.heappush(self.available, idx)
        else:
            self.stacks.append([val])
            if self.capacity > 1:
                heapq.heappush(self.available, len(self.stacks) - 1)

    def pop(self) -> int:
        while self.stacks and not self.stacks[-1]:
            self.stacks.pop()

        if not self.stacks:
            return -1

        val = self.stacks[-1].pop()
        heapq.heappush(self.available, len(self.stacks) - 1)

        while self.stacks and not self.stacks[-1]:
            self.stacks.pop()

        return val

    def popAtStack(self, index: int) -> int:
        if index >= len(self.stacks) or not self.stacks[index]:
            return -1

        val = self.stacks[index].pop()
        heapq.heappush(self.available, index)
        return val
```

```python
D = DinnerPlates(2)
D.push(1)
D.push(2)
D.push(3)
D.push(4)
D.push(5)
print(D.popAtStack(0))  # 2
D.push(20)
D.push(21)
print(D.popAtStack(0))  # 20
print(D.popAtStack(2))  # 21
print(D.pop())  # 5
print(D.pop())  # 4
print(D.pop())  # 3
print(D.pop())  # 1
print(D.pop())  # -1
```

# 232. Implement Queue using Stacks

```python
class MyQueue(object):

    def __init__(self):
        
        self.stack_in = []   # stack used for push operations
        self.stack_out = []  # stack used for pop/peek operations

    def push(self, x):
        """
        :type x: int
        :rtype: None
        """
        self.stack_in.append(x)

    def pop(self):
        """
        :rtype: int
        """
        self.move()
        return self.stack_out.pop()

    def peek(self):
        """
        :rtype: int
        """
        self.move()
        return self.stack_out[-1]

    def empty(self):
        """
        :rtype: bool
        """
        return not self.stack_out and not self.stack_in

    def move(self):
        """Helper function to move elements from stack_in to stack_out only when needed"""
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
# Example usage of MyQueue class
q = MyQueue()

q.push(1)       # queue = [1]
q.push(2)       # queue = [1, 2]
print(q.peek()) # Output: 1  (front of queue)
print(q.pop())  # Output: 1  (removes 1, queue = [2])
print(q.empty())# Output: False (queue still has [2])

q.push(3)       # queue = [2, 3]
q.push(4)       # queue = [2, 3, 4]
print(q.pop())  # Output: 2  (removes 2, queue = [3, 4])
print(q.peek()) # Output: 3  (front element)
print(q.pop())  # Output: 3  (queue = [4])
print(q.pop())  # Output: 4  (queue = [])
print(q.empty())# Output: True (queue is empty now)                
        
```

# üêæ Animal Shelter 

Design a system to manage an animal shelter that houses only cats (0) and dogs (1).
Implement:

enqueue([num, type]) ‚Üí add animal (FIFO order)

dequeueAny() ‚Üí adopt oldest animal

dequeueDog() ‚Üí adopt oldest dog

dequeueCat() ‚Üí adopt oldest cat

If no animal is available for adoption, return [-1, -1].

Example:

Input:
["AnimalShelf","enqueue","enqueue","enqueue","dequeueDog","dequeueCat","dequeueAny"]   
[[],[0,0],[1,0],[2,1],[],[],[]]

Output:

[null,null,null,null,[2,1],[0,0],[1,0]]

| Topic / Operation          | What it Means                          | How It Works Internally                                                                                    | Example / Result                           |
| -------------------------- | -------------------------------------- | ---------------------------------------------------------------------------------------------------------- | ------------------------------------------ |
| **Animal Format**          | Each animal is represented as `[i, j]` | `i` = arrival number (smaller = older) <br> `j` = type (`0 = cat`, `1 = dog`)                              | `[3, 0]` ‚Üí 3rd animal, Cat                 |
| **Data Structure Used**    | Two separate queues                    | `q[0]` ‚Üí queue for Cats <br> `q[1]` ‚Üí queue for Dogs                                                       | `q = [cats_queue, dogs_queue]`             |
| **Constructor `__init__`** | Initialize empty shelter               | Creates two empty queues using `deque`                                                                     | `q[0] = []`, `q[1] = []`                   |
| **enqueue(animal)**        | Add a new animal                       | If `j = 0` ‚Üí add `i` to `q[0]` <br> If `j = 1` ‚Üí add `i` to `q[1]`                                         | `enqueue([2,1])` ‚Üí Dogs queue = `[2]`      |
| **Why deque is used**      | Fast insert & remove                   | `append()` and `popleft()` both O(1)                                                                       | Efficient queue behavior                   |
| **dequeueAny()**           | Return the oldest animal (cat or dog)  | 1Ô∏è‚É£ If no cats ‚Üí take dog <br> 2Ô∏è‚É£ If dogs exist and `dog_id < cat_id` ‚Üí take dog <br> 3Ô∏è‚É£ Else ‚Üí take cat | Cats `[0,2]`, Dogs `[1]` ‚Üí returns `[0,0]` |
| **Comparison Logic**       | Decide oldest animal                   | Compare **front elements** of both queues                                                                  | Smaller id = arrived earlier               |
| **dequeueDog()**           | Return oldest dog                      | If `q[1]` empty ‚Üí `[-1,-1]` <br> Else `popleft()` from `q[1]`                                              | Dogs `[1,4]` ‚Üí returns `[1,1]`             |
| **dequeueCat()**           | Return oldest cat                      | If `q[0]` empty ‚Üí `[-1,-1]` <br> Else `popleft()` from `q[0]`                                              | Cats `[2,5]` ‚Üí returns `[2,0]`             |
| **Empty Case Handling**    | No animals available                   | All dequeue methods return `[-1,-1]`                                                                       | Safe edge-case handling                    |
| **Key Idea (One Line)**    | Core logic of problem                  | Maintain **separate queues** and compare arrival order                                                     | Oldest animal always adopted               |


| Step | Operation        | Action                          | Cats `q[0]` | Dogs `q[1]` | Return  |
| ---: | ---------------- | ------------------------------- | ----------- | ----------- | ------- |
|    0 | `AnimalShelf()`  | Initialize queues               | `[]`        | `[]`        | ‚Äì       |
|    1 | `enqueue([0,0])` | Cat with id 0 added             | `[0]`       | `[]`        | ‚Äì       |
|    2 | `enqueue([1,1])` | Dog with id 1 added             | `[0]`       | `[1]`       | ‚Äì       |
|    3 | `enqueue([2,0])` | Cat with id 2 added             | `[0,2]`     | `[1]`       | ‚Äì       |
|    4 | `dequeueAny()`   | Compare front ‚Üí cat(0) < dog(1) | `[2]`       | `[1]`       | `[0,0]` |
|    5 | `dequeueDog()`   | Remove oldest dog               | `[2]`       | `[]`        | `[1,1]` |
|    6 | `dequeueCat()`   | Remove oldest cat               | `[]`        | `[]`        | `[2,0]` |


```python
from collections import deque
from typing import List
class AnimalShelf:
    def __init__(self):
        self.q =  [deque(), deque()]   

    def enqueue(self,animal):
        i,j = animal 
        self.q[j].append(i)  

    def dequeueAny(self):
        if not self.q[0] or (self.q[1] and self.q[1][0] < self.q[0][0]):
            return self.dequeueDog()
        return self.dequeueCat()
    def dequeueDog(self):
        return [-1,-1] if not self.q[1] else [self.q[1].popleft(),1]
    
    def dequeueCat(self):
        return [-1,-1] if not self.q[0] else [self.q[0].popleft(),1]
                    
```

```python
shelter = AnimalShelf()

shelter.enqueue([0, 0])  # Cat 0
shelter.enqueue([1, 0])  # Cat 1
shelter.enqueue([2, 1])  # Dog 2

print(shelter.dequeueDog())  # Output: [2, 1]  (dog 2 adopted)
print(shelter.dequeueCat())  # Output: [0, 0]  (cat 0 adopted)
print(shelter.dequeueAny())  # Output: [1, 0]  (remaining cat adopted)
print(shelter.dequeueAny())  # Output: [-1, -1] (no animals left)
```

# 20 - Valid Parentheses

s = "{[()]}"

| Step | Current Char | Action      | Stack Before    | Stack After     | Explanation            |
| ---: | ------------ | ----------- | --------------- | --------------- | ---------------------- |
|    1 | `{`          | push `{`    | `[]`            | `['{']`         | Opening bracket ‚Üí push |
|    2 | `[`          | push `[`    | `['{']`         | `['{','[']`     | Opening bracket ‚Üí push |
|    3 | `(`          | push `(`    | `['{','[']`     | `['{','[','(']` | Opening bracket ‚Üí push |
|    4 | `)`          | pop & match | `['{','[','(']` | `['{','[']`     | `(` matches `)`        |
|    5 | `]`          | pop & match | `['{','[']`     | `['{']`         | `[` matches `]`        |
|    6 | `}`          | pop & match | `['{']`         | `[]`            | `{` matches `}`        |         

    
| Condition           | Result     |
| ------------------- | ---------- |
| Stack empty at end? | ‚úÖ Yes      |
| Valid Parentheses?  | ‚úÖ **True** |


```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        mapping ={
            ')' : '(',
            '}' : '{',
            ']' : '['
        }

        for ch in s:
            if ch in mapping.values():
                stack.append(ch) #'{','[','('
            else:
                if not stack or stack[-1] != mapping[ch]:
                    return False
                stack.pop()
        return stack == []    

# üîπ Testing
obj = Solution()

test1 = "()"
test2 = "()[]{}"
test3 = "(]"
test4 = "([)]"
test5 = "{[]}"

print(test1, "->", obj.isValid(test1))
print(test2, "->", obj.isValid(test2))
print(test3, "->", obj.isValid(test3))
print(test4, "->", obj.isValid(test4))
print(test5, "->", obj.isValid(test5))        
```

# 739 - Daily Temperatures

temps = [30, 40, 35]  
index     0   1   2

| i (day) | temp | stack (before) | comparison  | action | prev_day | ans update       | stack (after) |
| ------- | ---- | -------------- | ----------- | ------ | -------- | ---------------- | ------------- |
| 0       | 30   | []             | stack empty | push 0 | ‚Äì        | ‚Äì                | [0]           |
| 1       | 40   | [0]            | 40 > 30     | pop    | 0        | ans[0] = 1-0 = 1 | []            |
| 1       | 40   | []             | ‚Äì           | push 1 | ‚Äì        | ‚Äì                | [1]           |
| 2       | 35   | [1]            | 35 < 40     | no pop | ‚Äì        | ‚Äì                | [1, 2]        |


```python
class Solution:
    def dailyTemperatures(self, temperatures):
        n = len(temperatures)#3
        ans = [0] * n #[0] *3=0,0,0
        stack = []# 1
        for i in range(n): #i = 0,1,2
            while stack and temperatures[i] > temperatures[stack[-1]]: #35 >40 F
                prev_day = stack.pop() #-0
                ans[prev_day] = i - prev_day #ans[0] = 1-0 =1
            stack.append(i)
        return ans 
obj = Solution()
temps = [30,40,35]
print(obj.dailyTemperatures(temps))           
```

# 84 - Largest Rectangle in Histogram


### Dry Run ‚Äì Largest Rectangle in Histogram
heights = [2, 1]   
sentinel add ‚Üí [2, 1, 0]   
index            0  1  2   


stack = []  
maxArea = 0  

Rules:-
Stack stores ‚Üí indices  
Stack keeps bars in increasing height order  
PUSH rule  

üëâ If stack is empty OR
üëâ current height ‚â• height at stack top   
‚û°Ô∏è PUSH index  

POP rule
üëâ If current height < height at stack top  
‚û°Ô∏è POP

| Step | i | heights[i] | Stack BEFORE | Condition   | Action     | h | w     | Area  | Stack AFTER | maxArea |
| ---- | - | ---------- | ------------ | ----------- | ---------- | - | ----- | ----- | ----------- | ------- |
| 1    | 0 | 2          | []           | stack empty | **PUSH 0** | ‚Äì | ‚Äì     | ‚Äì     | [0]         | 0       |
| 2    | 1 | 1          | [0]          | 1 < 2 ‚úÖ     | **POP 0**  | 2 | i = 1 | 2√ó1=2 | []          | 2       |
| 3    | 1 | 1          | []           | stack empty | **PUSH 1** | ‚Äì | ‚Äì     | ‚Äì     | [1]         | 2       |
| 4    | 2 | 0          | [1]          | 0 < 1 ‚úÖ     | **POP 1**  | 1 | i = 2 | 1√ó2=2 | []          | 2       |
| 5    | 2 | 0          | []           | stack empty | **PUSH 2** | ‚Äì | ‚Äì     | ‚Äì     | [2]         | 2       |



```python
class Solution:
    def largestRectangleArea(self,heights):
        stack = [] # store indices(0)
        maxArea = 0
        heights.append(0) #sentinel
        for i in range(len(heights)): #3=> i =0,1,2
            while stack and heights[i] < heights[stack[-1]]: #1<2
                h = heights[stack.pop()]
                #width
                if stack:
                    w = i - stack[-1] 
                else: 
                    w =i #1
                maxArea = max(maxArea, h * w) #2*1 =2
            stack.append(i) #push
        return maxArea  

heights = [2,1,0]
obj = Solution()
result = obj.largestRectangleArea(heights)
print("Final Answer:", result)                  
```

# 394 - Decode String (Great stack application)

s = "3[a]2[bc]"

###  Dry Run: Decode String "3[a]2[bc]"

| Char | Action        | curr_num | curr_str  | Stack (prev_str, num) | Explanation |
|------|---------------|----------|-----------|----------------------|-------------|
| Start| Initialize    | 0        | ""        | []                   | Initialize variables |
| '3'  | Digit         | 3        | ""        | []                   | curr_num = 0*10 + 3 = 3 |
| '['  | '['           | 0        | ""        | [("", 3)]            | Push (curr_str="", curr_num=3) to stack, reset curr_str & curr_num |
| 'a'  | Letter        | 0        | "a"       | [("", 3)]            | Append letter to curr_str |
| ']'  | ']'           | 0        | "aaa"     | []                   | Pop ("",3) from stack, curr_str = "" + "a"*3 = "aaa" |
| '2'  | Digit         | 2        | "aaa"     | []                   | curr_num = 0*10 + 2 = 2 |
| '['  | '['           | 0        | ""        | [("aaa",2)]          | Push (curr_str="aaa", curr_num=2) to stack, reset curr_str & curr_num |
| 'b'  | Letter        | 0        | "b"       | [("aaa",2)]          | Append letter to curr_str |
| 'c'  | Letter        | 0        | "bc"      | [("aaa",2)]          | Append letter to curr_str |
| ']'  | ']'           | 0        | "aaabcbc" | []                   | Pop ("aaa",2) from stack, curr_str = "aaa" + "bc"*2 = "aaabcbc" |
| End  | Return        | 0        | "aaabcbc" | []                   | Finished decoding, return curr_str |


```python
class Solution:
    def decodeString(self, s: str) -> str:
        stack = []
        curr_str = "" #"aaa"
        curr_num = 0 #3
        for ch in s: #3
            if ch.isdigit():
                curr_num = curr_num * 10 +int(ch) #0*10+3 =3, 0*10+2 =2
            elif ch == '[':
                #push number and curr str
                stack.append((curr_str,curr_num))    #("", 3), ("aaa", 2)
                curr_str = ""
                curr_num = 0
            elif ch == ']':
                prev_str, num = stack.pop()#pop("",3), pop("aaa", 2)
                curr_str = prev_str + curr_str * num # ""+ "a" * 3 ="aaa", "aaa"+"bc"*2= aaabcbc

            else:
                curr_str += ch # bc
        return curr_str   #"aaabcbc"          
s = "3[a]2[bc]"
obj = Solution()
print(obj.decodeString(s))
```
