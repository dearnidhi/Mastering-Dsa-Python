### 1  Factorial of a Number

n=4
| Step | Function Call | Returned Value           |
| ---- | ------------- | ------------------------ |
| 1    | factorial(4)  | 4 √ó factorial(3)         |
| 2    | factorial(3)  | 3 √ó factorial(2)         |
| 3    | factorial(2)  | 2 √ó factorial(1)         |
| 4    | factorial(1)  | 1 (base case)            |
| üîÅ   | Return path   | 4 √ó 3 √ó 2 √ó 1 = **24** ‚úÖ |


```python
class Solution:
    def factorial(self, n:int)->int:
        #Base Case
        if n==0 or n==1:
            return 1
        # Recursive case
        return n * self.factorial(n-1)
print(Solution().factorial(4))  # Output ‚Üí 24
print(Solution().factorial(0))  # Output ‚Üí   1  
```

### 2. #509. Fibonacci Number

n=4

fib(4)

= fib(3) + fib(2)

= (fib(2) + fib(1)) + (fib(1) + fib(0))

= ((fib(1) + fib(0)) + 1) + (1 + 0)

= ((1 + 0) + 1) + (1 + 0)

= 3 ‚úÖ


```python
class Solution:
    def fib(self, n: int) -> int:
        # Base cases
        if n == 0:
            return 0
        if n == 1:
            return 1
        
        # Recursive step
        return self.fib(n - 1) + self.fib(n - 2)
print(Solution().fib(4))  # Output ‚Üí 3
    
```

# Dry Run: Sum of First N Natural Numbers (Recursive)

**Example:** `n = 5`

| Function Call       | Expression                     | Returns | Explanation |
|--------------------|--------------------------------|---------|-------------|
| sumOfNumbers(5)    | 5 + sumOfNumbers(4)           | ?       | Recursive call with n-1 |
| sumOfNumbers(4)    | 4 + sumOfNumbers(3)           | ?       | Recursive call with n-1 |
| sumOfNumbers(3)    | 3 + sumOfNumbers(2)           | ?       | Recursive call with n-1 |
| sumOfNumbers(2)    | 2 + sumOfNumbers(1)           | ?       | Recursive call with n-1 |
| sumOfNumbers(1)    | 1                              | 1       | Base case reached |

---

### Combine Results (Unwinding Recursion)

| Function Call       | Calculation           | Result |
|--------------------|----------------------|--------|
| sumOfNumbers(2)    | 2 + 1                | 3      |
| sumOfNumbers(3)    | 3 + 3                | 6      |
| sumOfNumbers(4)    | 4 + 6                | 10     |
| sumOfNumbers(5)    | 5 + 10               | 15     |

‚úÖ **Final Output:** 15


```python
class Solution:
    def sumOfNumbers(self, n: int) -> int:
        # Base case
        if n == 1:
            return 1
        # Recursive case
        return n + self.sumOfNumbers(n - 1)

sol = Solution()
print(sol.sumOfNumbers(5))  # Output: 15
```

```python
class Solution:
    def sumOfNumbers(self, n: int) -> int:
        # Formula: sum = n * (n + 1) // 2
        return n * (n + 1) // 2

sol = Solution()
print(sol.sumOfNumbers(5))  # Output: 15
```

#  4Ô∏è‚É£ 344 ‚Äì Reverse String

### üîÑ Dry Run for s = ["h","e","l","l","o"]
üü¢ Initial Call

reverse(0, 4)  # l = 0, r = 4
| Level | Function Call   | Condition       | Action (Swap + Recursive Call)            | Resulting List          |
| :---- | :-------------- | :-------------- | :---------------------------------------- | :---------------------- |
| 1     | `reverse(0, 4)` | `l < r ‚Üí True`  | Swap `s[0] ‚Üî s[4]` ‚Üí call `reverse(1, 3)` | `["o","e","l","l","h"]` |
| 2     | `reverse(1, 3)` | `l < r ‚Üí True`  | Swap `s[1] ‚Üî s[3]` ‚Üí call `reverse(2, 2)` | `["o","l","l","e","h"]` |
| 3     | `reverse(2, 2)` | `l < r ‚Üí False` | **Base Case ‚Üí Stop recursion**            | `["o","l","l","e","h"]` |

## üåÄ Unwinding the Stack
Level 3 returns ‚Üí Level 2 returns ‚Üí Level 1 returns

‚úÖ **Final Result:** ["o","l","l","e","h"]

---

## üß† Key Insight
Each recursive call shifts the pointers **inward**:

- `l` increases by 1 ‚Üí moves right  
- `r` decreases by 1 ‚Üí moves left  

That‚Äôs why we use:  
`reverse(l + 1, r - 1)`



#### üî∏ Step 1: reverse(0, len(s) - 1)

When the function starts,

l (left) = 0 ‚Üí the first index

r (right) = len(s) - 1 ‚Üí the last index

üëâ This means we are swapping the outermost characters first.
After swapping, the recursive call moves inward to handle the middle characters.

```python
from typing import List

class Solution:
    def reverseString(self, s: List[str]) -> None:
        def reverse(l, r):
            if l < r:
                # Swap characters
                s[l], s[r] = s[r], s[l]
                # Recursive call
                reverse(l + 1, r - 1)
        
        reverse(0, len(s) - 1)
sol = Solution()
# Test 1
s1 = ["h","e","l","l","o"]
sol.reverseString(s1)
print(s1)  # ["o","l","l","e","h"]        
```

## Unwinding the Stack

| Level | Expression                        | Result  |
|-------|----------------------------------|---------|
| 5     | "o"                               | "o"     |
| 4     | reverseString("o") + "l" ‚Üí "o" + "l" | "ol"    |
| 3     | reverseString("lo") + "l" ‚Üí "ol" + "l" | "oll"   |
| 2     | reverseString("llo") + "e" ‚Üí "oll" + "e" | "olle"  |

‚úÖ **Final Result:** `"olleh"`

---

## Visual Recursion Tree





```python
class Solution:
    def reverseString(self, s: str) -> str:
        # Base case: empty string or single character
        if len(s) <= 1:
            return s
        # Recursive case: reverse the rest and append the first character at the end
        return self.reverseString(s[1:]) + s[0]

# Example
sol = Solution()
print(sol.reverseString("hello"))  # Output: "olleh"
```

#  5Ô∏è‚É£ 24 ‚Äì Swap Nodes in Pairs ‚úÖ (new recursive linked list problem)


üß† Problem Goal

We need to swap every two nodes.

Example:

Input:  1 ‚Üí 2 ‚Üí 3 ‚Üí 4

Output: 2 ‚Üí 1 ‚Üí 4 ‚Üí 3


We‚Äôre NOT allowed to change the values inside nodes ‚Äî
we can only change the pointers (who points to whom).

# üîç Dry Run Example

Let‚Äôs take:
singly linked list = [1, 2, 3, 4]

We‚Äôll see how recursion works step-by-step üëá

#### üßæ Step 1: First Call

Function call: swapPairs(1‚Üí2‚Üí3‚Üí4)

| Step   | Variable | Meaning     |
| ------ | -------- | ----------- |
| first  | 1        | first node  |
| second | 2        | second node |



üìå Swap planned: (1, 2)
Next recursive call: swapPairs(3‚Üí4)

After swapping pointers:



(We‚Äôll hold until recursion finishes for the rest)

#### üßæ Step 2: Second Call

Function call: swapPairs(3‚Üí4)

| Step   | Variable | Meaning     |
| ------ | -------- | ----------- |
| first  | 3        | first node  |
| second | 4        | second node |

üìå Swap planned: (3, 4)

Next recursive call: swapPairs(None) ‚Üí end of list



4 ‚Üí 3 ‚Üí None



#### üßæ Step 3: Third Call (Base Case)

Function call: swapPairs(None)

Since list is empty ‚Üí just return None.

#### üåÄ Step 4: Unwinding the Recursion

Now we go back upward:

| Level | Returned List    | Combined Result |
| ----- | ---------------- | --------------- |
| 3     | `None`           | ‚Äî               |
| 2     | `4‚Üí3‚ÜíNone`       | done            |
| 1     | `2‚Üí1‚Üí(4‚Üí3‚ÜíNone)` | `2‚Üí1‚Üí4‚Üí3` ‚úÖ     |


output => 2 ‚Üí 1 ‚Üí 4 ‚Üí 3

### üìä Visual Summary (Table)
| Level | Function Call      | Swap Pair | Recursive Call  | Returned List |
| ----- | ------------------ | --------- | --------------- | ------------- |
| 1     | swapPairs(1‚Üí2‚Üí3‚Üí4) | (1, 2)    | swapPairs(3‚Üí4)  | 2‚Üí1‚Üí4‚Üí3       |
| 2     | swapPairs(3‚Üí4)     | (3, 4)    | swapPairs(None) | 4‚Üí3           |
| 3     | swapPairs(None)    | ‚Äî         | ‚Äî               | None          |


```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val 
        self.next = next

class Solution:
    def swapPairs(self, head: ListNode) -> ListNode:
        # Base case: if there are 0 or 1 nodes, nothing to swap
        if not head or not head.next:
            return head
        
        # Nodes to be swapped
        first = head # 1 #3
        second = head.next #2 #4

        # Swap recursively for the rest of the list
        first.next = self.swapPairs(second.next) #1.next =swapPairs(3->4), 3.next=swapPairs(None)

        # Swap current two nodes
        second.next = first  #2.next=1 ,4.next=3

        # Return new head (second node)
        return second # 2 # 4
```
