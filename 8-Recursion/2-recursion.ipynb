### 1  Factorial of a Number

n=4
| Step | Function Call | Returned Value           |
| ---- | ------------- | ------------------------ |
| 1    | factorial(4)  | 4 Ã— factorial(3)         |
| 2    | factorial(3)  | 3 Ã— factorial(2)         |
| 3    | factorial(2)  | 2 Ã— factorial(1)         |
| 4    | factorial(1)  | 1 (base case)            |
| ðŸ”   | Return path   | 4 Ã— 3 Ã— 2 Ã— 1 = **24** âœ… |


```python
class Solution:
    def factorial(self, n: int) -> int:
        # Base Case
        if n == 0 or n == 1:
            return 1
        
        # Recursive Step
        return n * self.factorial(n - 1)
print(Solution().factorial(5))  # Output â†’ 120
print(Solution().factorial(0))  # Output â†’ 1
    
```

### 2. #509. Fibonacci Number

n=4

fib(4)

= fib(3) + fib(2)

= (fib(2) + fib(1)) + (fib(1) + fib(0))

= ((fib(1) + fib(0)) + 1) + (1 + 0)

= ((1 + 0) + 1) + (1 + 0)

= 3 âœ…


```python
class Solution:
    def fib(self, n: int) -> int:
        # Base cases
        if n == 0:
            return 0
        if n == 1:
            return 1
        
        # Recursive step
        return self.fib(n - 1) + self.fib(n - 2)
```
