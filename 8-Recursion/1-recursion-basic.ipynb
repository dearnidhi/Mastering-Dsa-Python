# Recursion ‚Äî The Concept
### üß† What is Recursion?

Recursion means:
üëâ A function calls itself until it reaches a base condition.

It breaks a big problem into smaller subproblems

### üîë 3 Parts of Every Recursive Function:

Base Case ‚Üí When to stop calling itself

Work / Processing ‚Üí What to do in the current step

Recursive Call ‚Üí Call the same function with a smaller input

### üîÅ Basic Structure of a Recursive Function

```python
def func():
    # 1Ô∏è‚É£ Base case ‚Äì when to stop
    if condition:
        return something
    
    # 2Ô∏è‚É£ Recursive case ‚Äì call itself
    else:
        return func(smaller_input)
```

### 1: Print Numbers from 1 to N

| **Step** | **Function Call**       | **Condition Check / Action**                      | **Output** |
| -------- | ----------------------- | ------------------------------------------------- | ---------- |
| 1Ô∏è‚É£      | `print_nums(3)`         | `n = 3`, `n == 0? ‚ùå` ‚Üí Call `print_nums(2)`       | ‚Äî          |
| 2Ô∏è‚É£      | `print_nums(2)`         | `n = 2`, `n == 0? ‚ùå` ‚Üí Call `print_nums(1)`       | ‚Äî          |
| 3Ô∏è‚É£      | `print_nums(1)`         | `n = 1`, `n == 0? ‚ùå` ‚Üí Call `print_nums(0)`       | ‚Äî          |
| 4Ô∏è‚É£      | `print_nums(0)`         | `n = 0`, `n == 0? ‚úÖ` ‚Üí **Return** (Base case hit) | ‚Äî          |
| 5Ô∏è‚É£      | Back to `print_nums(1)` | Execute `print(1)` ‚Üí **Output 1**                 | ‚úÖ 1        |
| 6Ô∏è‚É£      | Back to `print_nums(2)` | Execute `print(2)` ‚Üí **Output 2**                 | ‚úÖ 1 2      |
| 7Ô∏è‚É£      | Back to `print_nums(3)` | Execute `print(3)` ‚Üí **Output 3**                 | ‚úÖ 1 2 3    |


```python
def print_nums(n):
    if n == 0:        # base case
        return
    print_nums(n - 1) # recursive call
    print(n)          # after recursion (ascending order)

print_nums(3)
```

# 3 Factorial Function

4! = 4 √ó 3 √ó 2 √ó 1 = 24


- Imagine you are multiplying numbers from n down to 1.
- Each function waits for the next one to finish ‚Äî
- at the end, all results get multiplied together.

| Step | Function Call | Condition Check | What Happens        | Returns            |
| ---- | ------------- | --------------- | ------------------- | ------------------ |
| 1    | `fac(4)`      | 4 not in [0,1]  | return `4 * fac(3)` | waits for `fac(3)` |
| 2    | `fac(3)`      | 3 not in [0,1]  | return `3 * fac(2)` | waits for `fac(2)` |
| 3    | `fac(2)`      | 2 not in [0,1]  | return `2 * fac(1)` | waits for `fac(1)` |
| 4    | `fac(1)`      | ‚úÖ 1 in [0,1]    | return `1`          | gives `1`          |


Now the returns start going back upward ‚¨ÜÔ∏è

| Returning To | Calculation | Result |
| ------------ | ----------- | ------ |
| `fac(2)`     | `2 * 1`     | `2`    |
| `fac(3)`     | `3 * 2`     | `6`    |
| `fac(4)`     | `4 * 6`     | `24`   |


```python
def fac(n): #4 
    if n in [0,1]:
        return 1
    else:
        return n *fac(n-1) #4*fac(3),3*fac(2),2*fac(1),1
print(fac(4))    
```

- Recursive factorial.
- Base case: 0! = 1 or 1! = 1
- Recursive case: n! = n * (n-1)!

# 4: Fibonacci Sequence

The Fibonacci sequence adds the two previous numbers to get the next one.

1, 1, 2, 3, 5, 8, 13, 21, ...


fib(n) = fib(n - 1) + fib(n - 2)

#### üß© Fibonacci Series Breakdown

fib(0) = 1  
fib(1) = 1  
fib(2) = fib(1) + fib(0) = 1 + 1 = 2  
fib(3) = fib(2) + fib(1) = 2 + 1 = 3  
fib(4) = fib(3) + fib(2) = 3 + 2 = 5  
fib(5) = fib(4) + fib(3) = 5 + 3 = 8  
fib(6) = fib(5) + fib(4) = 8 + 5 = 13  
fib(7) = fib(6) + fib(5) = 13 + 8 = 21


- Calculates nth Fibonacci number.
- Base cases: fibonacci(0)=0, fibonacci(1)=1
- Recursive case: fibonacci(n) = fibonacci(n-1) + fibonacci(n-2)

Output: 13
* Sequence: 0, 1, 1, 2, 3, 5, 8, 13

| Step | Function Call    | What Happens                    | Returns |
| ---- | ---------------- | ------------------------------- | ------- |
| 1    | `fib(7)`         | returns `fib(6) + fib(5)`       | waits   |
| 2    | `fib(6)`         | returns `fib(5) + fib(4)`       | waits   |
| 3    | `fib(5)`         | returns `fib(4) + fib(3)`       | waits   |
| 4    | `fib(4)`         | returns `fib(3) + fib(2)`       | waits   |
| 5    | `fib(3)`         | returns `fib(2) + fib(1)`       | waits   |
| 6    | `fib(2)`         | returns `fib(1) + fib(0)`       | ‚úÖ `2`   |
| 7    | `fib(3)`         | `fib(2) + fib(1)` ‚Üí `2 + 1 = 3` |         |
| 8    | Keep going up... | Adds results step by step       |         |
| 9    | `fib(7)`         | = `13 + 8 = 21` ‚úÖ               |         |


```python
def fib(n): #7
    if n in [0,1]:
        return 1
    else:
        return fib(n-1)+fib(n-2) #
print(fib(5))    
 
```

### 5: Sum of Digits

Find the sum of digits of a number using recursion.

sum_of_digits(1234) ‚Üí 1 + 2 + 3 + 4 = 10

#### üî¢ Sum of Digits (Recursion Visualization)

Let‚Äôs take **n = 1234**

---

#### üß† Step-by-Step Table

| Step | `n` Value | `n % 10` (Last Digit) | `n // 10` (Remaining) | Recursive Call | Return Value / Action |
|------|------------|-----------------------|------------------------|----------------|------------------------|
| 1 | 1234 | 4 | 123 | `sum_of_digits(1234) = 4 + sum_of_digits(123)` | Going deeper ‚Üì |
| 2 | 123 | 3 | 12 | `sum_of_digits(123) = 3 + sum_of_digits(12)` | Going deeper ‚Üì |
| 3 | 12 | 2 | 1 | `sum_of_digits(12) = 2 + sum_of_digits(1)` | Going deeper ‚Üì |
| 4 | 1 | 1 | 0 | `sum_of_digits(1) = 1 + sum_of_digits(0)` | Going deeper ‚Üì |
| 5 (Base Case) | 0 | - | - | Return `0` | Base case reached ‚úÖ |

---

#### üîÅ Flow of Calls (Recursion Tree)



Now going back up (adding everything):


---

#### ü™ú Returning Back (Unwinding Recursion)

| Return From | Calculation | Result |
|--------------|--------------|---------|
| sum_of_digits(0) | `0` | 0 |
| sum_of_digits(1) | `1 + 0` | 1 |
| sum_of_digits(12) | `2 + 1` | 3 |
| sum_of_digits(123) | `3 + 3` | 6 |
| sum_of_digits(1234) | `4 + 6` | **10** |

---

#### ‚úÖ Final Answer:
`sum_of_digits(1234) = 10`


```python
def sum_of_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_of_digits(n // 10)
# Example
print(sum_of_digits(1234))
```

### 6: Reverse a String

#### üß© Example: reverse_str("abc")

| Step          | Function Call                  | What Happens                             | Return Value          |
| ------------- | ------------------------------ | ---------------------------------------- | --------------------- |
| 1             | `reverse_str("abc")`           | Peel off `'a'`, call `reverse_str("bc")` | *waits for next call* |
| 2             | `reverse_str("bc")`            | Peel off `'b'`, call `reverse_str("c")`  | *waits for next call* |
| 3             | `reverse_str("c")`             | Peel off `'c'`, call `reverse_str("")`   | *waits for next call* |
| 4 (Base Case) | `reverse_str("")`              | Empty string ‚Üí return `""`               | `""`                  |
| 5             | Return to `reverse_str("c")`   | `"" + "c"` ‚Üí `"c"`                       | `"c"`                 |
| 6             | Return to `reverse_str("bc")`  | `"c" + "b"` ‚Üí `"cb"`                     | `"cb"`                |
| 7             | Return to `reverse_str("abc")` | `"cb" + "a"` ‚Üí `"cba"`                   | `"cba"`               |


#### ü™ú ‚ÄúGoing Down‚Äù vs ‚ÄúComing Up‚Äù

| Phase         | Action                                      | Example                         |
| ------------- | ------------------------------------------- | ------------------------------- |
| Going Down ü°£ | Breaking the string (call recursion)        | `"abc"` ‚Üí `"bc"` ‚Üí `"c"` ‚Üí `""` |
| Coming Up ü°°  | Building the reversed string (return phase) | `"" ‚Üí "c" ‚Üí "cb" ‚Üí "cba"`       |


output=> reverse_str("abc") = "cba"


```python
def reverse_str(s): #s=abc,bc,c,""
    if len(s) == 0:
        return ""
    return reverse_str(s[1:]) + s[0]  #bc+a, c+b ,""+c
# Example
print(reverse_str("abc"))
```

Real-Life Analogy:

- Imagine stacking letters one by one:a ‚Üí b ‚Üí c
- When you flip the stack upside down, you get c, b, a ‚Üí "cba" üéØ
