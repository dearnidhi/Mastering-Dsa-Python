### üèóÔ∏è How to Create a Tuple
‚úÖ Note:

Tuples can be created using () or the built-in tuple() constructor.

Tuples are immutable (cannot be changed after creation).

```python
newTuple = ('a', 'b', 'c', 'd', 'e')
newTuple1 = tuple('abcde')

print(newTuple1)     # Output: ('a', 'b', 'c', 'd', 'e')
```

### üéØ Access Tuple Elements

```python
print(newTuple[0])   # Output: 'a'
```

Access elements using indexing ‚Äî indexing starts from 0.

### üîÅ Traversing Through a Tuple

You can loop over tuples directly or by index.

```python
for i in newTuple:
    print(i)               # Direct iteration

for index in range(len(newTuple)):
    print(newTuple[index]) # Using index range
```

### üîç Searching for an Element in a Tuple

```python
print('a' in newTuple)   # Output: True
```

### Custom Search Function

```python
def searchInTuple(pTuple, element):
    for i in pTuple:
        if i == element:
            return pTuple.index(i)
    return 'The element does not exist'

print(searchInTuple(newTuple, 'a'))  # Output: 0
```

### ‚öôÔ∏è Tuple Operations / Functions

```python
myTuple = (1,4,3,2,5)
myTuple1 = (1,2,6,9,8,7)

print(myTuple + myTuple1)   # Concatenation
print(myTuple * 4)          # Repetition
print(2 in myTuple1)        # Membership check

print(myTuple1.count(2))    # Count occurrences of element
print(myTuple1.index(2))    # Find index of element
```
