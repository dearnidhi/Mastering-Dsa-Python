#### ü™Ñ Step 3 ‚Äî Tuple Count Formula

| **Product** | **Count (c)** | **Check** | **Calculation**                                 | **Meaning**               | **Running Total** |
| ----------- | ------------- | --------- | ----------------------------------------------- | ------------------------- | ----------------- |
| 6           | 1             | ‚ùå `c < 2` | skip                                            | only one pair             | 0                 |
| 8           | 1             | ‚ùå `c < 2` | skip                                            | only one pair             | 0                 |
| 12          | 2             | ‚úÖ `c ‚â• 2` | `total += c √ó (c - 1) √ó 4`<br>`= 2 √ó 1 √ó 4 = 8` | 2 unique pairs ‚Üí 8 tuples | 8                 |
| 18          | 1             | ‚ùå `c < 2` | skip                                            | only one pair             | 8                 |
| 24          | 1             | ‚ùå `c < 2` | skip                                            | only one pair             | 8                 |


#### ‚ú® Why Formula = c * (c - 1) * 4?
| **Part**      | **Meaning**                                            | **Example**                                 |
| ------------- | ------------------------------------------------------ | ------------------------------------------- |
| `c * (c - 1)` | Ways to pick **2 different pairs** having same product | For `c=2` ‚Üí `(A,B)` or `(B,A)`              |
| `√ó 4`         | Each pair combo has 4 possible orderings               | (2,6,3,4), (2,6,4,3), (6,2,3,4), (6,2,4,3)‚Ä¶ |

‚úÖ Total Tuples = 8

üéØ Final Output:

total = 8

````python
from collections import defaultdict
def tupleSameProduct(nums):
    product_count = defaultdict(int)
    total = 0
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            product = nums[i] * nums[j]
            product_count[product]+=1
    for c in product_count.values():
        if c>=2:
            total+= c*(c-1)*4
    return total
                   
print(tupleSameProduct([2, 3, 4, 6]))
````

# 2Ô∏è‚É£ 454 ‚Äì 4Sum II


A = [1, 2]  
B = [-2, -1]  
C = [-1, 2]  
D = [0, 2]

#### üßÆ Step 1Ô∏è‚É£ ‚Äî Build ab_sum (All possible A + B sums)

| a | b  | sum (a+b) | Action                           | ab_sum (after step) |
| - | -- | --------- | -------------------------------- | ------------------- |
| 1 | -2 | -1        | Add new key `-1:1`               | {-1: 1}             |
| 1 | -1 | 0         | Add new key `0:1`                | {-1: 1, 0: 1}       |
| 2 | -2 | 0         | Key `0` exists ‚Üí increment count | {-1: 1, 0: 2}       |
| 2 | -1 | 1         | Add new key `1:1`                | {-1: 1, 0: 2, 1: 1} |

‚úÖ Final ab_sum ‚Üí { -1: 1, 0: 2, 1: 1 }

#### üîç Step 2Ô∏è‚É£ ‚Äî Check all C + D combinations and find target = -(c + d)

| c  | d | sum (c+d) | target = -(c+d) | Found in ab_sum? | Add Count | Total Count |
| -- | - | --------- | --------------- | ---------------- | --------- | ----------- |
| -1 | 0 | -1        | 1               | ‚úÖ Yes (1:1)      | +1        | 1           |
| -1 | 2 | 1         | -1              | ‚úÖ Yes (-1:1)     | +1        | 2           |
| 2  | 0 | 2         | -2              | ‚ùå No             | +0        | 2           |
| 2  | 2 | 4         | -4              | ‚ùå No             | +0        | 2           |

‚úÖ Final count = 2

üéØ Matching Tuples Explanation
* From A + B, the sum 1 exists (‚Üí comes from A=2, B=-1), From C + D, the sum -1 exists (‚Üí comes from C=-1, D=0) , one valid tuple is (2, -1, -1, 0)

* From A + B, the sum -1 exists (‚Üí from A=1, B=-2),From C + D, the sum 1 exists (‚Üí from C=-1, D=2), another valid tuple is (1, -2, -1, 2)

| Tuple (A,B,C,D) | Equation                | Result |
| --------------- | ----------------------- | ------ |
| (2, -1, -1, 0)  | 2 + (-1) + (-1) + 0 = 0 | ‚úÖ      |
| (1, -2, -1, 2)  | 1 + (-2) + (-1) + 2 = 0 | ‚úÖ      |

‚úÖ There are 2 valid tuples that make the sum = 0.



````python
from typing import List
class Solution:
    def fourSumCount(self, A:List[int], B:List[int], C:List[int], D:List[int])-> int:
        ab_sum = {} # sum of A[i]+B[j]
        count = 0
        # store sum of A & B
        for a in A:
            for b in B:
                s = a + b
                ab_sum[s] = ab_sum.get(s,0)+1 #

        for c in C:
            for d in D:
                target =-(c+d)
                if target in ab_sum:
                    count += ab_sum[target]  
        return count                  
    
# ‚úÖ Example
A = [1, 2]
B = [-2, -1]
C = [-1, 2]
D = [0, 2]

print(Solution().fourSumCount(A, B, C, D))  # Output: 2    
````

````python
from typing import List

class Solution:
    def fourSumCount(self, A: List[int], B: List[int], C: List[int], D: List[int]) -> int:
        ab_sum = {}  # store sums of A[i] + B[j]
        count = 0

        # Step 1Ô∏è‚É£: Store all possible sums of A and B
        for a in A:
            for b in B:
                s = a + b
                ab_sum[s] = ab_sum.get(s, 0) + 1  # count frequency

        # Step 2Ô∏è‚É£: For every pair in C and D, look for -(c + d)
        for c in C:
            for d in D:
                target = -(c + d)
                if target in ab_sum:
                    count += ab_sum[target]  # add matching pairs

        return count


# ‚úÖ Example
A = [1, 2]
B = [-2, -1]
C = [-1, 2]
D = [0, 2]

print(Solution().fourSumCount(A, B, C, D))  # Output: 2
````

# 3Ô∏è‚É£ 447 ‚Äì Number of Boomerangs



## üß© Problem Statement
You are given `n` points in the plane that are all distinct, where  
`points[i] = [xi, yi]`.

A **boomerang** is a tuple of points `(i, j, k)` such that the **distance** between `i` and `j` equals the **distance** between `i` and `k`.  
*(The order of the tuple matters.)*

Return the **number of boomerangs**.

| Variable | Value                 |
| -------- | --------------------- |
| points   | `[[0,0],[1,0],[2,0]]` |
| total    | `0`                   |


| **Point i** | **Point j** | **Check `i==j`** | **Distance Calculation=** | **Distance (d)** | **dist Dictionary** | **Values** | **Calculation** = val * (val - 1)   | **Current Total** |
| ----------- | ----------- | ---------------- | ------------------------ | ---------------- | ------------------- | ---------- | ----------------- | ----------------- |
| [0,0]       | [0,0]       | ‚úÖ Skip           | -                        | -                | {}                  | -          | -                 | 0                 |
| [0,0]       | [1,0]       | ‚ùå                | (0‚àí1)¬≤ + (0‚àí0)¬≤ = 1      | 1                | {1:1}               | -          | -                 | 0                 |
| [0,0]       | [2,0]       | ‚ùå                | (0‚àí2)¬≤ + (0‚àí0)¬≤ = 4      | 4                | {1:1, 4:1}          | [1,1]      | 1√ó(1‚àí1)+1√ó(1‚àí1)=0 | 0                 |
| [1,0]       | [0,0]       | ‚ùå                | (1‚àí0)¬≤ + (0‚àí0)¬≤ = 1      | 1                | {1:1}               | -          | -                 | 0                 |
| [1,0]       | [1,0]       | ‚úÖ Skip           | -                        | -                | {1:1}               | -          | -                 | 0                 |
| [1,0]       | [2,0]       | ‚ùå                | (1‚àí2)¬≤ + (0‚àí0)¬≤ = 1      | 1                | {1:2}               | [2]        | 2√ó(2‚àí1)=2         | 2                 |
| [2,0]       | [0,0]       | ‚ùå                | (2‚àí0)¬≤ + (0‚àí0)¬≤ = 4      | 4                | {4:1}               | -          | -                 | 2                 |
| [2,0]       | [1,0]       | ‚ùå                | (2‚àí1)¬≤ + (0‚àí0)¬≤ = 1      | 1                | {4:1, 1:1}          | [1,1]      | 1√ó(1‚àí1)+1√ó(1‚àí1)=0 | 2                 |
| [2,0]       | [2,0]       | ‚úÖ Skip           | -                        | -                | {4:1, 1:1}          | -          | -                 | 2                 |


### Python Explanation

```python
dist[d] = dist.get(d, 0) + 1


dist = {}  # Empty dictionary
- **`dist.get(d, 0)`** ‚Üí Get the value of key `d` from the dictionary.  
  If the key doesn‚Äôt exist, take **0** as the default value.

- **`+ 1`** ‚Üí Add 1 to that value.

- **`dist[d] =`** ‚Üí Store the updated value back into the dictionary.

---

### Example

**Before:**  
`dist = {}` (empty dictionary)

**Step 1:**  
`d = 1`  
`dist.get(1, 0)` ‚Üí returns `0`  
`0 + 1 = 1`  
‚úÖ `dist = {1: 1}`

**Step 2:**  
`d = 1` again  
`dist.get(1, 0)` ‚Üí returns `1`  
`1 + 1 = 2`  
‚úÖ `dist = {1: 2}`

---

### In simple words

- If a **new distance** appears ‚Üí start its count at **1**  
- If the **same distance** appears again ‚Üí **increase its count by 1**


````python
class Solution:
    def numberOfBoomerangs(self,points):
        total = 0
        for i in points:
            dist = {}
            for j in points:
                if i == j:
                    continue
                d = (i[0]-j[0])**2 + (i[1]-j[1])**2 
                dist[d] = dist.get(d,0)+1
            for val in dist.values():
                total += val*(val-1)
        return total 
      
sol = Solution()    
print(sol.numberOfBoomerangs([[0,0],[1,0],[2,0]]))  # 2
print(sol.numberOfBoomerangs([[0,0],[1,0],[0,1]]))  # 2  
print(sol.numberOfBoomerangs([[0,0],[1,0]]))        # 0           
````

# 4Ô∏è‚É£ 1577 ‚Äì Number of Ways Where Square of Number Equals Product of Two Numbers

a¬≤ = b √ó c

````python
from collections import defaultdict

class Solution:
    def numTriplets(self, nums1, nums2):
        def count_triplets(A, B):
            count = 0                                    # Line 1: Result counter
            for a in A:                                  # Line 2: For each number in first array
                target = a * a                           # Line 3: Calculate square of current number
                freq = defaultdict(int)                  # Line 4: Create empty frequency dictionary
                
                for b in B:                              # Line 5: For each number in second array
                    if target % b == 0:                  # Line 6: Check if target is divisible by current b
                        needed = target // b             # Line 7: Calculate the number needed to complete the pair
                        count += freq[needed]            # Line 8: Add count of previously seen 'needed' numbers
                    freq[b] += 1                         # Line 9: Store current b in frequency map
            return count                                 # Line 10: Return count for this combination
        
        # Line 11: Calculate both types and return total
        return count_triplets(nums1, nums2) + count_triplets(nums2, nums1) #Line 11

# Test
nums1 = [7, 4]
nums2 = [5, 2, 8, 9]
print(Solution().numTriplets(nums1, nums2))  # Output: 1
````
