#### ‚ú® Why Formula = c * (c - 1) * 4?
| **Part**      | **Meaning**                                            | **Example**                                 |
| ------------- | ------------------------------------------------------ | ------------------------------------------- |
| `c * (c - 1)` | Ways to pick **2 different pairs** having same product | For `c=2` ‚Üí `(A,B)` or `(B,A)`              |
| `√ó 4`         | Each pair combo has 4 possible orderings               | (2,6,3,4), (2,6,4,3), (6,2,3,4), (6,2,4,3)‚Ä¶ |

‚úÖ Total Tuples = 8

üéØ Final Output:

total = 8

```python
from collections import defaultdict
def tupleSameProduct(nums):
    product_count = defaultdict(int)
    total = 0
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            product = nums[i] * nums[j]
            product_count[product]+=1
    for c in product_count.values():
        if c>=2:
            total+= c*(c-1)*4
    return total
                   
print(tupleSameProduct([2, 3, 4, 6]))
```
