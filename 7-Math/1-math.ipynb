````python
data = [
    {"name": "Alice", "age": 25, "score": 78},
    {"name": "Bob", "age": 30, "score": 45},
    {"name": "Charlie", "age": 22, "score": 92},
    {"name": "David", "age": 28, "score": 88}
]

def get_high_scores(data, threshold):
    """
    Returns a list of 'score' values from dictionaries that exceed the threshold.
   
    Args:
        data (list): List of dictionaries.
        threshold (int): Threshold value.
       
    Returns:
        list: List of scores above the threshold.
    """
    return [item["score"] for item in data if item["score"] > threshold]


# Example usage
result = get_high_scores(data, 80)
print(result)  # Output: [92, 88]
````

## 7 – Reverse Integer

### Step 1: Initialization
| Variable     | Value            |
| ------------ | ---------------- |
| `x`          | -123             |
| `MAX`        | 2147483647       |
| `sign`       | -1 (since x < 0) |
| `x = abs(x)` | 123              |
| `rev`        | 0                |


### 🔹 MAX = 2**31 - 1
It represents the maximum 32-bit integer value (2147483647).  
✅ Used to check if the reversed number goes beyond this limit (overflow).


### 🧮 Step 2: Loop Execution

### 🔁 Iteration 1
- **x = 123**, **rev = 0**  
- rev = rev * 10 + x % 10 = 0 * 10 + 3 = **3**  
- x = x // 10 = **12**  
✅ After iteration: rev = **3**, x = **12**

---

### 🔁 Iteration 2
- **rev = 3**, **x = 12**  
- rev = rev * 10 + x % 10 = 3 * 10 + 2 = **32**  
- x = x // 10 = **1**  
✅ After iteration: rev = **32**, x = **1**

---

### 🔁 Iteration 3
- **rev = 32**, **x = 1**  
- rev = rev * 10 + x % 10 = 32 * 10 + 1 = **321**  
- x = x // 10 = **0**  
✅ After iteration: rev = **321**, x = **0**

---

### 🧾 Step 3: Loop Ends
- Now `x = 0`, so loop stops.  
- `rev = 321`, but remember `sign = -1`.

---

### 🧩 Step 4: Final Return
```python
return rev * sign

👉 rev * sign = 321 * (-1) = -321


| Step | x   | rev | x % 10 | rev * 10 + (x % 10) | x // 10 |
| ---- | --- | --- | ------ | ------------------- | ------- |
| Init | 123 | 0   | -      | -                   | -       |
| 1    | 123 | 0   | 3      | 3                   | 12      |
| 2    | 12  | 3   | 2      | 32                  | 1       |
| 3    | 1   | 32  | 1      | 321                 | 0       |


➡️ Then return 321 * (-1) = -321

````python
class Solution:
    def reverse(self, x:int)-> int:
        MAX = 2**31 -1
        sign = -1 if x < 0 else 1
        x = abs(x)
        rev = 0

        while x:
            if rev >MAX//10:
                return
            rev = (rev * 10 + x % 10)
            x//=10
        return rev * sign        
                   
#Test cases
sol = Solution()
print(sol.reverse(-123))     # Output: 321  
print(sol.reverse(123))     # Output: 321
print(sol.reverse(-456))    # Output: -654
print(sol.reverse(120))     # Output: 21
print(sol.reverse(0))       # Output: 0  
````

#  9 – Palindrome Number


# 🧮 Dry Run — Palindrome Number

**Input:** `x = 1221`

### 🧠 Key Idea
We reverse only **half** of the number (rev_half).  
Then we check:
- **Even digits:** `x == rev_half`
- **Odd digits:** `x == rev_half // 10` → (remove middle digit)

---

## ✅ Test Case 1: x = 1221 (Even Digits)
| Step  | x (original) | remainder = x % 10 | rev_half (before) | rev_half (after) = rev_half * 10 + x % 10 | x //= 10 (new x) |
| ----- | ------------ | ------------------ | ----------------- | ----------------------------------------- | ---------------- |
| Start | 1221         | —                  | 0                 | —                                         | —                |
| 1     | 1221         | 1221%10=1                  | 0                 | 0×10 + 1 = **1**                          | 1221//10=122              |
| 2     | 122          | 122%10=2                  | 1                 | 1×10 + 2 = **12**                         | 122//10=12               |
| 3     | 12           | —                  | 12                | — (loop stops, since x == rev_half)       | —                |


### 🧮 Final Check
x == rev_half → 12 == 12 ✅

✅ Output → **True (Palindrome)**

---
rev_half = rev_half * 10 + x % 10   # ← here (x % 10) is remainder = last digit

x //= 10                            # ← removes that last digit from x

## ✅ Test Case 2: x = 12321 (Odd Digits)
| Step | x (original) | remainder = x % 10 | rev_half (before) | rev_half (after)=rev_half * 10 + x % 10  | x //= 10 (new x) |
| ---- | ------------ | ------------------ | ----------------- | ----------------- | ---------------- |
| 1️⃣  | 12321        | 12321 %10=1                  | 0                 | 0 * 10 + 1 = 1    | 12321//10=1232             |
| 2️⃣  | 1232         | 1232%10=2                  | 1                 | 1 * 10 + 2 = 12   | 1232//10=123              |
| 3️⃣  | 123          | 123%10=3                  | 12                | 12 * 10 + 3 = 123 | 123//10=12               |


### 🧮 Final Check
After loop end:

x = 12

rev_half = 123

| Check                 | Explanation                                                                  | Result  |
| --------------------- | ---------------------------------------------------------------------------- | ------- |
| `x == rev_half`       | `12 == 123`                                                                  | ❌ False |
| `x == rev_half // 10` | remove Middle digit (3) & compare → `12 == 123 // 10` → `12 == 12` | ✅ True  |


* x % 10 → We divide x by 10 and take the remainder — this gives the last digit
* x //= 10 → We divide x by 10 and take only the integer part — this removes the last digit
* rev_half → gradually builds reverse of the number’s right half

| Expression | What it does           | Example (x=123) | Result |
| ---------- | ---------------------- | --------------- | ------ |
| `x % 10`   | gives **last digit**   | 123 % 10        | 3      |
| `x // 10`  | removes **last digit** | 123 // 10       | 12     |


````python

class Solution:
    def isPalindrome(self, x:int)->bool:
        if x < 0 or (x % 10 == 0 and x != 0 ):
            return False
        rev_half = 0
        while x > rev_half : #1221 >0
            rev_half = rev_half * 10 + x % 10 # 0*10+ 1221%10 = 0+1=1
            x//=10 #1221//100=122
        return x ==rev_half or x == rev_half//10    
# examples
print(Solution().isPalindrome(1221))   # True
print(Solution().isPalindrome(12321))  # True
print(Solution().isPalindrome(-121))   # False
print(Solution().isPalindrome(10))     # False    
            
        
````

# 50 – Pow(x, n)


#### for Positive power

x = 2, n =5=(power), result = 1

| Step | n=n//2 | x (base) | n % 2    | Action              | result | Comment        |
| ---- | - | -------- | -------- | ------------------- | ------ | -------------- |
|1(initial)    | 5 | 2        | 5%2=1 (odd)  | result *= x → 1×2=2 | 2      | Multiply once  |
| 2    | 5//2=2 | 2²=4     | 0 (even) | skip multiply       | 2      | Just square    |
| 3    | 2//2=1 | 4²=16    | 1 (odd)  | result *= x → 2×4=8 | 8      | Multiply again |
| 4    | 0 | —        | —        | stop                | ✅      |                |
✅ Final result = 8

2^5=8
---
#### for Negative power

x = 2, n = -3

x = 1 / x   →  x = 1 / 2

n = -n      →  n = 3

| Step | n=n//2 | x          | n % 2   | Action                                   | result |
| ---- | - | ---------- | ------- | ---------------------------------------- | ------ |
| 1(initial)    | 3 | 0.5        | 1 (odd) | result = result * x → 1 * 0.5 = 0.5      | 0.5    |
| 2    | 1 | 0.25 (x*x) | 1 (odd) | result = result * x → 0.5 * 0.25 = 0.125 | 0.125  |
| 3    | 0 | —          | —       | Stop                                     | ✅      |




````python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        if n == 0:
            return 1
        if n < 0:
            x = 1/x
            n = -n
        result = 1
        while n > 0: 
            if n % 2 == 1:
                result *= x
            x *= x 
            n //= 2
        return result             
# 👇 Example calls (prints the output)
print(Solution().myPow(2, 10))   # 1024.0
print(Solution().myPow(2, -3))   # 0.125
print(Solution().myPow(3, 0))    # 1
print(Solution().myPow(5, 2))    # 25             
````

#  13 – Roman to Integer


"MCMXCIV" → 1994

| Step    | ch | value | prev_value | Check        | Operation          | total (after) | prev_value (after) |
| ------- | -- | ----- | ---------- | ------------ | ------------------ | ------------- | ------------------ |
| 1       | V  | 5     | 0          | 5 < 0 ❌      | total = 0 + 5      | 5             | 5                  |
| 2       | I  | 1     | 5          | 1 < 5 ✅      | total = 5 - 1      | 4             | 1                  |
| 3       | C  | 100   | 1          | 100 < 1 ❌    | total = 4 + 100    | 104           | 100                |
| 4       | X  | 10    | 100        | 10 < 100 ✅   | total = 104 - 10   | 94            | 10                 |
| 5       | M  | 1000  | 10         | 1000 < 10 ❌  | total = 94 + 1000  | 1094          | 1000               |
| 6       | C  | 100   | 1000       | 100 < 1000 ✅ | total = 1094 - 100 | 994           | 100                |
| 7       | M  | 1000  | 100        | 1000 < 100 ❌ | total = 994 + 1000 | 1994          | 1000               |
| ✅ Final | —  | —     | —          | —            | —                  | **1994**      | —                  |


````python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman_map = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50,
            'C': 100, 'D': 500, 'M': 1000
        }
        total = 0
        prev_value = 0
        for ch in reversed(s):#S= MCMXCIV, ch =v
            value = roman_map[ch] # ch =v =5
            if value < prev_value: #5<0
                total -= value
            else:
                total += value #  0 + 5 =5
            prev_value = value #5
        return total               
# 🧠 Example runs
print(Solution().romanToInt("III"))     # 3
print(Solution().romanToInt("IV"))      # 4
print(Solution().romanToInt("IX"))      # 9
print(Solution().romanToInt("LVIII"))   # 58
print(Solution().romanToInt("MCMXCIV")) # 1994
````

#  12 – Integer to Roman

num = 58
| Step | Symbol | Value | num (before) | num >= value? | Action    | Roman (res) | num (after) |
| ---- | ------ | ----- | ------------ | ------------- | --------- | ----------- | ----------- |
| 1    | M      | 1000  | 58           | ❌             | Skip      | —           | 58          |
| 2    | D      | 500   | 58           | ❌             | Skip      | —           | 58          |
| 3    | C      | 100   | 58           | ❌             | Skip      | —           | 58          |
| 4    | L      | 50    | 58           | ✅             | Add "L"   | L           | 8           |
| 5    | V      | 5     | 8            | ✅             | Add "V"   | LV          | 3           |
| 6    | I      | 1     | 3            | ✅ (3 times)   | Add "III" | LVIII       | 0           |


| Type | Rule                                                                   | Example        |
| ---- | ---------------------------------------------------------------------- | -------------- |
| 1️⃣  | Numbers are written from **left to right**                             | MCMXCIV (1994) |
| 2️⃣  | If a **smaller number comes before** a larger number → **subtract it** | IV = 5 - 1 = 4 |
| 3️⃣  | If a **smaller number comes after** a larger number → **add it**       | VI = 5 + 1 = 6 |


````python
class Solution:
    def intToRoman(self, num: int) -> str:
        roman_map = [
            ("M", 1000), ("CM", 900), ("D", 500), ("CD", 400),
            ("C", 100), ("XC", 90), ("L", 50), ("XL", 40),
            ("X", 10), ("IX", 9), ("V", 5), ("IV", 4), ("I", 1)
        ]
        res =" "
        for symbol, value in roman_map:
            while num >= value:
                res += symbol
                num -= value
        return res                   
print(Solution().intToRoman(58)) # 1994
````
