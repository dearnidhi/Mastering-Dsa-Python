# - 1929 – Concatenation of Array  

| Iteration | i | nums[i] | ans[i] after update | ans[i + n] after update | ans (current state) |
| --------- | - | ------- | ------------------- | ----------------------- | ------------------- |
| Start     | - | -       | -                   | -                       | [0, 0, 0, 0, 0, 0]  |
| 1         | 0 | 1       | ans[0] = 1          | ans[3] = 1              | [1, 0, 0, 1, 0, 0]  |
| 2         | 1 | 2       | ans[1] = 2          | ans[4] = 2              | [1, 2, 0, 1, 2, 0]  |
| 3         | 2 | 1       | ans[2] = 1          | ans[5] = 1              | [1, 2, 1, 1, 2, 1]  |


```python
class Solution(object):
    def getConcatenation(self, nums):
        n = len(nums)
        ans = [0] * (2 * n)   # naya array bana diya size 2n
        
        for i in range(n):
            ans[i] = nums[i]
            ans[i + n] = nums[i]
        
        return ans
nums = [1,2,1]
print(Solution().getConcatenation(nums))        
```

# - 1512 – Number of Good Pairs 

We need to count pairs (i, j) such that:

nums[i] == nums[j]

i < j

nums = [1, 2, 3, 1, 1, 3]   
Initial State:    

freq = {}    
count = 0

| Step | Current num | freq before     | Action Taken         | count after | freq after      |
| ---- | ----------- | --------------- | -------------------- | ----------- | --------------- |
| 1    | 1           | {}              | Not in freq → add 1  | 0           | {1:1}           |
| 2    | 2           | {1:1}           | Not in freq → add 2  | 0           | {1:1, 2:1}      |
| 3    | 3           | {1:1, 2:1}      | Not in freq → add 3  | 0           | {1:1, 2:1, 3:1} |
| 4    | 1           | {1:1, 2:1, 3:1} | In freq → count += 1 | 1           | {1:2, 2:1, 3:1} |
| 5    | 1           | {1:2, 2:1, 3:1} | In freq → count += 2 | 3           | {1:3, 2:1, 3:1} |
| 6    | 3           | {1:3, 2:1, 3:1} | In freq → count += 1 | 4           | {1:3, 2:1, 3:2} |


```python
class Solution(object):
    def numIdenticalPairs(self, nums):
        freq = {}
        count = 0
        
        for num in nums:
            if num in freq:
                count += freq[num]
                freq[num] += 1
            else:
                freq[num] = 1
        
        return count
nums = [1, 2, 3, 1, 1, 3]   
print(Solution().numIdenticalPairs(nums))
```

# - 2011 – Final Value of Variable After Performing Operations  

You are given an array of strings like:

["--X","X++","X++"]


Each operation modifies a variable X starting from 0.

Rules:

"++X" or "X++" → increment X by 1

"--X" or "X--" → decrement X by 1

Return the final value of X.

| Step  | Current Operation | '+' in op? | Action     | x after step |
| ----- | ----------------- | ---------- | ---------- | ------------ |
| Start | -                 | -          | Initialize | 0            |
| 1     | "--X"             | No         | x = x - 1  | -1           |
| 2     | "X++"             | Yes        | x = x + 1  | 0            |
| 3     | "X++"             | Yes        | x = x + 1  | 1            |


```python
class Solution(object):
    def finalValueAfterOperations(self, operations):
        x = 0
        
        for op in operations:
            if '+' in op:
                x += 1
            else:
                x -= 1
        
        return x
print(Solution().finalValueAfterOperations("++"))
```
