### ‚úÖ What is an Array?
An array is a collection of elements (of same data type) stored at contiguous memory locations.

```python
arr = [10, 20, 30, 40]  # Python array (list)
```

# ‚úÖ Why Use Arrays?

Easy to access elements by index.

Efficient for storing multiple values.

Useful in building complex data structures (stacks, queues, etc.).

# ‚úÖ Properties of Array (with Code)

* Fixed Size (in low-level languages)
Python lists grow dynamically, but in C/C++, array size is fixed.

* Same Data Type
Python allows mixed types, but in DSA logic, we assume same types.

* Index-based Access 

```python
arr = [5, 10, 15, 20]
print(arr[2])  # Output: 15
```

```python
print(arr[3])  # Output: 20
```

* Contiguous Memory Allocation (C/C++ specific)

## üîç What is Traversal in Array?
Traversal means visiting each element of an array one by one, usually to:

Print

Search

Modify

Apply operations (like sort, sum, etc.)

üìå Examples

```python
arr = [10, 20, 30, 40, 50]
# Output: 10 20 30 40 50
```

# üìÇ Types of Array Traversal
# ‚úÖ 1. Linear Traversal

‚û°Ô∏è Visiting elements from start to end.

```python
aar = [1,2,3,4,5]
print("Linear Traversal:", end = " ")
for i in aar:
    print(i, end = " ")
```

# üîÅ 2. Reverse Traversal

‚û°Ô∏è Visiting elements from end to start.

range(start, stop, step)

start ‚Üí where the loop begins

stop ‚Üí where the loop stops (exclusive)

step ‚Üí increment or decrement

üí° Here:

start = len(arr) - 1 ‚Üí last index of the array

stop = -1 ‚Üí because Python‚Äôs range excludes the stop value, this makes the loop go down to index 0

step = -1 ‚Üí means we are moving backward

```python
arr = [1,2,3,4,5]
print("Reverse Traversal:", end = " ")
for i in range(len(arr)-1,-1,-1): # range(5-1, -1 ,-1) = range(4, -1, -1)
    print(arr[i], end = " ")
```

# üîÑ Methods to Traverse an Array


# üåÄ 1. Using For Loop


```python
arr = [10, 20, 30, 40, 50]
print("Traverse unsing for loop:", end = " ")
for i in range(len(arr)):
    print(arr[i], end = " ")
```

# üîÅ 2. Using While Loop

```python
arr = [10, 20, 30, 40, 50]
i = 0
while i < len(arr):  # 0<5, 1<5 ,2<5, 3<5 4<5
    print(arr[i], end = " ") #arr[0] =10, arr[1] =20, arr[2]= 30  arr[3]= 40, arr[4]= 50
    i+=1
```

# üßæ 3. Using Foreach Loop (in Python: for value in arr)

```python

arr = [10,20,30,40,50]
for value in arr:
    print(value, end =" ")
```

# üõ†Ô∏è Inserting Elements in an Array


# ‚úÖ 1. Inserting at the Beginning


```python

arr = [10,20,30,40,50]
ele = 50 
arr.insert(0,ele)
print("After insertion:", arr)
```

# üéØ 2. Inserting at a Given Position

```python

arr = [10,20,30,40,50]
pos = 2
ele = 50 
arr.insert(pos-1,ele)  #(2-1,50) = (1,50)
print("After insertion at given position:", arr)

```

# üöÄ 3. Inserting at the End

```python
arr = [10,20,30,40,50]
ele = 50 
arr.append(ele)  #(2-1,50) = (1,50)
print("After append ele at end:", arr)

```

# 4. Searching in Array



```python
# üîç Linear Search in Unsorted Array
def FindElements(arr, n, key):
    for i in range(n):
        if arr[i] == key: 
            return i
arr = [12, 34,10,6,40]
key = 40
n= len(arr)
index = FindElements(arr, n, key)
if index != -1:
    print("element found at index:", index)
else:
    print("element not found")    


```

# 3. Deletion in Array
