# 1. Missing Number


# Dry Run

Array: [3, 0, 1]

Length n = 3 

total_sum = 3*(3+1)/2 = 6 = n*(n+1)//2

Sum of array = 3+0+1 = 4 = sum_arr

Missing = 6 - 4 = 2 ✅-> total_sum - sum_arr

```python
from typing import List

class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        n = len(nums)
        total = n * (n + 1) // 2
        current_sum = 0
        
        for i in range(len(nums)):
            current_sum += nums[i]
        
        return total - current_sum


arr = [3, 0, 1]
print(Solution().missingNumber(arr))
```

| ⏱️ **Time Complexity**            | 🧠 **Space Complexity**                               |
| --------------------------------- | ----------------------------------------------------- |
| Compute total → O(1)              | Only integers `n`, `total`, `current_sum`, `i` → O(1) |
| Loop over `nums` → O(n)           | No extra array/list used                              |
| Addition per element → O(1)       | Return value → O(1)                                   |
| **✅ Final Time Complexity:** O(n) | **✅ Total Space Complexity:** O(1)                    |


```python
from typing import List
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        n = len(nums) #3
        total = n*(n+1)//2 # 6
        arr = sum(nums) # 4
        missing = total - arr
        return  missing

arr = [3,0,1]
print(Solution().missingNumber(arr))      
```

# 2.Two Sum


## 🔹 Step-by-Step Dry Run

### Outer Loop → i = 0 (nums[0] = 2)
Inner loop starts from j = i+1 = 1.

Input: nums = [2,7,11,15], target = 9


| i | j | nums[i] | nums[j] | nums[i] + nums[j] | Target | Match? | Result     |
|---|---|---------|---------|-------------------|--------|--------|------------|
| 0 | 1 | 2       | 7       | 2 + 7 = 9         | 9      | ✅ Yes  | Return [0,1] |

---

## 🔹 Visual Flow (Narration for Video)

1. Start: `nums = [2,7,11,15]`, `target = 9`.  
2. Outer loop `i = 0` → `nums[i] = 2`.  
3. Inner loop `j = 1` → `nums[j] = 7`.  
4. Check: `nums[i] + nums[j] = 2 + 7 = 9`.  
5. Condition satisfied ✅ → return indices `[0, 1]`.  
6. Execution stops here.


```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        n = len(nums) #4
        for i in range(n):
            for j in range(i+1, n):
                if nums[i] + nums[j] == target:
                    return [i,j]
                
sol = Solution()
print(sol.twoSum([2,7,11,15], 9))   # Output: [0, 1]     
```

| ⏱️ **Time Complexity**                         | 🧠 **Space Complexity**                 |
| ---------------------------------------------- | --------------------------------------- |
| Outer loop: i in 0..n-1 → O(n)                 | Only loop variables `i,j` → O(1)        |
| Inner loop: j in i+1..n → O(n-i) → total O(n²) | No extra list used except return → O(1) |
| Check sum `nums[i]+nums[j]` → O(1)             | Return `[i,j]` → O(2) = O(1)            |
| **✅ Final Time Complexity:** O(n²)             | **✅ Total Space Complexity:** O(1)      |


# 3. Best Time to Buy and Sell Stock (LeetCode 121)

# Profit Calculation
Profit = Sell Price - Buy Price

---

## 🔹 Dry Run: Brute-Force Method (Example: [7,1,5])

**Example:** [7, 1, 5]

**Prices:**  7   1   5  
**Days:**    1   2   3

### Step by Step Check (Buy → Sell)

**Buy Day 1 (Price 7)**  
- Sell Day 2 (Price 1) → Profit = 1 - 7 = -6 ❌ (loss)  
- Sell Day 3 (Price 5) → Profit = 5 - 7 = -2 ❌ (loss)  

**Buy Day 2 (Price 1)**  
- Sell Day 3 (Price 5) → Profit = 5 - 1 = 4 ✅ (profit)  

---

✅ **Final Max Profit = 4** (Buy at 1, Sell at 5)


```python

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        max_profit = 0
        n = len(prices)
        for i in range(n): #Buy day => price[i=0] => 7
            for j in range(i+1, n): # sell day => prices[i+1] =prices[0+1]= 1
                profit = prices[j] - prices[i]
                if profit > max_profit:
                    max_profit = profit
        return max_profit            

# Example test
sol = Solution()
print(sol.maxProfit([7,1,5]))  # Output: 4
```

| ⏱️ **Time Complexity**                 | 🧠 **Space Complexity**                        |
| -------------------------------------- | ---------------------------------------------- |
| Loop over `prices` → O(n)              | Only `min_price`, `max_profit`, `price` → O(1) |
| Comparison and subtraction → O(1) each | No extra list or array used                    |
| Single pass → very efficient           | Return value → O(1)                            |
| **✅ Final Time Complexity:** O(n)      | **✅ Total Space Complexity:** O(1)             |


## 🔹 Dry Run: Optimized Method (Example: [7,1,5,3,6,4])

| Day | Price | Min Price so far | Profit if Sell Today | Max Profit so far |
|-----|-------|-----------------|--------------------|-----------------|
| 1   | 7     | 7               | 7 - 7 = 0          | 0               |
| 2   | 1     | 1               | 1 - 1 = 0          | 0               |
| 3   | 5     | 1               | 5 - 1 = 4          | 4               |
| 4   | 3     | 1               | 3 - 1 = 2          | 4               |
| 5   | 6     | 1               | 6 - 1 = 5          | 5               |
| 6   | 4     | 1               | 4 - 1 = 3          | 5               |

✅ **Final Max Profit = 5** (Buy at 1, Sell at 6)


```python
class Solution:
    def maxProfit(self, prices):
        min_price = float('inf')  # Cheapest buy so far
        max_profit = 0

        for price in prices:
            if price < min_price:
                min_price = price  # Update cheapest buy
            elif price - min_price > max_profit:
                max_profit = price - min_price  # Update max profit

        return max_profit

# Example test
sol = Solution()
print(sol.maxProfit([7,1,5,3,6,4]))  # Output: 5
```

# 5 Max Product of two integer

## 🔹 Dry Run: Optimized Maximum Product (Example: [3,4,2])

| Step | num | max1 | max2 | Explanation                      |
| ---- | --- | ---- | ---- | -------------------------------- |
| 1    | 3   | 3    | 0    | First number → max1 updated      |
| 2    | 4   | 4    | 3    | 4 > max1 → max2 = max1, max1 = 4 |
| 3    | 2   | 4    | 3    | 2 < max2 → no change             |

✅ Final Max Product = (4-1)*(3-1) = 3 * 2 = 6


```python
def maxProduct(nums):
    max1,max2 = 0,0
    for num in nums: # nums =3,4
        if num > max1: #3>0,4>3
           max2 = max1 #0
           max1 = num  # 3
        elif num > max2:
           max2 = num  
    return (max1-1) * (max2 - 1)
print(maxProduct(nums=[3,4,1]))       
```

| ⏱️ **Time Complexity**                           | 🧠 **Space Complexity**                |
| ------------------------------------------------ | -------------------------------------- |
| Loop over `nums` → O(n)                          | Only two variables `max1, max2` → O(1) |
| Comparisons and assignments → O(1) per iteration | No extra list or array                 |
| Final multiplication → O(1)                      | Return value → O(1)                    |
| **✅ Total Time Complexity:** O(n)                | **✅ Total Space Complexity:** O(1)     |

# 7 - 2D Diagonal List

n = len(mat) = 3



**Formula:** `mat[i][n-1-i]` → picks elements from top-right → bottom-left

| i (row) | n-1-i (col) | mat[i][n-1-i] | Explanation                  |
|----------|------------|----------------|------------------------------|
| 0        | 2          | 3              | Top-right element             |
| 1        | 1          | 5              | Middle element                |
| 2        | 0          | 7              | Bottom-left element           |

✅ Secondary diagonal elements: 3, 5, 7  



- Primary diagonal → top-left → bottom-right → `mat[i][i]` → 1, 5, 9  
- Secondary diagonal → top-right → bottom-left → `mat[i][n-1-i]` → 3, 5, 7  

| i (row) | Primary mat[i][i] | Secondary mat[i][n-1-i] | Total so far |
|----------|-----------------|-------------------------|-------------|
| 0        | 1               | 3                       | 4           |
| 1        | 5               | 5                       | 14          |
| 2        | 9               | 7                       | 30          |

- n = 3 (odd) → middle element `mat[1][1] = 5` counted twice → subtract once  
✅ Final Sum = 30 - 5 = 25  

**Notes:**  
- Primary diagonal → `mat[i][i]`  
- Secondary diagonal → `mat[i][n-1-i]`   


```python
from typing import List
class Solution:
    def diagonalSum(self, mat:List[List[int]])-> int:
        n = len(mat)
        total = 0
        for i in range(n):
            total += mat[i][i]
            total += mat[i][n-1-i]
        if n % 2 == 1:
            total -= mat[n//2][n//2]
        return total   

# Example test
sol = Solution()
print(sol.diagonalSum([[1,2,3],
                       [4,5,6],
                       [7,8,9]]))  # Output: 25     
```

| ⏱️ **Time Complexity**                           | 🧠 **Space Complexity**                  |
| ------------------------------------------------ | ---------------------------------------- |
| Loop: `for i in range(n)` → O(n)                 | Only integer `total`, `i` → O(1)         |
| Access `mat[i][i]` & `mat[i][n-1-i]` → O(1) each | No extra list or matrix → in-place       |
| Conditional check `if n % 2 == 1` → O(1)         | Temporary integer for subtraction → O(1) |
| **✅ Final Time Complexity:** O(n)                | **✅ Total Space Complexity:** O(1)       |


# 8 Third Maximum Number

| Step | Operation          | Array/Value | Explanation                       |
| ---- | ------------------ | ----------- | --------------------------------- |
| 1    | Remove duplicates  | `[3,2,1]`   | No duplicates                     |
| 2    | Sort descending    | `[3,2,1]`   | Sorted in reverse order           |
| 3    | Check length >= 3  | Yes         | Enough elements to pick third max |
| 4    | Return 3rd element | `1`         | Third maximum number              |


```python
from typing import List

class Solution:
    def thirdMax(self, nums: List[int]) -> int:
        # Step 1️⃣: Remove duplicates
        unique_nums = list(set(nums))
        
        # Step 2️⃣: Sort in descending order
        unique_nums.sort(reverse=True)
        
        # Step 3️⃣: Return 3rd max if it exists, else the largest
        return unique_nums[2] if len(unique_nums) >= 3 else unique_nums[0]

# 🔹 Example Runs
sol = Solution()
print(sol.thirdMax([3, 2, 1]))      # 1
print(sol.thirdMax([1, 2]))         # 2
print(sol.thirdMax([2, 2, 3, 1]))   # 1
```

| ⏱️ **Time Complexity**                                                  | 🧠 **Space Complexity**                               |
| ----------------------------------------------------------------------- | ----------------------------------------------------- |
| `set(nums)` → O(n)                                                      | `unique_nums` list → O(n), stores all unique elements |
| `list(set(nums))` → O(n)                                                | Temporary set → O(n)                                  |
| `unique_nums.sort(reverse=True)` → O(k log k), k = #unique elements ≤ n | Sorting in-place, no extra memory → O(1) extra        |
| Index access `unique_nums[2]` → O(1)                                    | Loop variables / pointers → O(1)                      |
| **✅ Final Time Complexity:** O(n + k log k) ≈ O(n log n) worst case     | **✅ Total Space Complexity:** O(n)                    |


# 9 Remove Duplicates from Sorted Array


#### 🧩 Dry Run: Remove Duplicates from [0,0,1,1,2]

## Initial State
nums = [0,0,1,1,2]  
i = 0  # slow pointer (position to place next unique element)

---

## Step-by-Step Table

| Step | i  | j  | nums[j] | nums[i] | Condition (`nums[j] != nums[i]`) | Action                      | Array State       |
|------|----|----|---------|---------|----------------------------------|-----------------------------|-----------------|
| 1️⃣   | 0  | 1  | 0       | 0       | ❌ False                        | skip                        | [0,0,1,1,2]     |
| 2️⃣   | 0  | 2  | 1       | 0       | ✅ True                         | i += 1 → nums[1] = 1       | [0,1,1,1,2]     |
| 3️⃣   | 1  | 3  | 1       | 1       | ❌ False                        | skip                        | [0,1,1,1,2]     |
| 4️⃣   | 1  | 4  | 2       | 1       | ✅ True                         | i += 1 → nums[2] = 2       | [0,1,2,1,2]     |

---

## Final Output

- **Length of unique elements:** `i + 1 = 3`  
- **Array after removing duplicates:** `[0,1,2]` (first 3 elements)

---

### ✅ How it works

1. `i` tracks **where to put the next unique element**.  
2. `j` scans the array from index 1 to end.  
3. If `nums[j] != nums[i]` → increment `i` and copy `nums[j]` to `nums[i]`.  
4. Return `i+1` → number of unique elements.  

---



nums = [0,0,1,1,1,2,2,3,3,4]


| Step  | j | Condition (`nums[j] != nums[i]`) | Action                 | i (after) | nums (after update)     |
| ----- | - | -------------------------------- | ---------------------- | --------- | ----------------------- |
| Start | - | -                                | Initialize `i = 0`     | 0         | `[0,0,1,1,1,2,2,3,3,4]` |
| 1     | 1 | `0 == 0 → False`                 | Skip duplicate         | 0         | `[0,0,1,1,1,2,2,3,3,4]` |
| 2     | 2 | `1 != 0 → True`                  | `i=1`, set `nums[1]=1` | 1         | `[0,1,1,1,1,2,2,3,3,4]` |
| 3     | 3 | `1 == 1 → False`                 | Skip duplicate         | 1         | `[0,1,1,1,1,2,2,3,3,4]` |
| 4     | 4 | `1 == 1 → False`                 | Skip duplicate         | 1         | `[0,1,1,1,1,2,2,3,3,4]` |
| 5     | 5 | `2 != 1 → True`                  | `i=2`, set `nums[2]=2` | 2         | `[0,1,2,1,1,2,2,3,3,4]` |
| 6     | 6 | `2 == 2 → False`                 | Skip duplicate         | 2         | `[0,1,2,1,1,2,2,3,3,4]` |
| 7     | 7 | `3 != 2 → True`                  | `i=3`, set `nums[3]=3` | 3         | `[0,1,2,3,1,2,2,3,3,4]` |
| 8     | 8 | `3 == 3 → False`                 | Skip duplicate         | 3         | `[0,1,2,3,1,2,2,3,3,4]` |
| 9     | 9 | `4 != 3 → True`                  | `i=4`, set `nums[4]=4` | 4         | `[0,1,2,3,4,2,2,3,3,4]` |


```python
from typing import List

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if not nums:
            return 0
        i = 0
        for j in range(1, len(nums)):
            if nums[j]!= nums[i]:
                 i+=1
                 nums[i] = nums[j]
        return i+1         
sol = Solution()
print(sol.removeDuplicates([0,0,1,1,1,2,2,3,3,4]))
```

| ⏱️ **Time Complexity**                               | 🧠 **Space Complexity**                             |
| ---------------------------------------------------- | --------------------------------------------------- |
| Loop: `for j in range(1, len(nums))` → O(n)          | In-place update → no extra list used → O(1)         |
| Comparison `nums[j] != nums[i]` → O(1) per iteration | Loop variables `i, j` → O(1)                        |
| Assignment `nums[i] = nums[j]` → O(1)                | No extra memory for output, array modified in-place |
| **✅ Final Time Complexity:** O(n)                    | **✅ Total Space Complexity:** O(1)                  |
| Reason: Single pass through array of length n        | Reason: In-place, only a few integer variables      |

# 10 Piar Sum

🧩 Problem Description

Given an integer array arr and a target value target, find all unique pairs of numbers that add up to the target sum.
Return them in string format like "a+b".

🧠 Algorithm Explanation

Initialize an empty list result.

Use two nested loops:

Outer loop iterates through each number arr[i].

Inner loop checks the next numbers arr[j].

If the sum equals the target, store the pair.

Finally, return all pairs found.

## 🧮 Step-by-Step Dry Run

| Step   | i  | j  | arr[i] | arr[j] | Sum  | Check (== 5) | Action        |
|--------|----|----|--------|--------|------|---------------|---------------|
| 1️⃣     | 0  | 1  | 2      | 4      | 6    | ❌            | continue      |
| 2️⃣     | 0  | 2  | 2      | 3      | 5    | ✅            | add "2+3"    |
| 3️⃣     | 0  | 3  | 2      | 5      | 7    | ❌            | continue      |
| 4️⃣     | 0  | 4  | 2      | 6      | 8    | ❌            | continue      |
| 5️⃣     | 0  | 5  | 2      | -1     | 1    | ❌            | continue      |
| 6️⃣     | 1  | 2  | 4      | 3      | 7    | ❌            | continue      |
| 7️⃣     | 1  | 3  | 4      | 5      | 9    | ❌            | continue      |
| 8️⃣     | 1  | 4  | 4      | 6      | 10   | ❌            | continue      |
| 9️⃣     | 1  | 5  | 4      | -1     | 3    | ❌            | continue      |
| 🔟      | 2  | 3  | 3      | 5      | 8    | ❌            | continue      |
| 1️⃣1️⃣   | 2  | 4  | 3      | 6      | 9    | ❌            | continue      |
| 1️⃣2️⃣   | 2  | 5  | 3      | -1     | 2    | ❌            | continue      |
| 1️⃣3️⃣   | 3  | 4  | 5      | 6      | 11   | ❌            | continue      |
| 1️⃣4️⃣   | 3  | 5  | 5      | -1     | 4    | ❌            | continue      |
| 1️⃣5️⃣   | 4  | 5  | 6      | -1     | 5    | ✅            | add "6+-1"   |


```python
def twoSum(arr, target):
    result = []
    n = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            if arr[i] + arr[j] == target:
                result.append(f"{arr[i]}+ {arr[j]}")
    return result            
print(twoSum([2, 4, 3, 5, 6, -1], 5))
```

| ⏱️ **Time Complexity**                                      | 🧠 **Space Complexity**                                                 |
| ----------------------------------------------------------- | ----------------------------------------------------------------------- |
| Outer loop: `for i in range(n)` → O(n)                      | `result` list → O(k), k = number of pairs matching target               |
| Inner loop: `for j in range(i+1, n)` → O(n-i) → total O(n²) | Temporary strings in `result.append(f"{arr[i]}+{arr[j]}")` → O(k) extra |
| Check `if arr[i] + arr[j] == target` → O(1)                 | Loop variables `i, j` → O(1)                                            |
| **✅ Final Time Complexity:** O(n²)                          | **✅ Total Space Complexity:** O(k) (depends on number of pairs)         |


# 11 . contain Duplicates

Dry Run

nums = [1,2,3,1]

Start: seen = {}

n=1 → not in set → add → seen={1}

n=2 → not in set → add → seen={1,2}

n=3 → not in set → add → seen={1,2,3}

n=1 → already in set → return True

```python
class Solution:
    def ContainDuplicates(self, nums:List[int])-> bool:
        seen = set()
        for n in nums:
            if n in seen:
                return True
            seen.add(n)
        return False  
sol=Solution()     
print(sol.ContainDuplicates([1,2,3,1]))     
```

| ⏱️ **Time Complexity**                                    | 🧠 **Space Complexity**                                           |
| --------------------------------------------------------- | ----------------------------------------------------------------- |
| `seen = set()` → O(1)                                     | `seen` set → O(n) in worst case, stores all elements              |
| Loop: `for n in nums:` → O(n)                             | Loop variable `n` → O(1)                                          |
| Check `if n in seen` → O(1) average                       | Add `seen.add(n)` → O(1) average                                  |
| Return early if duplicate found → best case O(1)          | No extra memory besides set                                       |
| **✅ Final Time Complexity:** O(n)                         | **✅ Total Space Complexity:** O(n)                                |
| Reason: Each element is checked and inserted at most once | Reason: In worst case all elements unique → set stores n elements |


# 12 Permutation Check

Problem:

Given two strings s1 and s2, check if s2 contains a permutation of s1.

Return True if yes, otherwise False.

Example:

s1 = "ab"

s2 = "eidbaooo"


Algorithm:

Sort s1 → s1_sorted = ['a','b']

Slide a window of length len(s1) over s2.

Check if the sorted substring matches s1_sorted.

### Dry Run: Permutation in String

**Input:**  
s1 = "ab"  
s2 = "eidbaooo"

| Step | i | substring (`s2[i:i+len(s1)]`) | sorted(substring) | Comparison with `s1_sorted` | Result / Action                  |
|------|---|-------------------------------|-----------------|-----------------------------|---------------------------------|
| 1    | 0 | "ei"                          | ['e','i']       | ['e','i'] != ['a','b']      | No match, continue               |
| 2    | 1 | "id"                          | ['d','i']       | ['d','i'] != ['a','b']      | No match, continue               |
| 3    | 2 | "db"                          | ['b','d']       | ['b','d'] != ['a','b']      | No match, continue               |
| 4    | 3 | "ba"                          | ['a','b']       | ['a','b'] == ['a','b']      | Match found → return True        |




**Explanation:**  
- Humne `s2` ke har substring of length 2 (same as `s1`) ko check kiya.  
- Jab i = 3 pe substring `"ba"` mili, jo sorted hone ke baad `['a','b']` ban gayi.  
- Ye `s1_sorted` ke equal hai → permutation exists → **return True**.

```python
class Solution:
    def checkInclusion(self,s1:str, s2:str)->bool:
        s1_sorted = sorted(s1)
        len_s1= len(s1)
        for i in range(len(s2)-len_s1+1): #range(8-2+1)=range(7)
            sub_string = s2[i:i+len_s1]
            if sorted(sub_string) == s1_sorted:
                return True
        return False    
```

```python
# Test
sol = Solution()
print(sol.checkInclusion("ab", "eidbaooo"))  # True
print(sol.checkInclusion("ab", "eidboaoo"))  # False
```

| ⏱️ **Time Complexity**                                     | 🧠 **Space Complexity**                                  |
| ---------------------------------------------------------- | -------------------------------------------------------- |
| `s1_sorted = sorted(s1)` → O(n log n) where n = len(s1)    | `s1_sorted` → O(n) stores sorted s1                      |
| Loop: `for i in range(len(s2)-len_s1+1)` → O(m-n+1) ≈ O(m) | `sub_string = s2[i:i+len_s1]` → O(n) temporary substring |
| `sorted(sub_string)` → O(n log n) per iteration            | Sorting creates temporary list → O(n)                    |
| Comparison `== s1_sorted` → O(n) per iteration             | Comparison uses O(n) extra memory for temp list          |
| **Total time:** O((m-n+1) * n log n) ≈ O(m * n log n)      | **Total space:** O(n) + O(n) temporary → O(n)            |


# 13 #48 Rotate Matrix

We need to rotate the matrix 90° clockwise, in-place (without creating a new matrix).

Example Input:

matrix = 

[[1, 2, 3],

  [4, 5, 6],

  [7, 8, 9]]


Expected Output:

[[7, 4, 1],

  [8, 5, 2],

  [9, 6, 3]]

### ⚙️ Step 1: Transpose the Matrix

Transpose means — convert rows into columns (swap matrix[i][j] ↔ matrix[j][i]).

| i | j | Swap                       | Matrix after swap           |
| - | - | -------------------------- | --------------------------- |
| 0 | 0 | same                       | no change                   |
| 0 | 1 | (0,1) ↔ (1,0) → swap 2 ↔ 4 | `[[1,4,3],[2,5,6],[7,8,9]]` |
| 0 | 2 | (0,2) ↔ (2,0) → swap 3 ↔ 7 | `[[1,4,7],[2,5,6],[3,8,9]]` |
| 1 | 1 | same                       | no change                   |
| 1 | 2 | (1,2) ↔ (2,1) → swap 6 ↔ 8 | `[[1,4,7],[2,5,8],[3,6,9]]` |
| 2 | 2 | same                       | no change                   |


```python
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        n = len(matrix) #n=3
        for i in range(n):
            for j in range(i,n):
                matrix[i][j], matrix[j][i] = matrix[j][i],matrix[i][j]
        for row in matrix:
            row.reverse()   

# Example usage
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

sol = Solution()
sol.rotate(matrix)

# Print final rotated matrix
for row in matrix:
    print(row)                 

```

| ⏱️ **Time Complexity**                                                      | 🧠 **Space Complexity**                                |
| --------------------------------------------------------------------------- | ------------------------------------------------------ |
| Outer loop: `for i in range(n)` → O(n)                                      | Only uses `i, j` → O(1)                                |
| Inner loop: `for j in range(i,n)` → O(n-i) per i → total O(n²/2) ≈ O(n²)    | Swaps done in-place → O(1)                             |
| Swap: `matrix[i][j], matrix[j][i]` → O(1) per swap                          | Temporary variable for swap → O(1)                     |
| Second loop: `for row in matrix: row.reverse()` → O(n²)                     | Row reversal done in-place → O(1)                      |
| **✅ Final Time Complexity:** O(n²)                                          | **✅ Total Space Complexity:** O(1)                     |
| Reason: Each element is accessed and swapped at most twice → quadratic time | Reason: No extra matrix or list created, only pointers |
