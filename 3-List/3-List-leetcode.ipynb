# 6 -350 Intersection of Two Arrays II

We are given two integer arrays `nums1` and `nums2`.  
We need to **return an array that represents their intersection**, where each element appears **as many times as it shows in both arrays**.

### Example:
nums1 = [1,2,2,1],
nums2 = [2,2],
Output = [2,2]
âœ… Because `2` appears twice in both arrays.


### ðŸ§  Dictionary.get() Method & Intersection Algorithm â€” Dry Run

---

### ðŸ” Understanding `dict.get(key, default)`

| **Concept** | **Description** |
|--------------|-----------------|
| **Syntax** | `dictionary.get(key, default_value)` |
| **Purpose** | Returns the value for a key if it exists, else returns the default value. |
| **Use Case** | Avoids errors when trying to access missing keys in a dictionary. |

---

### âœ… Example 1: When Key Exists

| Code | Output |
|------|--------|
| `count1 = {1: 2, 2: 2}` |  |
| `count1.get(2, 0)` | `2` |
| `count1.get(1, 0)` | `2` |

ðŸ‘‰ **Explanation:** Both keys `1` and `2` exist, so actual values are returned.

---

### ðŸš« Example 2: When Key Doesnâ€™t Exist

| Code | Output |
|------|--------|
| `count1 = {1: 2, 2: 2}` |  |
| `count1.get(5, 0)` | `0` |
| `count1.get(3, 0)` | `0` |

ðŸ‘‰ **Explanation:** Keys `5` and `3` are missing, so the default value `0` is returned safely.

---

### âš™ï¸ Complete Intersection Algorithm â€” Step-by-Step

| **Input** | **Description** |
|------------|----------------|
| `nums1 = [1, 2, 2, 1]` | First array |
| `nums2 = [2, 2]` | Second array |

---

### ðŸ§© Initialization

| Variable | Value |
|-----------|--------|
| `count1` | `Counter({1: 2, 2: 2})` |
| `result` | `[]` |

---

### ðŸ” Iteration Details

| Step | num (from nums2) | count1.get(num, 0) | Condition | Action | Updated `count1` | `result` |
|------|-------------------|--------------------|------------|---------|------------------|----------|
| 1 | 2 | 2 | `2 > 0 â†’ True` | Append 2 | `{1:2, 2:1}` | `[2]` |
| 2 | 2 | 1 | `1 > 0 â†’ True` | Append 2 | `{1:2, 2:0}` | `[2, 2]` |

---

### âœ… Final Output

| **Result** | `[2, 2]` |
|-------------|----------|

---

### ðŸ§¾ Key Points Summary

| Concept | Explanation |
|----------|-------------|
| `dict.get(key, default)` | Safely retrieves values, prevents `KeyError` |
| Default = 0 | Ensures missing numbers donâ€™t cause issues |
| Decrement step | Keeps track of how many times a number can be reused |


---



---

## âš™ï¸ Approach

We use **Counter** (from `collections` module) to store the frequency of each element in `nums1`.  
Then, we loop through each element in `nums2` and check:

- If the number exists in `count1` (its count > 0)
  - Add it to the result.
  - Decrease its count by 1 (because one occurrence is now used).

This ensures that we count duplicates correctly â€”  
each number appears **only as many times as it appears in both arrays**.

---


```python
from collections import Counter
from typing import List
class Solution:
    def intersect(self, nums1:List[int], nums2:List[int])->List[int]:
        count1 = Counter(nums1)
        result =[]
        for num in nums2:
            if count1.get(num,0) > 0:
                result.append(num)
                count1[num] -= 1
        return result 
print(Solution().intersect([1,2,2,1],[2,2]))         
```

## 7. 88 â€“ Merge Sorted Array âœ… *(new)*  

nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3


| Step | Description             | Operation / Details                                                                          | `nums1` after Step       |
| ---- | ----------------------- | -------------------------------------------------------------------------------------------- | ------------------------ |
| 1    | Start                   | Initial arrays                                                                               | `[1, 2, 3, 0, 0, 0]`     |
| 2    | `nums1[m:] = nums2[:n]` | `m=3 â†’ nums1[3:] = [0,0,0]`<br>`n=3 â†’ nums2[:3] = [2,5,6]`<br>Replace `[3:6]` with `[2,5,6]` | `[1, 2, 3, 2, 5, 6]`     |
| 3    | `nums1.sort()`          | Sort all elements in non-decreasing order                                                    | `[1, 2, 2, 3, 5, 6]`     |
| âœ…    | **Final Result**        | Merged and sorted array                                                                      | **`[1, 2, 2, 3, 5, 6]`** |


```python
class Solution:
    def merge(self,nums1, m, nums2,n):
        nums1[m:] = nums2[:n]
        nums1.sort()
#Test        
sol = Solution()    
nums1 = [1, 2, 3, 0, 0, 0]
nums2 = [2, 5, 6]    
sol.merge(nums1, 3, nums2, 3)
print(nums1)  # Output: [1, 2, 2, 3, 5, 6]        

```

# 8. 283 â€“ Move Zeroes âœ… *(new)*  


Given an integer array nums, move all 0â€™s to the end of it while maintaining the relative order of the non-zero elements.
You must do this in-place, without making a copy of the array.

nums = [0,1,0,3,12]

| Step | i | j | Condition (`nums[i] != 0`) | Action                              | `nums` after step |
| ---- | - | - | -------------------------- | ----------------------------------- | ----------------- |
| 1    | 0 | 0 | âŒ (`nums[0]=0`)            | Skip                                | `[0,1,0,3,12]`    |
| 2    | 1 | 0 | âœ… (`nums[1]=1`)            | Swap `nums[1]` and `nums[0]`; `j=1` | `[1,0,0,3,12]`    |
| 3    | 2 | 1 | âŒ (`nums[2]=0`)            | Skip                                | `[1,0,0,3,12]`    |
| 4    | 3 | 1 | âœ… (`nums[3]=3`)            | Swap `nums[3]` and `nums[1]`; `j=2` | `[1,3,0,0,12]`    |
| 5    | 4 | 2 | âœ… (`nums[4]=12`)           | Swap `nums[4]` and `nums[2]`; `j=3` | `[1,3,12,0,0]`    |

âœ… Final Output: [1,3,12,0,0]

```python
class Solution:
    def moveZeroes(self, nums:List[int])-> None:
        j = 0
        for i in range(len(nums)):#5
            if nums[i] != 0: #3
                nums[j], nums[i] = nums[i], nums[j] 
                j += 1
        return nums     

# Example usage:
print(Solution().moveZeroes([0,1,0,3,12]))  # [1,3,12,0,0]
print(Solution().moveZeroes([0]))           # [0]
print(Solution().moveZeroes([1,0,1]))       # [1,1,0]
```

# 9. 118 â€“ Pascalâ€™s Triangle âœ… *(new)*  


Pascalâ€™s Triangle is a special number pattern where each row is built using the sum of two adjacent numbers from the previous row.

ðŸ§  Basic Rules

The first and last number in every row is always 1.

Each middle number is the sum of the two numbers directly above it from the previous row.

### ðŸ”¹ Example Structure
| Row No. | Formula                 | Result            |
| ------- | ----------------------- | ----------------- |
| Row 0   | `[1]`                   | `[1]`             |
| Row 1   | `[1, 1]`                | `[1, 1]`          |
| Row 2   | `[1, 1+1, 1]`           | `[1, 2, 1]`       |
| Row 3   | `[1, 1+2, 2+1, 1]`      | `[1, 3, 3, 1]`    |
| Row 4   | `[1, 1+3, 3+3, 3+1, 1]` | `[1, 4, 6, 4, 1]` |


# ðŸ§® Dry Run (numRows = 5)

| Step | i (Row No.) | Initial Row       | Inner Loop Range                     |row[j]=triangle[i - 1][j - 1] + triangle[i - 1][j]                                                                                                                                                                                                 | Final Row         | Triangle So Far                                           |
| ---- | ----------- | ----------------- | ------------------------------------ | -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- | ----------------- | --------------------------------------------------------- |
| 1    | 0           | `[1]`             | `range(1, 0)` â†’ `[]` â†’ No iterations | â€”                                                                                                                                                                                                              | `[1]`             | `[[1]]`                                                   |
| 2    | 1           | `[1, 1]`          | `range(1, 1)` â†’ `[]` â†’ No iterations | â€”                                                                                                                                                                                                              | `[1, 1]`          | `[[1], [1, 1]]`                                           |
| 3    | 2           | `[1, 1, 1]`       | `range(1, 2)` â†’ `[1]`                | **j = 1** â†’ `row[1] = triangle[1][0] + triangle[1][1] = 1 + 1 = 2`                                                                                                                                             | `[1, 2, 1]`       | `[[1], [1, 1], [1, 2, 1]]`                                |
| 4    | 3           | `[1, 1, 1, 1]`    | `range(1, 3)` â†’ `[1, 2]`             | **j = 1** â†’ `row[1] = triangle[2][0] + triangle[2][1] = 1 + 2 = 3`<br>**j = 2** â†’ `row[2] = triangle[2][1] + triangle[2][2] = 2 + 1 = 3`                                                                       | `[1, 3, 3, 1]`    | `[[1], [1, 1], [1, 2, 1], [1, 3, 3, 1]]`                  |
| 5    | 4           | `[1, 1, 1, 1, 1]` | `range(1, 4)` â†’ `[1, 2, 3]`          | **j = 1** â†’ `row[1] = triangle[3][0] + triangle[3][1] = 1 + 3 = 4`<br>**j = 2** â†’ `row[2] = triangle[3][1] + triangle[3][2] = 3 + 3 = 6`<br>**j = 3** â†’ `row[3] = triangle[3][2] + triangle[3][3] = 3 + 1 = 4` | `[1, 4, 6, 4, 1]` | `[[1], [1, 1], [1, 2, 1], [1, 3, 3, 1], [1, 4, 6, 4, 1]]` |



ðŸ’¡ Key Notes:

range(1, i) â†’ gives indices of middle elements only.

First (j=0) and last (j=i) elements are always 1.

Each row[j] = triangle[i-1][j-1] + triangle[i-1][j].

So the inner loop builds the middle part of each row.

```python
class Solution:
    def generate(self, numRows: int) -> list[list[int]]:
        triangle = []

        for i in range(numRows):
            # Create a row with (i+1) elements initialized to 1
            row = [1] * (i + 1)

            # Fill in the middle elements using the previous row
            for j in range(1, i):
                row[j] = triangle[i - 1][j - 1] + triangle[i - 1][j]

            # Add the completed row to the triangle
            triangle.append(row)

        return triangle

# Example usage:
print(Solution().generate(5))
```

# 10 -485. Max Consecutive Ones

You are given a binary array (only contains 0s and 1s). You need to find the longest sequence of consecutive 1's in the array.

Input: [1,1,0,1,1,1]

Groups: [1,1] + [1,1,1]

Longest: 3

Output: 3

Input: nums = [1, 1, 0, 1, 1, 1]

| **Step** | **Operation**    | **Explanation**                                     | **Result**                       |
| -------- | ---------------- | --------------------------------------------------- | -------------------------------- |
| 1        | `map(str, nums)` | Convert each number in the list to a string         | `['1', '1', '0', '1', '1', '1']` |
| 2        | `''.join(...)`   | Join all string elements into one continuous string | `'110111'`                       |
| 3        | `.split('0')`    | Split the string at every `'0'`                     | `['11', '111']`                  |
| 4        | `map(len, ...)`  | Get the length of each string segment               | `[2, 3]`                         |
| 5        | `max(...)`       | Find the maximum length value                       | `3`                              |

return max(map(len, ''.join(map(str, nums)).split('0')))


```python
from typing import List
class Solution:
    def findMaxConsecutiveOnes(self, nums:List[int])->int:
        return max(map(len, ''.join(map(str, nums)).split('0')))
# Example 1
nums = [1,1,0,1,1,1]
print(Solution().findMaxConsecutiveOnes(nums))  # Output: 3  

# Example 2  
nums = [1,0,1,1,0,1]
print(Solution().findMaxConsecutiveOnes(nums))  # Output: 2    
```
