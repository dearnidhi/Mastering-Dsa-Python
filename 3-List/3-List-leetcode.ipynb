# 6 -350 Intersection of Two Arrays II

We are given two integer arrays `nums1` and `nums2`.  
We need to **return an array that represents their intersection**, where each element appears **as many times as it shows in both arrays**.

### Example:
nums1 = [1,2,2,1],
nums2 = [2,2],
Output = [2,2]
✅ Because `2` appears twice in both arrays.


### 🧠 Dictionary.get() Method & Intersection Algorithm — Dry Run

---

### 🔍 Understanding `dict.get(key, default)`

| **Concept** | **Description** |
|--------------|-----------------|
| **Syntax** | `dictionary.get(key, default_value)` |
| **Purpose** | Returns the value for a key if it exists, else returns the default value. |
| **Use Case** | Avoids errors when trying to access missing keys in a dictionary. |

---

### ✅ Example 1: When Key Exists

| Code | Output |
|------|--------|
| `count1 = {1: 2, 2: 2}` |  |
| `count1.get(2, 0)` | `2` |
| `count1.get(1, 0)` | `2` |

👉 **Explanation:** Both keys `1` and `2` exist, so actual values are returned.

---

### 🚫 Example 2: When Key Doesn’t Exist

| Code | Output |
|------|--------|
| `count1 = {1: 2, 2: 2}` |  |
| `count1.get(5, 0)` | `0` |
| `count1.get(3, 0)` | `0` |

👉 **Explanation:** Keys `5` and `3` are missing, so the default value `0` is returned safely.

| **Input** | **Description** |
|------------|----------------|
| `nums1 = [1, 2, 2, 1]` | First array |
| `nums2 = [2, 2]` | Second array |

| Step | num (from nums2) | count1.get(num, 0) | Condition | Action | Updated `count1` | `result` |

|----------|-------------|

## ⚙️ Approach

We use **Counter** (from `collections` module) to store the frequency of each element in `nums1`.  
Then, we loop through each element in `nums2` and check:

- If the number exists in `count1` (its count > 0)
  - Add it to the result.
  - Decrease its count by 1 (because one occurrence is now used).

This ensures that we count duplicates correctly —  
each number appears **only as many times as it appears in both arrays**.

---


```python
from collections import Counter
from typing import List
class Solution:
    def intersect(self, nums1:List[int], nums2:List[int])->List[int]:
        count1 = Counter(nums1)
        result = []
        for num in nums2:
            if count1.get(num,0)> 0:
                result.append(num)
                count1[num] -= 1
        return result
      
print(Solution().intersect([1,2,2,1],[2,2]))         
```

| ⏱️ **Time Complexity**                                                  | 🧠 **Space Complexity**                                                  |
| ----------------------------------------------------------------------- | ------------------------------------------------------------------------ |
| **`count1 = Counter(nums1)`** → O(n)<br>build frequency map for `nums1` | **`count1` dictionary** → O(n)<br>stores counts of elements from `nums1` |
| **`for num in nums2:`** → O(m)<br>loop over all elements in `nums2`     | **`result` list** → O(min(n, m))<br>stores intersection elements         |
| **`count1.get(num,0)`** → O(1) per lookup                               | Loop variable `num` → O(1)                                               |
| **`result.append(num)`** → O(1) amortized                               | Updating `count1[num]` → O(1)                                            |
| **`count1[num] -= 1`** → O(1)                                           | No other extra storage → O(1)                                            |
| **`return result`** → O(1)                                              | Returning reference → O(1)                                               |
| **✅ Final Time Complexity:** O(n + m)<br>n = len(nums1), m = len(nums2) | **✅ Total Space Complexity:** O(n + min(n, m))                           |


## 7. 88 – Merge Sorted Array ✅ *(new)*  

nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3


| Step | Description             | Operation / Details                                                                          | `nums1` after Step       |
| ---- | ----------------------- | -------------------------------------------------------------------------------------------- | ------------------------ |
| 1    | Start                   | Initial arrays                                                                               | `[1, 2, 3, 0, 0, 0]`     |
| 2    | `nums1[m:] = nums2[:n]` | `m=3 → nums1[3:] = [0,0,0]`<br>`n=3 → nums2[:3] = [2,5,6]`<br>Replace `[3:6]` with `[2,5,6]` | `[1, 2, 3, 2, 5, 6]`     |
| 3    | `nums1.sort()`          | Sort all elements in non-decreasing order                                                    | `[1, 2, 2, 3, 5, 6]`     |
| ✅    | **Final Result**        | Merged and sorted array                                                                      | **`[1, 2, 2, 3, 5, 6]`** |


```python
class Solution:
    def merge(self,nums1, m, nums2,n):
        nums1[m:] = nums2[:n]
        nums1.sort()
        
#Test        
sol = Solution()    
nums1 = [1, 2, 3, 0, 0, 0]
nums2 = [2, 5, 6]    
sol.merge(nums1, 3, nums2, 3)
print(nums1)  # Output: [1, 2, 2, 3, 5, 6]        

```

| ⏱️ **Time Complexity**                                              | 🧠 **Space Complexity**                                                                                   |
| ------------------------------------------------------------------- | --------------------------------------------------------------------------------------------------------- |
| **`nums1[m:] = nums2[:n]`** → O(n) <br>copy `n` elements from nums2 | **Temporary slice `nums2[:n]`** → O(n) <br>creates a temporary list                                       |
| **`nums1.sort()`** → O((m+n) log(m+n)) <br>sort combined array      | **Sort in-place** → O(1) if Timsort uses in-place optimizations <br>otherwise O(m+n) worst-case auxiliary |
| **✅ Final Time Complexity:** O((m+n) log(m+n))                      | **✅ Total Space Complexity:** O(n) (due to temporary slice)                                               |


# or

nums1 = [1,2,3,0,0,0], m=3     
nums2 = [2,5,6], n=3  
Pointer    
| Pointer | Initial Value | Meaning                    |
| ------- | ------------- | -------------------------- |
| `p1`    | 2             | last valid index in nums1  |
| `p2`    | 2             | last index in nums2        |
| `p`     | 5             | last index in merged array |

Dry Run   

| Step | nums1 state   | nums2 state | p1 | p2 | p | Action / Explanation                                             |
| ---- | ------------- | ----------- | -- | -- | - | ---------------------------------------------------------------- |
| 0    | [1,2,3,0,0,0] | [2,5,6]     | 2  | 2  | 5 | Initialize pointers                                              |
| 1    | [1,2,3,0,0,6] | [2,5,6]     | 2  | 1  | 4 | nums2[2]=6 > nums1[2]=3 → place 6 at nums1[5], decrement p2 & p  |
| 2    | [1,2,3,0,5,6] | [2,5,6]     | 2  | 0  | 3 | nums2[1]=5 > nums1[2]=3 → place 5 at nums1[4], decrement p2 & p  |
| 3    | [1,2,3,3,5,6] | [2,5,6]     | 2  | 0  | 3 | nums1[2]=3 > nums2[0]=2 → place 3 at nums1[3], decrement p1 & p  |
| 4    | [1,2,2,3,5,6] | [2,5,6]     | 1  | 0  | 2 | nums2[0]=2 >= nums1[1]=2 → place 2 at nums1[2], decrement p2 & p |
| 5    | [1,2,2,3,5,6] | [2,5,6]     | 1  | -1 | 1 | p2 < 0 → stop loop                                               |
                                          |


```python
class Solution:
    def merge(self, nums1, m, nums2, n):
        # Last se start karo
        p1 = m - 1 # nums1 last valid index
        p2 = n - 1 #nums2 last element
        p = m + n -1 # marged array ka last pos
        while p2 >= 0:
            if p1 >= 0 and nums1[p1] > nums2[p2]:
                nums1[p] = nums1[p1]
                p1 -= 1
            else:
                nums1[p] = nums2[p2]
                p2 -= 1
            p -= 1             

# Test
sol = Solution()
nums1 = [1, 2, 3, 0, 0, 0]
nums2 = [2, 5, 6]
sol.merge(nums1, 3, nums2, 3)
print(nums1)  # [1, 2, 2, 3, 5, 6]
```

| ⏱️ **Time Complexity**                                                 | 🧠 **Space Complexity**                                    |
| ---------------------------------------------------------------------- | ---------------------------------------------------------- |
| **Initialization of pointers** → O(1)<br>`p1, p2, p = ...`             | **Pointers `p1, p2, p`** → O(1)<br>constant memory         |
| **While loop `while p2 >= 0`** → O(m + n)<br>each element visited once | **In-place array `nums1`** → O(1)<br>no extra list created |
| **Comparison `nums1[p1] > nums2[p2]`** → O(1) per iteration            | Loop variable `p, p1, p2` → O(1)                           |
| **Assignment `nums1[p] = ...`** → O(1)                                 | No extra storage for merging → O(1)                        |
| **Pointer decrement `p1/p2/p -= 1`** → O(1)                            |                                                            |
| **Return (in-place)** → O(1)                                           |                                                            |
| **✅ Final Time Complexity:** O(m + n)                                  | **✅ Total Space Complexity:** O(1)                         |


# 8. 283 – Move Zeroes ✅ *(new)*  


Given an integer array nums, move all 0’s to the end of it while maintaining the relative order of the non-zero elements.
You must do this in-place, without making a copy of the array.

nums = [0,1,0,3,12]

| Step | i | j | Condition (`nums[i] != 0`) | Action                              | `nums` after step |
| ---- | - | - | -------------------------- | ----------------------------------- | ----------------- |
| 1    | 0 | 0 | ❌ (`nums[0]=0`)            | Skip                                | `[0,1,0,3,12]`    |
| 2    | 1 | 0 | ✅ (`nums[1]=1`)            | Swap `nums[1]` and `nums[0]`; `j=1` | `[1,0,0,3,12]`    |
| 3    | 2 | 1 | ❌ (`nums[2]=0`)            | Skip                                | `[1,0,0,3,12]`    |
| 4    | 3 | 1 | ✅ (`nums[3]=3`)            | Swap `nums[3]` and `nums[1]`; `j=2` | `[1,3,0,0,12]`    |
| 5    | 4 | 2 | ✅ (`nums[4]=12`)           | Swap `nums[4]` and `nums[2]`; `j=3` | `[1,3,12,0,0]`    |

✅ Final Output: [1,3,12,0,0]

```python
class Solution:
    def moveZeroes(self, nums:List[int])-> None:
        j = 0
        for i in range(len(nums)):
            if nums[i] != 0:
                nums[j],nums[i] = nums[i], nums[j]
                j += 1
        return nums        

# Example usage:
print(Solution().moveZeroes([0,1,0,3,12]))  # [1,3,12,0,0]
print(Solution().moveZeroes([0]))           # [0]
print(Solution().moveZeroes([1,0,1]))       # [1,1,0]
```

| ⏱️ **Time Complexity**                                              | 🧠 **Space Complexity**                                          |
| ------------------------------------------------------------------- | ---------------------------------------------------------------- |
| **`j = 0`** → O(1)<br>initialize pointer                            | **`j` variable** → O(1)<br>constant memory                       |
| **`for i in range(len(nums))`** → O(n)<br>iterate over all elements | **`i` loop variable** → O(1)<br>reused each iteration            |
| **`if nums[i] != 0`** → O(1)<br>comparison                          | **swap `nums[j], nums[i]`** → O(1)<br>in-place, no extra memory  |
| **`nums[j], nums[i] = nums[i], nums[j]`** → O(1)<br>swap operation  | Operations inside loop → O(1)<br>no extra data structures        |
| **`j += 1`** → O(1)<br>increment pointer                            | **nums array** → in-place modification → O(1) extra space        |
| **`return nums`** → O(1)<br>return reference                        | Return value → no extra memory                                   |
| **✅ Final Time Complexity:** O(n)<br>single pass through array      | **✅ Total Space Complexity:** O(1)<br>only constant extra memory |


# 9. 118 – Pascal’s Triangle ✅ *(new)*  


Pascal’s Triangle is a special number pattern where each row is built using the sum of two adjacent numbers from the previous row.

🧠 Basic Rules

The first and last number in every row is always 1.

Each middle number is the sum of the two numbers directly above it from the previous row.

### 🔹 Example Structure
| Row No. | Formula                 | Result            |
| ------- | ----------------------- | ----------------- |
| Row 0   | `[1]`                   | `[1]`             |
| Row 1   | `[1, 1]`                | `[1, 1]`          |
| Row 2   | `[1, 1+1, 1]`           | `[1, 2, 1]`       |
| Row 3   | `[1, 1+2, 2+1, 1]`      | `[1, 3, 3, 1]`    |
| Row 4   | `[1, 1+3, 3+3, 3+1, 1]` | `[1, 4, 6, 4, 1]` |


# 🧮 Dry Run (numRows = 5)

| Step | i (Row No.) | Initial Row       | Inner Loop Range                     |row[j]=triangle[i - 1][j - 1] + triangle[i - 1][j]                                                                                                                                                                                                 | Final Row         | Triangle So Far                                           |
| ---- | ----------- | ----------------- | ------------------------------------ | -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- | ----------------- | --------------------------------------------------------- |
| 1    | 0           | `[1]`             | `range(1, 0)` → `[]` → No iterations | —                                                                                                                                                                                                              | `[1]`             | `[[1]]`                                                   |
| 2    | 1           | `[1, 1]`          | `range(1, 1)` → `[]` → No iterations | —                                                                                                                                                                                                              | `[1, 1]`          | `[[1], [1, 1]]`                                           |
| 3    | 2           | `[1, 1, 1]`       | `range(1, 2)` → `[1]`                | **j = 1** → `row[1] = triangle[1][0] + triangle[1][1] = 1 + 1 = 2`                                                                                                                                             | `[1, 2, 1]`       | `[[1], [1, 1], [1, 2, 1]]`                                |
| 4    | 3           | `[1, 1, 1, 1]`    | `range(1, 3)` → `[1, 2]`             | **j = 1** → `row[1] = triangle[2][0] + triangle[2][1] = 1 + 2 = 3`<br>**j = 2** → `row[2] = triangle[2][1] + triangle[2][2] = 2 + 1 = 3`                                                                       | `[1, 3, 3, 1]`    | `[[1], [1, 1], [1, 2, 1], [1, 3, 3, 1]]`                  |
| 5    | 4           | `[1, 1, 1, 1, 1]` | `range(1, 4)` → `[1, 2, 3]`          | **j = 1** → `row[1] = triangle[3][0] + triangle[3][1] = 1 + 3 = 4`<br>**j = 2** → `row[2] = triangle[3][1] + triangle[3][2] = 3 + 3 = 6`<br>**j = 3** → `row[3] = triangle[3][2] + triangle[3][3] = 3 + 1 = 4` | `[1, 4, 6, 4, 1]` | `[[1], [1, 1], [1, 2, 1], [1, 3, 3, 1], [1, 4, 6, 4, 1]]` |



💡 Key Notes:

range(1, i) → gives indices of middle elements only.

First (j=0) and last (j=i) elements are always 1.

Each row[j] = triangle[i-1][j-1] + triangle[i-1][j].

So the inner loop builds the middle part of each row.

```python
class Solution:
    def generate(self, numRows: int) -> list[list[int]]:
        triangle = []
        for i in range(numRows):
            row = [1] * (i + 1)
            for j in range(1, i):  
                row[j]= triangle[i - 1][j - 1] + triangle[i - 1][j]
            triangle.append(row)
        return triangle        
# Example usage:
print(Solution().generate(5))
```

| ⏱️ **Time Complexity**                                                    | 🧠 **Space Complexity**                                                                     |
| ------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------- |
| **`triangle = []`** → O(1)<br>initialize empty list                       | **`triangle = []`** → O(1)<br>empty list                                                    |
| **Outer loop** → O(numRows²)<br>nested loop fills rows                    | **Outer loop** → triangle grows row by row<br>total elements stored = 1 + 2 + … + n = O(n²) |
| **`row = [1] * (i + 1)`** → O(numRows²)<br>fills row with 1s              | **`row` list** → size = i + 1<br>stored temporarily → O(i)                                  |
| **Inner loop `for j in range(1,i)`** → O(numRows²)<br>computes sums       | Inner loop uses **row[j]** → in-place → O(1) extra                                          |
| **`row[j] = triangle[i-1][j-1] + triangle[i-1][j]`** → O(1) per iteration | Updates row in place → no extra memory                                                      |
| **`triangle.append(row)`** → O(1) amortized                               | Adds row to triangle → memory accumulates → O(n²) total                                     |
| **`return triangle`** → O(1)                                              | Return reference → no extra memory                                                          |
| **✅ Final Time Complexity:** O(n²)<br>nested loop over row length         | **✅ Total Space Complexity:** O(n²)<br>triangle stores all values                           |


# 10 -485. Max Consecutive Ones

You are given a binary array (only contains 0s and 1s). You need to find the longest sequence of consecutive 1's in the array.

Input: [1,1,0,1,1,1]

Groups: [1,1] + [1,1,1]

Longest: 3

Output: 3

Input: nums = [1, 1, 0, 1, 1, 1]

| Step | num | count | max_count BEFORE | max(max_count, count) | max_count AFTER | Explanation  |
| ---- | --- | ----- | ---------------- | --------------------- | --------------- | ------------ |
| 1    | 1   | 1     | 0                | max(0,1)=1            | 1               | first streak |
| 2    | 1   | 2     | 1                | max(1,2)=2            | 2               | new best     |
| 3    | 0   | 0     | 2                | —                     | 2               | reset only   |
| 4    | 1   | 1     | 2                | max(2,1)=2            | 2               | best stays   |
| 5    | 1   | 2     | 2                | max(2,2)=2            | 2               | same         |
| 6    | 1   | 3     | 2                | max(2,3)=3            | 3               | new best     |



```python
class Solution:
    def findMaxConsecutiveOnes(self, nums):
        max_count = 0 #2
        count = 0 #2

        for num in nums:
            if num == 1:
                count += 1
                max_count = max(max_count, count)
            else:
                count = 0
        return max_count            

       
# Example 1
nums = [1,1,0,1,1,1]
print(Solution().findMaxConsecutiveOnes(nums))  # Output: 3          
```

| ⏱️ **Time Complexity**                                     | 🧠 **Space Complexity**                                    |
| ---------------------------------------------------------- | ---------------------------------------------------------- |
| **Initialize variables** → O(1)<br>constant setup          | **`max_count` variable** → O(1)<br>stores best streak      |
| **Loop through array** → O(n)<br>visits each element once  | **`count` variable** → O(1)<br>stores current streak       |
| **Comparison (`num == 1`)** → O(1)<br>constant check       | **loop variable `num`** → O(1)<br>reused each iteration    |
| **Increment `count`** → O(1)<br>simple arithmetic          | **no extra data structures** → O(1)<br>in-place processing |
| **Update max using `max()`** → O(1)<br>constant comparison | **return value** → O(1)<br>existing value returned         |
| **Reset `count`** → O(1)<br>constant                       |                                                            |
| **Return result** → O(1)                                   |                                                            |
| **👉 Final: O(n)**<br>single linear pass                   | **👉 Auxiliary Space: O(1)**<br>**👉 Total Space: O(1)**   |
