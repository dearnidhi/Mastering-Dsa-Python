# 1Ô∏è‚É£ 260 ‚Äì Single Number III

freq[num] = freq.get(num, 0) + 1
- freq.get(num, 0)	=> Take old count of num, or 0 if it‚Äôs not there
-  +1	Add one for this new occurrence
- freq[num] = ...	Save updated count back

‚úÖ In short:

If number exists ‚Üí increase count by 1, else start from 1.

### üß† Dry Run ‚Äî nums = [1, 2, 1, 3, 2, 5]
| **Step / Iteration** | **Operation / Code Line** | **Formula Applied**            | **Explanation (What‚Äôs Happening)**                                         | **Dictionary (`freq`) State** |
| -------------------- | ------------------------- | ------------------------------ | -------------------------------------------------------------------------- | ----------------------------- |
| **Step 1**           | Initialize                | ‚Äî                              | Start with an **empty dictionary** to store number frequencies.            | `{}`                          |
| **Iteration 1**      | `num = 1`                 | `freq[1] = freq.get(1, 0) + 1` | `get(1, 0)` ‚Üí returns `0` (since 1 not found) ‚Üí `0 + 1 = 1` ‚Üí add `{1: 1}` | `{1: 1}`                      |
| **Iteration 2**      | `num = 2`                 | `freq[2] = freq.get(2, 0) + 1` | `get(2, 0)` ‚Üí `0` ‚Üí `0 + 1 = 1` ‚Üí add `{2: 1}`                             | `{1: 1, 2: 1}`                |
| **Iteration 3**      | `num = 1`                 | `freq[1] = freq.get(1, 0) + 1` | `get(1, 0)` ‚Üí `1` (already exists) ‚Üí `1 + 1 = 2` ‚Üí update `{1: 2}`         | `{1: 2, 2: 1}`                |
| **Iteration 4**      | `num = 3`                 | `freq[3] = freq.get(3, 0) + 1` | `get(3, 0)` ‚Üí `0` ‚Üí `0 + 1 = 1` ‚Üí add `{3: 1}`                             | `{1: 2, 2: 1, 3: 1}`          |
| **Iteration 5**      | `num = 2`                 | `freq[2] = freq.get(2, 0) + 1` | `get(2, 0)` ‚Üí `1` ‚Üí `1 + 1 = 2` ‚Üí update `{2: 2}`                          | `{1: 2, 2: 2, 3: 1}`          |
| **Iteration 6**      | `num = 5`                 | `freq[5] = freq.get(5, 0) + 1` | `get(5, 0)` ‚Üí `0` ‚Üí `0 + 1 = 1` ‚Üí add `{5: 1}`                             | `{1: 2, 2: 2, 3: 1, 5: 1}`    |


### ü™Ñ Step 3 ‚Äî Filter numbers with count == 1

   Now use the list comprehension:

  [num for num, count in freq.items() if count == 1]

| **Pair (num, count)** | **Condition (count == 1?)** | **Include?** |
| --------------------- | --------------------------- | ------------ |
| (1, 2)                | ‚ùå No                        | Skip         |
| (2, 2)                | ‚ùå No                        | Skip         |
| (3, 1)                | ‚úÖ Yes                       | Include      |
| (5, 1)                | ‚úÖ Yes                       | Include      |


‚úÖ Output: [3, 5]



````python
class Solution:
    def singleNumber(self,nums):
        freq = {} #key=Numbers, value = count
        for num in nums:
            freq[num] = freq.get(num, 0) + 1
        return [num for num, count in freq.items() if count ==1 ]   
print(Solution().singleNumber([1, 2, 1, 3, 2, 5]))
# Output: [3, 5]     
````


# 2Ô∏è‚É£ 720 ‚Äì Longest Word in Dictionary


['a', 'ap', 'app', 'appl', 'apple', 'apply']

valid = set()

result = ""


| Word      | Prefix (`word[:-1]`) | Condition True? | Action        | Updated `valid`                    | Updated `result`      | Reason                                 |
| --------- | -------------------- | --------------- | ------------- | ---------------------------------- | --------------------- | -------------------------------------- |
| `'a'`     | `''`                 | ‚úÖ (length 1)    | Add `'a'`     | `{a}`                              | `'a'`                 | First single-letter word               |
| `'ap'`    | `'a'`                | ‚úÖ               | Add `'ap'`    | `{a, ap}`                          | `'ap'`                | `'a'` already valid                    |
| `'app'`   | `'ap'`               | ‚úÖ               | Add `'app'`   | `{a, ap, app}`                     | `'app'`               | `'ap'` valid, longer word              |
| `'appl'`  | `'app'`              | ‚úÖ               | Add `'appl'`  | `{a, ap, app, appl}`               | `'appl'`              | `'app'` valid                          |
| `'apple'` | `'appl'`             | ‚úÖ               | Add `'apple'` | `{a, ap, app, appl, apple}`        | `'apple'`             | `'appl'` valid                         |
| `'apply'` | `'appl'`             | ‚úÖ               | Add `'apply'` | `{a, ap, app, appl, apple, apply}` | `'apple'` (no change) | Equal length but `'apple'` < `'apply'` |


````python
from typing import List
class Solution:
    def longestWord(self, words:List[str])->str:
        words.sort()
        valid = set()
        result = " "
        for word in words:
            if len(word)==1 or word[:-1] in valid:
                valid.add(word)
                if len(word) > len(result) or (len(word) == len(result) and word < result):
                    result = word
        return result 

# Correct usage examples:
print(Solution().longestWord(["a", "ap", "app", "appl", "apple", "application"]))  # Multiple words    
print(Solution().longestWord(["w", "wo", "wor", "worl", "world"]))  # Example from problem 
````

diff_count = 0

 [('h','h'), ('e','h'), ('l','l'), ('l','l'), ('o','o')]


| Index | word[i] | searchWord[i] | Same? | diff_count |
| ----- | ------- | ------------- | ----- | ---------- |
| 0     | h       | h             | ‚úÖ     | 0          |
| 1     | e       | h             | ‚ùå     | 1          |
| 2     | l       | l             | ‚úÖ     | 1          |
| 3     | l       | l             | ‚úÖ     | 1          |
| 4     | o       | o             | ‚úÖ     | 1          |


‚úÖ diff_count == 1 ‚Üí return True

True

So "hhllo" matches "hello" after changing exactly one character (‚Äòe‚Äô ‚Üí ‚Äòh‚Äô).

````python
class MagicDictionary:
    def __init__(self):
        self.words = []
    def buildDict(self, dictionary:list[str])-> None:
        self.words = dictionary
    def search(self, searchWord:str)->bool:
        for word in self.words:
            if len(word) != len(searchWord):
                continue
            diff_count = 0
            for a, b in zip(word, searchWord):  
                if a != b:
                    diff_count += 1
                    if diff_count > 1:
                        break
            if diff_count == 1:
                return True
        return False    

magic = MagicDictionary()
magic.buildDict(["hello", "leetcode"])
magic.search("hhllo")              

    
````

# üîç **Word Break Dry Run with Memoization**

**Example:**  
`s = "catdog"`, `wordDict = ["cat", "dog"]`

---

Line 17: call canForm("catdog")  

  Line 4: "catdog" not in memo  
  Line 6: "catdog" not in word_set  

  Line 9: i=1 ‚Üí prefix="c", suffix="atdog" ‚Üí "c" not in word_set  
  Line 9: i=2 ‚Üí prefix="ca", suffix="tdog" ‚Üí "ca" not in word_set  
  Line 9: i=3 ‚Üí prefix="cat", suffix="dog" ‚Üí "cat" in word_set ‚úÖ  

    Line 12: call canForm("dog") [RECURSION]  
    
      Line 4: "dog" not in memo  
      Line 6: "dog" in word_set ‚úÖ  
        Line 7: memo["dog"] = True  
        Line 8: return True  

  Line 12: True and True ‚Üí  
    Line 13: memo["catdog"] = True  
    Line 14: return True  

---

üìä **Final Memo State:**  
```python
memo = {
    "dog": True,
    "catdog": True
}



**üéØ Simple Flow:** Check memo ‚Üí Check whole word ‚Üí Try all splits ‚Üí Store result ‚Üí Return

````python
from typing import List

class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        word_set = set(wordDict)           # Line 1
        memo = {}                          # Line 2
        
        def canForm(word):                 # Line 3
            if word in memo:               # Line 4
                return memo[word]          # Line 5
                
            if word in word_set:           # Line 6
                memo[word] = True          # Line 7
                return True                # Line 8
            
            for i in range(1, len(word)):  # Line 9
                prefix = word[:i]          # Line 10
                suffix = word[i:]          # Line 11
                
                if prefix in word_set and canForm(suffix):  # Line 12
                    memo[word] = True      # Line 13
                    return True            # Line 14
            
            memo[word] = False             # Line 15
            return False                   # Line 16
        
        return canForm(s)                  # Line 17
    
sol = Solution()
print(sol.wordBreak("leetcode", ["leet", "code"]))  # True
print(sol.wordBreak("applepenapple", ["apple", "pen"]))  # True
print(sol.wordBreak("catsandog", ["cats","dog","sand","and","cat"]))  # False    
````

# 5Ô∏è‚É£ 884 ‚Äì Uncommon Words from Two Sentences


## üß© Problem
You are given two sentences `s1` and `s2`.  A **word** is *uncommon* if it appears **exactly once** in one sentence  
and **does not appear** in the other sentence.

You must return a list of all such uncommon words.

---

## üß† Example
**Input:**
```python
s1 = "this apple is sweet"
s2 = "this apple is sour"

Output:

["sweet", "sour"]

üßæ Logic (Simple Steps)

Split both sentences into words.

Count the occurrences of each word using a dictionary or collections.Counter.

Return words that appear exactly once in total.

### üîç Dry Run

**s1 = "this apple is sweet"**  
**s2 = "this apple is sour"**

---

### **Combined words ‚Üí**
["this", "apple", "is", "sweet", "this", "apple", "is", "sour"]


---

### **Count ‚Üí**
```python
{
  "this": 2,
  "apple": 2,
  "is": 2,
  "sweet": 1,
  "sour": 1
}

Filter freq == 1 ‚Üí

‚úÖ ["sweet", "sour"]

‚úÖ Final Output:

["sweet", "sour"]

````python
from typing import List
from collections import Counter
class Solution:
    def uncommonFromSentences(self, s1:str, s2: str)->List[str]:
        words = (s1 +" " + s2).split()
        count = Counter(words)
        return [word for word , freq in count.items() if freq ==1]
    

# üîç Example usage
sol = Solution()
s1 = "this apple is sweet"
s2 = "this apple is sour"
result = sol.uncommonFromSentences(s1, s2)
print(result)  # ‚úÖ Output: ['sweet', 'sour']       
````
