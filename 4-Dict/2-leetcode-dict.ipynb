# 1ï¸âƒ£ 260 â€“ Single Number III

freq[num] = freq.get(num, 0) + 1
- freq.get(num, 0)	=> Take old count of num, or 0 if itâ€™s not there
-  +1	Add one for this new occurrence
- freq[num] = ...	Save updated count back

âœ… In short:

If number exists â†’ increase count by 1, else start from 1.

### ðŸ§  Dry Run â€” nums = [1, 2, 1, 3, 2, 5]
| **Step / Iteration** | **Operation / Code Line** | **Formula Applied**            | **Explanation (Whatâ€™s Happening)**                                         | **Dictionary (`freq`) State** |
| -------------------- | ------------------------- | ------------------------------ | -------------------------------------------------------------------------- | ----------------------------- |
| **Step 1**           | Initialize                | â€”                              | Start with an **empty dictionary** to store number frequencies.            | `{}`                          |
| **Iteration 1**      | `num = 1`                 | `freq[1] = freq.get(1, 0) + 1` | `get(1, 0)` â†’ returns `0` (since 1 not found) â†’ `0 + 1 = 1` â†’ add `{1: 1}` | `{1: 1}`                      |
| **Iteration 2**      | `num = 2`                 | `freq[2] = freq.get(2, 0) + 1` | `get(2, 0)` â†’ `0` â†’ `0 + 1 = 1` â†’ add `{2: 1}`                             | `{1: 1, 2: 1}`                |
| **Iteration 3**      | `num = 1`                 | `freq[1] = freq.get(1, 0) + 1` | `get(1, 0)` â†’ `1` (already exists) â†’ `1 + 1 = 2` â†’ update `{1: 2}`         | `{1: 2, 2: 1}`                |
| **Iteration 4**      | `num = 3`                 | `freq[3] = freq.get(3, 0) + 1` | `get(3, 0)` â†’ `0` â†’ `0 + 1 = 1` â†’ add `{3: 1}`                             | `{1: 2, 2: 1, 3: 1}`          |
| **Iteration 5**      | `num = 2`                 | `freq[2] = freq.get(2, 0) + 1` | `get(2, 0)` â†’ `1` â†’ `1 + 1 = 2` â†’ update `{2: 2}`                          | `{1: 2, 2: 2, 3: 1}`          |
| **Iteration 6**      | `num = 5`                 | `freq[5] = freq.get(5, 0) + 1` | `get(5, 0)` â†’ `0` â†’ `0 + 1 = 1` â†’ add `{5: 1}`                             | `{1: 2, 2: 2, 3: 1, 5: 1}`    |


### ðŸª„ Step 3 â€” Filter numbers with count == 1

   Now use the list comprehension:

  [num for num, count in freq.items() if count == 1]

| **Pair (num, count)** | **Condition (count == 1?)** | **Include?** |
| --------------------- | --------------------------- | ------------ |
| (1, 2)                | âŒ No                        | Skip         |
| (2, 2)                | âŒ No                        | Skip         |
| (3, 1)                | âœ… Yes                       | Include      |
| (5, 1)                | âœ… Yes                       | Include      |


âœ… Output: [3, 5]



```python
class Solution:
    def singleNumber(self,nums):
        freq = {} #key=Numbers, value = count
        for num in nums:
            freq[num] = freq.get(num, 0) + 1
        return [num for num, count in freq.items() if count ==1 ]   
print(Solution().singleNumber([1, 2, 1, 3, 2, 5]))
# Output: [3, 5]     
   

```


# 2ï¸âƒ£ 720 â€“ Longest Word in Dictionary


```python
class Solution:
    def longestWord(self, words: List[str]) -> str:
        words.sort()
        valid = set()
        result = ""
        
        for word in words:
            if len(word) == 1 or word[:-1] in valid:
                valid.add(word)
                if len(word) > len(result) or (len(word) == len(result) and word < result):
                    result = word
        return result
```
