# 1️⃣ 260 – Single Number III

freq[num] = freq.get(num, 0) + 1
- freq.get(num, 0)	=> Take old count of num, or 0 if it’s not there
-  +1	Add one for this new occurrence
- freq[num] = ...	Save updated count back

✅ In short:

If number exists → increase count by 1, else start from 1.

### 🧠 Dry Run — nums = [1, 2, 1, 3, 2, 5]
| **Step / Iteration** | **Operation / Code Line** | **Formula Applied**            | **Explanation (What’s Happening)**                                         | **Dictionary (`freq`) State** |
| -------------------- | ------------------------- | ------------------------------ | -------------------------------------------------------------------------- | ----------------------------- |
| **Step 1**           | Initialize                | —                              | Start with an **empty dictionary** to store number frequencies.            | `{}`                          |
| **Iteration 1**      | `num = 1`                 | `freq[1] = freq.get(1, 0) + 1` | `get(1, 0)` → returns `0` (since 1 not found) → `0 + 1 = 1` → add `{1: 1}` | `{1: 1}`                      |
| **Iteration 2**      | `num = 2`                 | `freq[2] = freq.get(2, 0) + 1` | `get(2, 0)` → `0` → `0 + 1 = 1` → add `{2: 1}`                             | `{1: 1, 2: 1}`                |
| **Iteration 3**      | `num = 1`                 | `freq[1] = freq.get(1, 0) + 1` | `get(1, 0)` → `1` (already exists) → `1 + 1 = 2` → update `{1: 2}`         | `{1: 2, 2: 1}`                |
| **Iteration 4**      | `num = 3`                 | `freq[3] = freq.get(3, 0) + 1` | `get(3, 0)` → `0` → `0 + 1 = 1` → add `{3: 1}`                             | `{1: 2, 2: 1, 3: 1}`          |
| **Iteration 5**      | `num = 2`                 | `freq[2] = freq.get(2, 0) + 1` | `get(2, 0)` → `1` → `1 + 1 = 2` → update `{2: 2}`                          | `{1: 2, 2: 2, 3: 1}`          |
| **Iteration 6**      | `num = 5`                 | `freq[5] = freq.get(5, 0) + 1` | `get(5, 0)` → `0` → `0 + 1 = 1` → add `{5: 1}`                             | `{1: 2, 2: 2, 3: 1, 5: 1}`    |


### 🪄 Step 3 — Filter numbers with count == 1

   Now use the list comprehension:

  [num for num, count in freq.items() if count == 1]

| **Pair (num, count)** | **Condition (count == 1?)** | **Include?** |
| --------------------- | --------------------------- | ------------ |
| (1, 2)                | ❌ No                        | Skip         |
| (2, 2)                | ❌ No                        | Skip         |
| (3, 1)                | ✅ Yes                       | Include      |
| (5, 1)                | ✅ Yes                       | Include      |


✅ Output: [3, 5]



````python
class Solution:
    def singleNumber(self,nums):
        freq = {} #key=Numbers, value = count
        #{1:2, 2:1,3:1, 5:1}
        for num in nums:#1
            freq[num] = freq.get(num, 0) + 1 #1
            #freq[1] = freq.get(1,0)+1 = 0 +1 =1
            #freq[2] = freq.get(2,0)+1 = 0 +1 =1
            #freq[1] = freq.get(1,0)+1 = 1 +1 =2
            #freq[3] = freq.get(3,0)+1 = 0 +1 =1
        return [num for num , count in freq.items() if count ==1]         
print(Solution().singleNumber([1, 2, 1, 3, 2, 5]))
# Output: [3, 5]     
````


# 2️⃣ 720 – Longest Word in Dictionary


['a', 'ap', 'app', 'appl', 'apple', 'apply']

valid = set()

result = ""


| Word      | Prefix (`word[:-1]`) | Condition True? | Action        | Updated `valid`                    | Updated `result`      | Reason                                 |
| --------- | -------------------- | --------------- | ------------- | ---------------------------------- | --------------------- | -------------------------------------- |
| `'a'`     | `''`                 | ✅ (length 1)    | Add `'a'`     | `{a}`                              | `'a'`                 | First single-letter word               |
| `'ap'`    | `'a'`                | ✅               | Add `'ap'`    | `{a, ap}`                          | `'ap'`                | `'a'` already valid                    |
| `'app'`   | `'ap'`               | ✅               | Add `'app'`   | `{a, ap, app}`                     | `'app'`               | `'ap'` valid, longer word              |
| `'appl'`  | `'app'`              | ✅               | Add `'appl'`  | `{a, ap, app, appl}`               | `'appl'`              | `'app'` valid                          |
| `'apple'` | `'appl'`             | ✅               | Add `'apple'` | `{a, ap, app, appl, apple}`        | `'apple'`             | `'appl'` valid                         |
| `'apply'` | `'appl'`             | ✅               | Add `'apply'` | `{a, ap, app, appl, apple, apply}` | `'apple'` (no change) | Equal length but `'apple'` < `'apply'` |


````python
from typing import List
class Solution:
    def longestWord(self, words:List[str])->str:
        words.sort()
        valid = set()
        result = ""

        for word in words:
            if len(word) == 1 or word[:-1] in valid:
                valid.add(word)
                if len(word) > len(result) or (len(word) == len(result) and word < result):
                    result = word
        return result            
# Correct usage examples:
print(Solution().longestWord(["a", "ap", "app", "appl", "apple"]))  # Multiple words    
#print(Solution().longestWord(["w", "wo", "wor", "worl", "world"]))  # Example from problem 
````

# 3️⃣ 676 – Implement Magic Dictionary


🔎 Search for "hhllo"

We loop through each word in self.words.

1️⃣ Compare with "hello"

✅ Both have length 5 → continue

diff_count = 0

zip("hello", "hhllo")

 [('h','h'), ('e','h'), ('l','l'), ('l','l'), ('o','o')]


| Index | word[i] | searchWord[i] | Same? | diff_count |
| ----- | ------- | ------------- | ----- | ---------- |
| 0     | h       | h             | ✅     | 0          |
| 1     | e       | h             | ❌     | 1          |
| 2     | l       | l             | ✅     | 1          |
| 3     | l       | l             | ✅     | 1          |
| 4     | o       | o             | ✅     | 1          |


✅ diff_count == 1 → return True

True

So "hhllo" matches "hello" after changing exactly one character (‘e’ → ‘h’).

````python
class MagicDictionary:
    def __init__(self):
        self.words = []
    def buildDict(self, dictionary:list[str])-> None:
        self.words = dictionary
    def search(self, searchWord:str)->bool:
        for word in self.words:
            if len(word) != len(searchWord):
                continue
            diff_count = 0
            for a, b in zip(word, searchWord):
                if a !=b:
                    diff_count += 1
                    if diff_count > 1:
                        break
                    if diff_count == 1:
                        return True
        return False                
magic = MagicDictionary()
magic.buildDict(["hello", "leetcode"])
magic.search("hhllo")              

    
````


# 4️⃣ 139 – Word Break

Given a string s and a list of words wordDict, return True if s can be segmented into a sequence of one or more dictionary words.


# ✅ Word Break – Proper Dry Run with j Explanation

## Input

s = "cat"
wordDict = ["ca", "t"]

word_set = {"ca", "t"}
dp = [True, False, False, False]

---

## 🔹 Important Understanding

- i = length of prefix we are trying to validate
- j = possible split position
- s[j:i] = substring starting at index j and ending BEFORE i

Python slicing rule:
s[start:end] → includes start, excludes end

---

## String Index Reference

Index:   0   1   2  
Chars:   c   a   t  

Prefixes:  
i = 1 → "c"  
i = 2 → "ca"  
i = 3 → "cat"  

---

# 🔁 Step-by-Step Execution (Single Combined Table)

| i | range(i) gives | j | dp[j] | s[j:i] | In Dictionary? | Action | dp After |
|---|---------------|---|--------|--------|----------------|--------|----------|
| 1 | [0] | 0 | True  | "c"   | No  | dp[1] = False | [T, F, F, F] |
| 2 | [0,1] | 0 | True  | "ca"  | Yes | dp[2] = True (break) | [T, F, T, F] |
| 3 | [0,1,2] | 0 | True  | "cat" | No  | Continue | - |
| 3 | [0,1,2] | 1 | False | "at"  | No  | Skip (dp[1] is False) | - |
| 3 | [0,1,2] | 2 | True  | "t"   | Yes | dp[3] = True (break) | [T, F, T, T] |

---

# 📌 Final dp

Index:   0     1     2     3  
Prefix: ""    "c"   "ca"  "cat"  
dp:     [T,    F,    T,    T]

---

# 🔎 Why j = 1 Came?

Because:

For i = 3  
range(3) → [0, 1, 2]

That is how Python works.  
range(i) means numbers from 0 to i-1.

---

# 🔑 Core Condition Meaning

if dp[j] and s[j:i] in word_set

Means:

Left part (0 to j) is already valid  
AND  
Right part (j to i) is a dictionary word  

If both true → entire prefix 0 to i becomes valid.

````python
class Solution(object):
    def wordBreak(self, s, wordDict):
        word_set = set(wordDict)
        dp = [True] + [False] * len(s)
        #dp = [T] +[F] * (3)= T F F F
        for i in range(1, len(s)+1): #i =(1, 4)
            for j in range(i):
                if dp[j] and s[j:i] in word_set:
                    dp[i]= True
                    break
        return dp[-1]        
sol = Solution()
print(sol.wordBreak("leetcode", ["leet", "code"]))  # True        
````

# 5️⃣ 884 – Uncommon Words from Two Sentences


## 🧩 Problem
You are given two sentences `s1` and `s2`.  A **word** is *uncommon* if it appears **exactly once** in one sentence  
and **does not appear** in the other sentence.

You must return a list of all such uncommon words.

---

## 🧠 Example
**Input:**
```python
s1 = "this apple is sweet"
s2 = "this apple is sour"

Output:

["sweet", "sour"]

🧾 Logic (Simple Steps)

Split both sentences into words.

Count the occurrences of each word using a dictionary or collections.Counter.

Return words that appear exactly once in total.

### 🔍 Dry Run

**s1 = "this apple is sweet"**  
**s2 = "this apple is sour"**

---

### **Combined words →**
["this", "apple", "is", "sweet", "this", "apple", "is", "sour"]


---

### **Count →**
```python
{
  "this": 2,
  "apple": 2,
  "is": 2,
  "sweet": 1,
  "sour": 1
}

Filter freq == 1 →

✅ ["sweet", "sour"]

✅ Final Output:

["sweet", "sour"]

````python
from typing import List
from collections import Counter
class Solution:
    def uncommonFromSentences(self, s1:str, s2: str)->List[str]:
        words = (s1 + " " + s2).split()
        count = Counter(words)
        return [word for word , freq in count.items() if freq == 1]    

# 🔍 Example usage
sol = Solution()
s1 = "this apple is sweet"
s2 = "this apple is sour"
result = sol.uncommonFromSentences(s1, s2)
print(result)  # ✅ Output: ['sweet', 'sour']       
````
