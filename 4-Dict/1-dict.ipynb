# Dict
- Dictionaries are used to store data values in key:value pairs.
- A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
- It holds a key: value pair in which we can easily access a value if the key is known. 
- A dictionary (also called a hash map) is a key-value data structure.

📦 Think of it like a locker system:

Each key = locker number

Each value = item stored inside

👉 You don’t need to search linearly — just use the key to directly find the value.

### ⚡ 3. Add / Update / Delete

```python
# Add new key-value

student = {"name":"Nidhi", "age":25, "Course":"DSA"}
student["city"] ="goa"
# Update
student["age"] = 21

# Delete
del student["Course"]
# {'name': 'Nidhi', 'age': 22, 'city': 'Bengaluru'}
print(student)
```

```python
#  Update / add an element to the dictionary


my_dict = {"name" :"eddy", "age": 29}
my_dict["address"] = "London"
print(my_dict)
```

```python

#  Traverse through a dictionary
def traversDict(dict):
    for key in dict:
        print(key, dict[key])
traversDict(my_dict)        

       
```

```python
#  Searching a dictionary
def searchDict(dict, value):
    for key in dict: #name
        if dict[key] == value:
            return key , value
    return "value not found"        

#Test
newdict = {'name':'edy', 'age':29, 'address':'london'}
#  Delete or remove a dictionary

print(searchDict(newdict, 29))

```

```python
# sorted method
my_D = {'eooo':1, 'aas':2, 'add':3, 'sseo':4, 'werwi':5}
print(sorted(my_D, key = len))

```

# ⚡ Difference Between dict and HashMap in Python (DSA Context)
| Point | `dict`                                 | `HashMap`                                          |
| ----- | -------------------------------------- | -------------------------------------------------- |
| 1️⃣   | Built-in Python data structure         | Custom class made while learning DSA               |
| 2️⃣   | Already optimized and implemented in C | Manually implemented by you (using list + hashing) |
| 3️⃣   | Used directly in coding                | Used to **understand how hashing works**           |
| 4️⃣   | Handles collisions internally          | You must handle collisions (like chaining)         |
| 5️⃣   | Used in real projects                  | Used mainly in DSA learning & interviews           |


```python
# dict example
my_dict = {"name": "Nidhi", "age": 21}
print(my_dict["name"])   # Output: Nidhi
```

```python
# HashMap example
class HashMap:
    def __init__(self):
        self.size = 10
        self.map = [None] * self.size
    
    def _get_hash(self, key):
        return hash(key) % self.size
```

🧠 Summary:

➡️ dict = ready-made, built-in hash table
