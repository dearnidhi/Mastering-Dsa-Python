# Arrays
- Dictionaries are used to store data values in key:value pairs.
- A dictionary is a collection which is ordered*, changeable and do not allow duplicates.
- It holds a key: value pair in which we can easily access a value if the key is known. 
- A dictionary (also called a hash map) is a key-value data structure.

üì¶ Think of it like a locker system:

Each key = locker number

Each value = item stored inside

üëâ You don‚Äôt need to search linearly ‚Äî just use the key to directly find the value.

### ‚öôÔ∏è 2. Creating a Dictionary

```python
# empty dictionary
student = {}

# with values
student = {"name": "Nidhi", "age": 21, "course": "DSA"}
```

```python
### Access by key:
print(student["name"])   # Output: Nidhi
```

### ‚ö° 3. Add / Update / Delete

```python
# Add new key-value
student["city"] = "Bengaluru"

# Update
student["age"] = 22

# Delete
del student["course"]

print(student)
# {'name': 'Nidhi', 'age': 22, 'city': 'Bengaluru'}
```

```python
#  Update / add an element to the dictionary

myDict = {'name': 'Edy', 'age': 26}
myDict['address'] = 'London'
print(myDict)
```

```python

#  Traverse through a dictionary
def traverseDict(dict):
    for key in dict:
        print(key, dict[key])

traverseDict(myDict)
```

```python
#  Searching a dictionary
def searchDict(dict, value):
    for key in dict:
        if dict[key] == value:
            return key, value
    return 'The value does not exist'
print(searchDict(myDict, 27))

#  Delete or remove a dictionary

myDict.pop('name')
```

```python
# sorted method
myDict = {'eooooa': 1, 'aas': 2, 'udd': 3, 'sseo': 4, 'werwi': 5}
print(sorted(myDict, key=len))
```

```python
# HashMap example
class HashMap:
    def __init__(self):
        self.size = 10
        self.map = [None] * self.size
    
    def _get_hash(self, key):
        return hash(key) % self.size
```
