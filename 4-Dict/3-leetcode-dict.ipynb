#  6️⃣ 953 – Verifying an Alien Dictionary


### 👽 **Problem Statement – Verifying an Alien Dictionary**

You are given:

- A list of words: **`words`**  
- A string: **`order`**, which represents the **alphabetical order** of characters in an alien language.

Your task is to determine **if the given list of words is sorted** according to this alien dictionary order.

---

#### ✅ **Rules:**
1. Each word must appear **lexicographically before or equal to** the next word, based on the given `order`.  
2. If two words share a prefix, the **shorter word** should come first (e.g., `"app"` should come before `"apple"`).  

---

#### 🧠 **Return:**
- **`True`** → if the words are sorted correctly.  
- **`False`** → if the order is violated anywhere.

---

#### 🧪 **Example:**
```python
isAlienSorted(["hello", "leetcode"], "hlabcdefgijkmnopqrstuvwxyz") ➜ True  
isAlienSorted(["word", "world", "row"], "worldabcefghijkmnpqstuvxyz") ➜ False  
isAlienSorted(["apple", "app"], "abcdefghijklmnopqrstuvwxyz") ➜ False


Lexicographical = compare letter by letter using a given alphabet order (normal or alien).


## **Input:**
- `words = ["hello", "leetcode"]`
- `order = "hlabcdefgijkmnopqrstuvwxyz"`

---

## **Step 1: Create Rank Dictionary**
```python
rank = {ch: i for i, ch in enumerate(order)}
# rank = {'h':0, 'l':1, 'a':2, 'b':3, 'c':4, ...}

👉 Means: 'h' comes before 'l', 'l' before 'a', etc.
This gives every character a numeric rank according to its order in the alien language.

| Index | w1[j] | w2[j] | rank[w1[j]] | rank[w2[j]] | Comparison | Result                  |
| ----- | ----- | ----- | ----------- | ----------- | ---------- | ----------------------- |
| 0     | h     | l     | 0           | 1           | 0 < 1      | ✅ correct order → break |


✅ Since 'h' comes before 'l', words are in correct order.
No more pairs → return True.

Output: True

#### 🧩 Example 2

**Input:**
```python
words = ["word", "world", "row"]
order = "worldabcefghijkmnpqstuvxyz"

rank = {'w':0, 'o':1, 'r':2, 'l':3, 'd':4, 'a':5, ...}

Compare ("word", "world")


| j | w1[j] | w2[j] | rank[w1[j]] | rank[w2[j]] | Result                |
| - | ----- | ----- | ----------- | ----------- | --------------------- |
| 0 | w     | w     | same        | same        | continue              |
| 1 | o     | o     | same        | same        | continue              |
| 2 | r     | r     | same        | same        | continue              |
| 3 | d     | l     | 4           | 3           | ❌ wrong order (4 > 3) |

👉 'd' comes after 'l' in alien order → ❌ not sorted → return False.

Output: False

#### ❌ Example 3
words = ["apple", "app"], order = "abcdefghijklmnopqrstuvwxyz"

Compare ("apple", "app")

Check letter by letter: a == a, p == p, p == p

No difference found (loop ended)

Then check:

if len(w1) > len(w2):

    return False

✅ len("apple") = 5 and len("app") = 3 → longer word first ❌ (prefix rule broken)

Output: False    



| Words                    | Order      | Output  | Why                                 |
| ------------------------ | ---------- | ------- | ----------------------------------- |
| ["hello", "leetcode"]    | "hl..."    | ✅ True  | ‘h’ comes before ‘l’                |
| ["word", "world", "row"] | "world..." | ❌ False | ‘d’ comes after ‘l’                 |
| ["apple", "app"]         | "abc..."   | ❌ False | longer word comes before its prefix |


````python
def isAlienSorted(words, order):
    # create dictionary for order lookup
    rank = {ch: i for i, ch in enumerate(order)}
    
    # check each consecutive pair
    for i in range(len(words)- 1):
        w1, w2  = words[i], words[i+1]
        for j in range(min(len(w1),len(w2))): #5
            if w1[j] != w2[j]: # h != l
                if rank[w1[j]] > rank[w2[j]]:# 0> 1
                    return False
                break
            else:
                if len(w1) > len(w2):
                    return False
    return True                    
# ✅ Example
print(isAlienSorted(["hello", "leetcode"], "hlabcdefgijkmnopqrstuvwxyz"))  # True
print(isAlienSorted(["word", "world", "row"], "worldabcefghijkmnpqstuvxyz"))  # False
print(isAlienSorted(["apple", "app"], "abcdefghijklmnopqrstuvwxyz"))  # False
````

# 7️⃣ 127 – Word Ladder


### 🧩 **Problem Statement – Word Ladder**

You are given:

- Two words: **`beginWord`** and **`endWord`**  
- A list of words: **`wordList`**

Your task is to **find the length of the shortest transformation sequence** from `beginWord` to `endWord`, such that:

1. Only **one letter** can be changed at a time.  
2. Each transformed word **must exist in `wordList`**.

Return the **number of words** in the shortest transformation sequence,  
**including both `beginWord` and `endWord`**.  

If no such transformation exists, **return `0`**.

### 🧩Dry Run 

### **Input**
```python
beginWord = "hit"
endWord = "cog"
wordList = ["hot","dot","dog","lot","log","cog"]


| Step | Current Word | Steps | Generated Words                 | Found in Dictionary | Action Taken               | Queue After Step         | Words Remaining           |
| ---- | ------------ | ----- | ------------------------------- | ------------------- | -------------------------- | ------------------------ | ------------------------- |
| 1    | **hit**      | 1     | ait, bit, cit, ..., **hot**     | ✅ hot               | Add ("hot", 2) to queue    | [("hot", 2)]             | {dot, dog, lot, log, cog} |
| 2    | **hot**      | 2     | aot, bot, ..., **dot**, **lot** | ✅ dot, lot          | Add ("dot", 3), ("lot", 3) | [("dot", 3), ("lot", 3)] | {dog, log, cog}           |
| 3    | **dot**      | 3     | aot, bot, ..., **dog**          | ✅ dog               | Add ("dog", 4)             | [("lot", 3), ("dog", 4)] | {log, cog}                |
| 4    | **lot**      | 3     | aot, bot, ..., **log**          | ✅ log               | Add ("log", 4)             | [("dog", 4), ("log", 4)] | {cog}                     |
| 5    | **dog**      | 4     | aog, bog, ..., **cog**          | ✅ cog               | Found endWord! return 5    | —                        | —                         |


Transformation Path=>hit → hot → dot → dog → cog


````python
from collections import deque
class Solution:
    def ladderLength(self, beginWord, endWord, wordList):
        words = set(wordList) # words ={"hot", "dot","dog", "lot","log","cog"}
        queue = deque([(beginWord, 1)]) # [("hit", 1)]
        while queue:
            word, steps = queue.popleft() #take 1 word from the front of the que
            # word = "hit" , step =1
            if word == endWord: # hit == cog, step = 1
                return steps
            for i in range(len(word)): # i =0 to 5   
                for c in "abcdefghijklmnopqrstuvwxyz":
                    new_word = word[:i] + c + word[i+1:] #dot, i= 2
                    #word[:i] = "ot" 
                    # c = a
                    # word[3:] = ""
                    if new_word in words: #hot
                        words.remove(new_word) # remove hot
                        queue.append((new_word, steps+1)) # ("hot", 2)
        return 0            
    # ✅ Example run
sol = Solution()
result = sol.ladderLength("hit", "cog", ["hot","dot","dog","lot","log","cog"])
print("Output:", result)  # 🟢 Expected Output: 5
````

# 8️⃣ 269 – Alien Dictionary


You are given a list of words that are already sorted according to the rules of some alien language.

Your job is to figure out the order of letters (the alien alphabet).

words = ["wrt", "wrf", "er", "ett", "rftt"]

These words are already sorted in alien dictionary order.

We must find:
👉 What order of characters (like ‘w’, ‘e’, ‘r’, ‘t’, ‘f’, …) makes this sorting possible?

| Compare         | First Difference | Meaning                  |
| --------------- | ---------------- | ------------------------ |
| "wrt" vs "wrf"  | `'t'` vs `'f'`   | `'t'` comes before `'f'` |
| "wrf" vs "er"   | `'w'` vs `'e'`   | `'w'` comes before `'e'` |
| "er" vs "ett"   | `'r'` vs `'t'`   | `'r'` comes before `'t'` |
| "ett" vs "rftt" | `'e'` vs `'r'`   | `'e'` comes before `'r'` |


w → e → r → t → f


✅ Final Answer

The alien alphabet order is:

"wertf"

### 🧩 Dry Run — `alienOrder(["wrt", "wrf", "er", "ett", "rftt"])`

### Step 1️⃣ — Input
words = ["wrt", "wrf", "er", "ett", "rftt"]


We’ll compare each **consecutive pair**:  
→ ("wrt", "wrf"), ("wrf", "er"), ("er", "ett"), ("ett", "rftt")

---

### Step 2️⃣ — Process Each Pair

| Pair | Compare letters | First difference found | Relationship (a→b) | Append to `order_pairs` |
|------|------------------|------------------------|---------------------|--------------------------|
| ("wrt", "wrf") | 'w'='w', 'r'='r', **'t' ≠ 'f'** | t→f | ✅ Append ('t','f') |
| ("wrf", "er") | **'w' ≠ 'e'** | w→e | ✅ Append ('w','e') |
| ("er", "ett") | 'e'='e', **'r' ≠ 't'** | r→t | ✅ Append ('r','t') |
| ("ett", "rftt") | **'e' ≠ 'r'** | e→r | ✅ Append ('e','r') |

---

### Step 3️⃣ — Final `order_pairs`

order_pairs = [('t','f'), ('w','e'), ('r','t'), ('e','r')]


✅ These show the alien letter dependencies:

w → e → r → t → f
---
### 🏁 Output:
```python
[('t','f'), ('w','e'), ('r','t'), ('e','r')]

````python
def alienOrder(words):
    order_pairs = []

    # find all (a → b) relationships
    for w1, w2 in zip(words, words[1:]):## ("wrt", "wrf"), ("wrf", "er"), ("er", "ett"), ("ett", "rftt")
        for a, b in zip(w1, w2):
            if a != b:
                order_pairs.append((a,b)) #('t', 'f')
                break
            else:
                if len(w1) > len(w2):#5>3
                    return ""

    return order_pairs 
                 
print(alienOrder(["wrt","wrf","er","ett","rftt"]))

````

w1 = "apple" = 5, w2 = "app" =3

All letters match (a == a, p == p, p == p)

Loop ends without break

→ so else runs

| Word 1 | Word 2 | First Different Letters | Added Pair |
| ------ | ------ | ----------------------- | ---------- |
| wrt    | wrf    | t ≠ f                   | ('t','f')  |
| wrf    | er     | w ≠ e                   | ('w','e')  |
| er     | ett    | r ≠ t                   | ('r','t')  |
| ett    | rftt   | e ≠ r                   | ('e','r')  |


output:-[('t','f'), ('w','e'), ('r','t'), ('e','r')]


What [('t','f'), ('w','e'), ('r','t'), ('e','r')] Means

Each tuple (a, b) means:

🔹 In the alien language, letter a comes before b

So, your list actually means:

| Tuple      | Meaning   |
| ---------- | --------- |
| ('t', 'f') | `t` → `f` |
| ('w', 'e') | `w` → `e` |
| ('r', 't') | `r` → `t` |
| ('e', 'r') | `e` → `r` |


If we connect them like a chain (directed graph):

w → e → r → t → f


So the Final Order Is:
"wertf"

✅ That’s the alien alphabet order which makes your given words sorted correctly.

# 9️⃣ 49 – Group Anagrams ✅
 

Given an array of strings strs, group the anagrams together.
You can return the answer in any order.

Two strings are anagrams if they contain the same characters in a different order.

### 🧩 Dry Run for: ["eat", "tea", "tan", "ate", "nat", "bat"]
| Step        | Word    | Sorted Key | Action                                      | anagram_dict (After Step)                                               |
| ----------- | ------- | ---------- | ------------------------------------------- | ----------------------------------------------------------------------- |
| **Initial** | —       | —          | Start with empty dict                       | `{}`                                                                    |
| **1**       | `"eat"` | `'aet'`    | Key not in dict → create `{"aet": ["eat"]}` | `{"aet": ["eat"]}`                                                      |
| **2**       | `"tea"` | `'aet'`    | Key exists → append `"tea"`                 | `{"aet": ["eat", "tea"]}`                                               |
| **3**       | `"tan"` | `'ant'`    | Key not in dict → create `{"ant": ["tan"]}` | `{"aet": ["eat", "tea"], "ant": ["tan"]}`                               |
| **4**       | `"ate"` | `'aet'`    | Key exists → append `"ate"`                 | `{"aet": ["eat", "tea", "ate"], "ant": ["tan"]}`                        |
| **5**       | `"nat"` | `'ant'`    | Key exists → append `"nat"`                 | `{"aet": ["eat", "tea", "ate"], "ant": ["tan", "nat"]}`                 |
| **6**       | `"bat"` | `'abt'`    | Key not in dict → create `{"abt": ["bat"]}` | `{"aet": ["eat", "tea", "ate"], "ant": ["tan", "nat"], "abt": ["bat"]}` |


{
  "aet": ["eat", "tea", "ate"],

  "ant": ["tan", "nat"],
  
  "abt": ["bat"]
}


output :- [["eat", "tea", "ate"], ["tan", "nat"], ["bat"]]

````python
from typing import List
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        anagrams = {}  # dictionary to group words
        for word in strs:
            key = ''. join(sorted(word))
            if key not in anagrams:
                anagrams[key] = [] #{"aet":[]}
            anagrams[key].append(word)#  {"aet":["eat"]} 
        return list(anagrams.values())     
print(Solution().groupAnagrams(["eat","tea","tan","ate","nat","bat"]))
````


 # 🔟 1160 – Find Words That Can Be Formed by Characters ✅

 🔹 Find the total length of all words that can be formed using the letters from a given string chars.

Each letter in chars can only be used as many times as it appears in chars.

### 🔍 Dry Run — `countCharacters(["cat","bt","hat","tree"], "atach")`

**Initial:** `total = 0`

words = ["cat", "bt", "hat", "tree"]

chars = "atach"

| Step | Word  | Char | `word.count(char)` | `chars.count(char)` | Check (`<=`?) | Action | `total` after step |
|:----:|:------|:-----:|:------------------:|:-------------------:|:-------------:|:-------|:------------------:|
| 1 | "cat" | c | 1 | 1 | ✅ 1 ≤ 1 | continue checking | 0 |
| 2 | "cat" | a | 1 | 2 | ✅ 1 ≤ 2 | continue checking | 0 |
| 3 | "cat" | t | 1 | 1 | ✅ 1 ≤ 1 | all chars ok → add len("cat") = 3 | **3** |
| 4 | "bt"  | b | 1 | 0 | ❌ 1 ≤ 0 false | break → skip word "bt" | 3 |
| 5 | "hat" | h | 1 | 1 | ✅ 1 ≤ 1 | continue checking | 3 |
| 6 | "hat" | a | 1 | 2 | ✅ 1 ≤ 2 | continue checking | 3 |
| 7 | "hat" | t | 1 | 1 | ✅ 1 ≤ 1 | all chars ok → add len("hat") = 3 | **6** |
| 8 | "tree"| t | 1 | 1 | ✅ 1 ≤ 1 | continue checking | 6 |
| 9 | "tree"| r | 1 | 0 | ❌ 1 ≤ 0 false | break → skip word "tree" | 6 |

---

**Final Result:** `total = 6`


can_form is a boolean flag that shows whether a word can be formed using the given characters — ✅ True means all letters are available so far, ❌ False means a letter is missing, so stop checking that word.

````python
def countCharacters(words, chars):
    total = 0
    for word in words: #cat, bt, hat
        can_form = True
        for c in word: #char =c
            if word.count(c) > chars.count(c): # 1 > 0
                can_form = False
                break
        if can_form:
            total += len(word)
    return total            
print(countCharacters(["cat", "bt", "hat", "tree"], "atach"))
````
