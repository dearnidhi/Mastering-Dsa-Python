#  6Ô∏è‚É£ 953 ‚Äì Verifying an Alien Dictionary


### üëΩ **Problem Statement ‚Äì Verifying an Alien Dictionary**

You are given:

- A list of words: **`words`**  
- A string: **`order`**, which represents the **alphabetical order** of characters in an alien language.

Your task is to determine **if the given list of words is sorted** according to this alien dictionary order.

---

#### ‚úÖ **Rules:**
1. Each word must appear **lexicographically before or equal to** the next word, based on the given `order`.  
2. If two words share a prefix, the **shorter word** should come first (e.g., `"app"` should come before `"apple"`).  

---

#### üß† **Return:**
- **`True`** ‚Üí if the words are sorted correctly.  
- **`False`** ‚Üí if the order is violated anywhere.

---

#### üß™ **Example:**
```python
isAlienSorted(["hello", "leetcode"], "hlabcdefgijkmnopqrstuvwxyz") ‚ûú True  
isAlienSorted(["word", "world", "row"], "worldabcefghijkmnpqstuvxyz") ‚ûú False  
isAlienSorted(["apple", "app"], "abcdefghijklmnopqrstuvwxyz") ‚ûú False


Lexicographical = compare letter by letter using a given alphabet order (normal or alien).


## **Input:**
- `words = ["hello", "leetcode"]`
- `order = "hlabcdefgijkmnopqrstuvwxyz"`

---

## **Step 1: Create Rank Dictionary**
```python
rank = {ch: i for i, ch in enumerate(order)}
# rank = {'h':0, 'l':1, 'a':2, 'b':3, 'c':4, ...}

üëâ Means: 'h' comes before 'l', 'l' before 'a', etc.
This gives every character a numeric rank according to its order in the alien language.

| Index | w1[j] | w2[j] | rank[w1[j]] | rank[w2[j]] | Comparison | Result                  |
| ----- | ----- | ----- | ----------- | ----------- | ---------- | ----------------------- |
| 0     | h     | l     | 0           | 1           | 0 < 1      | ‚úÖ correct order ‚Üí break |


‚úÖ Since 'h' comes before 'l', words are in correct order.
No more pairs ‚Üí return True.

Output: True

#### üß© Example 2

**Input:**
```python
words = ["word", "world", "row"]
order = "worldabcefghijkmnpqstuvxyz"

rank = {'w':0, 'o':1, 'r':2, 'l':3, 'd':4, 'a':5, ...}

Compare ("word", "world")


| j | w1[j] | w2[j] | rank[w1[j]] | rank[w2[j]] | Result                |
| - | ----- | ----- | ----------- | ----------- | --------------------- |
| 0 | w     | w     | same        | same        | continue              |
| 1 | o     | o     | same        | same        | continue              |
| 2 | r     | r     | same        | same        | continue              |
| 3 | d     | l     | 4           | 3           | ‚ùå wrong order (4 > 3) |

üëâ 'd' comes after 'l' in alien order ‚Üí ‚ùå not sorted ‚Üí return False.

Output: False

#### ‚ùå Example 3
words = ["apple", "app"], order = "abcdefghijklmnopqrstuvwxyz"

Compare ("apple", "app")

Check letter by letter: a == a, p == p, p == p

No difference found (loop ended)

Then check:

if len(w1) > len(w2):

    return False

‚úÖ len("apple") = 5 and len("app") = 3 ‚Üí longer word first ‚ùå (prefix rule broken)

Output: False    



| Words                    | Order      | Output  | Why                                 |
| ------------------------ | ---------- | ------- | ----------------------------------- |
| ["hello", "leetcode"]    | "hl..."    | ‚úÖ True  | ‚Äòh‚Äô comes before ‚Äòl‚Äô                |
| ["word", "world", "row"] | "world..." | ‚ùå False | ‚Äòd‚Äô comes after ‚Äòl‚Äô                 |
| ["apple", "app"]         | "abc..."   | ‚ùå False | longer word comes before its prefix |


````python
def isAlienSorted(words, order):
    # create dictionary for order lookup
    rank = {ch: i for i, ch in enumerate(order)}
    
    # check each consecutive pair
    for i in range(len(words) - 1): #len(words) = 3

        w1, w2 = words[i], words[i + 1]
        for j in range(min(len(w1), len(w2))):
            if w1[j] != w2[j]:
                # if order of letters is wrong
                if rank[w1[j]] > rank[w2[j]]:
                    return False
                break
        else:
            # no difference found ‚Üí prefix case
            if len(w1) > len(w2):
                return False
    return True


# ‚úÖ Example
print(isAlienSorted(["hello", "leetcode"], "hlabcdefgijkmnopqrstuvwxyz"))  # True
print(isAlienSorted(["word", "world", "row"], "worldabcefghijkmnpqstuvxyz"))  # False
print(isAlienSorted(["apple", "app"], "abcdefghijklmnopqrstuvwxyz"))  # False
````

# 7Ô∏è‚É£ 127 ‚Äì Word Ladder


### üß© **Problem Statement ‚Äì Word Ladder**

You are given:

- Two words: **`beginWord`** and **`endWord`**  
- A list of words: **`wordList`**

Your task is to **find the length of the shortest transformation sequence** from `beginWord` to `endWord`, such that:

1. Only **one letter** can be changed at a time.  
2. Each transformed word **must exist in `wordList`**.

Return the **number of words** in the shortest transformation sequence,  
**including both `beginWord` and `endWord`**.  

If no such transformation exists, **return `0`**.

### üß©Dry Run 

### **Input**
```python
beginWord = "hit"
endWord = "cog"
wordList = ["hot","dot","dog","lot","log","cog"]


| Step | Current Word | Steps | Generated Words                 | Found in Dictionary | Action Taken               | Queue After Step         | Words Remaining           |
| ---- | ------------ | ----- | ------------------------------- | ------------------- | -------------------------- | ------------------------ | ------------------------- |
| 1    | **hit**      | 1     | ait, bit, cit, ..., **hot**     | ‚úÖ hot               | Add ("hot", 2) to queue    | [("hot", 2)]             | {dot, dog, lot, log, cog} |
| 2    | **hot**      | 2     | aot, bot, ..., **dot**, **lot** | ‚úÖ dot, lot          | Add ("dot", 3), ("lot", 3) | [("dot", 3), ("lot", 3)] | {dog, log, cog}           |
| 3    | **dot**      | 3     | aot, bot, ..., **dog**          | ‚úÖ dog               | Add ("dog", 4)             | [("lot", 3), ("dog", 4)] | {log, cog}                |
| 4    | **lot**      | 3     | aot, bot, ..., **log**          | ‚úÖ log               | Add ("log", 4)             | [("dog", 4), ("log", 4)] | {cog}                     |
| 5    | **dog**      | 4     | aog, bog, ..., **cog**          | ‚úÖ cog               | Found endWord! return 5    | ‚Äî                        | ‚Äî                         |


Transformation Path=>hit ‚Üí hot ‚Üí dot ‚Üí dog ‚Üí cog


````python
from collections import deque

class Solution:
    def ladderLength(self, beginWord, endWord, wordList):
        words = set(wordList) #words = {"hot", "dot", "dog", "lot", "log", "cog"}
        queue = deque([(beginWord, 1)])# Queue looks like: [("hit", 1)]
        
        while queue:
            word, steps = queue.popleft() #Take one word from the front of the queue.At first:-word = "hit", steps = 1
            
            if word == endWord: #Did we reach 'cog' yet?‚ÄùNo ‚Äî continue exploring neighbors.
                return steps
            
            for i in range(len(word)):
                for c in 'abcdefghijklmnopqrstuvwxyz':
                    new_word = word[:i] + c + word[i+1:]##word ="hit", i=0,new_word=[:0]+a+[0+1:]=""+"a"+"it" = "ait"
                    if new_word in words:#"hot" found in dict!  
                        words.remove(new_word)#So:-Remove "hot" (mark as visited)
                        queue.append((new_word, steps + 1))# Add to queue: ("hot", 2)
        
        return 0
    # ‚úÖ Example run
sol = Solution()
result = sol.ladderLength("hit", "cog", ["hot","dot","dog","lot","log","cog"])
print("Output:", result)  # üü¢ Expected Output: 5
````

# 8Ô∏è‚É£ 269 ‚Äì Alien Dictionary


You are given a list of words that are already sorted according to the rules of some alien language.

Your job is to figure out the order of letters (the alien alphabet).

words = ["wrt", "wrf", "er", "ett", "rftt"]

These words are already sorted in alien dictionary order.

We must find:
üëâ What order of characters (like ‚Äòw‚Äô, ‚Äòe‚Äô, ‚Äòr‚Äô, ‚Äòt‚Äô, ‚Äòf‚Äô, ‚Ä¶) makes this sorting possible?

| Compare         | First Difference | Meaning                  |
| --------------- | ---------------- | ------------------------ |
| "wrt" vs "wrf"  | `'t'` vs `'f'`   | `'t'` comes before `'f'` |
| "wrf" vs "er"   | `'w'` vs `'e'`   | `'w'` comes before `'e'` |
| "er" vs "ett"   | `'r'` vs `'t'`   | `'r'` comes before `'t'` |
| "ett" vs "rftt" | `'e'` vs `'r'`   | `'e'` comes before `'r'` |


w ‚Üí e ‚Üí r ‚Üí t ‚Üí f


‚úÖ Final Answer

The alien alphabet order is:

"wertf"

### üß© Dry Run ‚Äî `alienOrder(["wrt", "wrf", "er", "ett", "rftt"])`

### Step 1Ô∏è‚É£ ‚Äî Input
words = ["wrt", "wrf", "er", "ett", "rftt"]


We‚Äôll compare each **consecutive pair**:  
‚Üí ("wrt", "wrf"), ("wrf", "er"), ("er", "ett"), ("ett", "rftt")

---

### Step 2Ô∏è‚É£ ‚Äî Process Each Pair

| Pair | Compare letters | First difference found | Relationship (a‚Üíb) | Append to `order_pairs` |
|------|------------------|------------------------|---------------------|--------------------------|
| ("wrt", "wrf") | 'w'='w', 'r'='r', **'t' ‚â† 'f'** | t‚Üíf | ‚úÖ Append ('t','f') |
| ("wrf", "er") | **'w' ‚â† 'e'** | w‚Üíe | ‚úÖ Append ('w','e') |
| ("er", "ett") | 'e'='e', **'r' ‚â† 't'** | r‚Üít | ‚úÖ Append ('r','t') |
| ("ett", "rftt") | **'e' ‚â† 'r'** | e‚Üír | ‚úÖ Append ('e','r') |

---

### Step 3Ô∏è‚É£ ‚Äî Final `order_pairs`

order_pairs = [('t','f'), ('w','e'), ('r','t'), ('e','r')]


‚úÖ These show the alien letter dependencies:

w ‚Üí e ‚Üí r ‚Üí t ‚Üí f
---
### üèÅ Output:
```python
[('t','f'), ('w','e'), ('r','t'), ('e','r')]

````python
def alienOrder(words):
    order_pairs = []

    # find all (a ‚Üí b) relationships
    for w1, w2 in zip(words, words[1:]): ## ("wrt","wrf")("wrf","er") ("er","ett") ("ett","rftt") 
        for a, b in zip(w1, w2): ## For ("wrt","wrf") ->compare 'w' vs 'w', compare 'r' vs 'r', compare 't' vs 'f'
            if a != b: # # ‚Äúwrt‚Äù vs ‚Äúwrf‚Äù ‚Üí difference at 't' != 'f'
                order_pairs.append((a, b)) #  Add ('t','f')
                break
        else:
            if len(w1) > len(w2): # 5> 3
                return ""  # invalid prefix case

    return order_pairs
print(alienOrder(["wrt","wrf","er","ett","rftt"]))

````

w1 = "apple" = 5, w2 = "app" =3

All letters match (a == a, p == p, p == p)

Loop ends without break

‚Üí so else runs

| Word 1 | Word 2 | First Different Letters | Added Pair |
| ------ | ------ | ----------------------- | ---------- |
| wrt    | wrf    | t ‚â† f                   | ('t','f')  |
| wrf    | er     | w ‚â† e                   | ('w','e')  |
| er     | ett    | r ‚â† t                   | ('r','t')  |
| ett    | rftt   | e ‚â† r                   | ('e','r')  |


output:-[('t','f'), ('w','e'), ('r','t'), ('e','r')]


What [('t','f'), ('w','e'), ('r','t'), ('e','r')] Means

Each tuple (a, b) means:

üîπ In the alien language, letter a comes before b

So, your list actually means:

| Tuple      | Meaning   |
| ---------- | --------- |
| ('t', 'f') | `t` ‚Üí `f` |
| ('w', 'e') | `w` ‚Üí `e` |
| ('r', 't') | `r` ‚Üí `t` |
| ('e', 'r') | `e` ‚Üí `r` |


If we connect them like a chain (directed graph):

w ‚Üí e ‚Üí r ‚Üí t ‚Üí f


So the Final Order Is:
"wertf"

‚úÖ That‚Äôs the alien alphabet order which makes your given words sorted correctly.

# 9Ô∏è‚É£ 49 ‚Äì Group Anagrams ‚úÖ
 

Given an array of strings strs, group the anagrams together.
You can return the answer in any order.

Two strings are anagrams if they contain the same characters in a different order.

### üß© Dry Run for: ["eat", "tea", "tan", "ate", "nat", "bat"]
| Step        | Word    | Sorted Key | Action                                      | anagram_dict (After Step)                                               |
| ----------- | ------- | ---------- | ------------------------------------------- | ----------------------------------------------------------------------- |
| **Initial** | ‚Äî       | ‚Äî          | Start with empty dict                       | `{}`                                                                    |
| **1**       | `"eat"` | `'aet'`    | Key not in dict ‚Üí create `{"aet": ["eat"]}` | `{"aet": ["eat"]}`                                                      |
| **2**       | `"tea"` | `'aet'`    | Key exists ‚Üí append `"tea"`                 | `{"aet": ["eat", "tea"]}`                                               |
| **3**       | `"tan"` | `'ant'`    | Key not in dict ‚Üí create `{"ant": ["tan"]}` | `{"aet": ["eat", "tea"], "ant": ["tan"]}`                               |
| **4**       | `"ate"` | `'aet'`    | Key exists ‚Üí append `"ate"`                 | `{"aet": ["eat", "tea", "ate"], "ant": ["tan"]}`                        |
| **5**       | `"nat"` | `'ant'`    | Key exists ‚Üí append `"nat"`                 | `{"aet": ["eat", "tea", "ate"], "ant": ["tan", "nat"]}`                 |
| **6**       | `"bat"` | `'abt'`    | Key not in dict ‚Üí create `{"abt": ["bat"]}` | `{"aet": ["eat", "tea", "ate"], "ant": ["tan", "nat"], "abt": ["bat"]}` |


{
  "aet": ["eat", "tea", "ate"],

  "ant": ["tan", "nat"],
  
  "abt": ["bat"]
}


output :- [["eat", "tea", "ate"], ["tan", "nat"], ["bat"]]

````python
from typing import List
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        anagrams = {}  # dictionary to group words

        for word in strs:
            key = ''.join(sorted(word))  # sort letters to form key

            if key not in anagrams:
                anagrams[key] = [] #{"aet": []}

            anagrams[key].append(word) # {"aet": ["eat"]}

        return list(anagrams.values())

print(Solution().groupAnagrams(["eat","tea","tan","ate","nat","bat"]))
````


 # üîü 1160 ‚Äì Find Words That Can Be Formed by Characters ‚úÖ

 üîπ Find the total length of all words that can be formed using the letters from a given string chars.

Each letter in chars can only be used as many times as it appears in chars.

### üîç Dry Run ‚Äî `countCharacters(["cat","bt","hat","tree"], "atach")`

**Initial:** `total = 0`

words = ["cat", "bt", "hat", "tree"]

chars = "atach"

| Step | Word  | Char | `word.count(char)` | `chars.count(char)` | Check (`<=`?) | Action | `total` after step |
|:----:|:------|:-----:|:------------------:|:-------------------:|:-------------:|:-------|:------------------:|
| 1 | "cat" | c | 1 | 1 | ‚úÖ 1 ‚â§ 1 | continue checking | 0 |
| 2 | "cat" | a | 1 | 2 | ‚úÖ 1 ‚â§ 2 | continue checking | 0 |
| 3 | "cat" | t | 1 | 1 | ‚úÖ 1 ‚â§ 1 | all chars ok ‚Üí add len("cat") = 3 | **3** |
| 4 | "bt"  | b | 1 | 0 | ‚ùå 1 ‚â§ 0 false | break ‚Üí skip word "bt" | 3 |
| 5 | "hat" | h | 1 | 1 | ‚úÖ 1 ‚â§ 1 | continue checking | 3 |
| 6 | "hat" | a | 1 | 2 | ‚úÖ 1 ‚â§ 2 | continue checking | 3 |
| 7 | "hat" | t | 1 | 1 | ‚úÖ 1 ‚â§ 1 | all chars ok ‚Üí add len("hat") = 3 | **6** |
| 8 | "tree"| t | 1 | 1 | ‚úÖ 1 ‚â§ 1 | continue checking | 6 |
| 9 | "tree"| r | 1 | 0 | ‚ùå 1 ‚â§ 0 false | break ‚Üí skip word "tree" | 6 |

---

**Final Result:** `total = 6`


can_form is a boolean flag that shows whether a word can be formed using the given characters ‚Äî ‚úÖ True means all letters are available so far, ‚ùå False means a letter is missing, so stop checking that word.

````python
def countCharacters(words, chars):
    total = 0
    
    for word in words:
        can_form = True
        for c in word:
            # agar char count jyada ho gaya to break
            if word.count(c) > chars.count(c):
                can_form = False
                break
        if can_form:
            total += len(word)
    return total


print(countCharacters(["cat", "bt", "hat", "tree"], "atach"))
````
