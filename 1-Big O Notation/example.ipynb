# 2. O(n) â†’ Linear Time & Space
ðŸ“Œ Input jitna bada, utne steps

```python
def printAll(arr):
    for item in arr:
        print(item)
```

# 4. O(log n) â†’ Logarithmic Time
ðŸ“Œ Jaise Binary Search

```python
def binarySearch(arr, target):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

Time: O(log n) â†’ har baar array half hota

Space: O(1) â†’ koi extra memory nahi

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
```

# . O(n!) â†’ Factorial Time
ðŸ“Œ Jaise Permutations

```python
def permute(s, l, r):
    if l == r:
        print("".join(s))
    else:
        for i in range(l, r+1):
            s[l], s[i] = s[i], s[l]
            permute(s, l+1, r)
            s[l], s[i] = s[i], s[l]
```

Time: O(n!)

Space: O(n) â†’ recursion depth
