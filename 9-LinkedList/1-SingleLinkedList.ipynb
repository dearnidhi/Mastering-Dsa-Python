# What is a Linked List?

A Linked List is a data structure that stores data in nodes.

# Each node has:

* Data (value of the node)

* Next pointer (address/reference of the next node)

ðŸ‘‰ Unlike arrays/lists, elements are not stored in continuous memory locations.

Example (visual):

Head â†’ [10|Next] â†’ [20|Next] â†’ [30|Next] â†’ None

# Linked List vs Array/List
# Array (Python list)

arr = [10, 20, 30]

memory contiguous -> indexing fast (arr[1] = 20)

Linked List -> scattered in memory

head -> [10|next] -> [20|next] -> [30|None]


Array â†’ fast indexing , slow insert/delete 

Linked List â†’ slow indexing, fast insert/delete 

Use Array if you need fast access.

Use Linked List if you need frequent insertions/deletions.

# Types of Linked List

# Singly Linked List
Each node points to the next node only.

10 â†’ 20 â†’ 30 â†’ None




# Doubly Linked List
Each node has two pointers: prev and next.

None â† 10 â‡† 20 â‡† 30 â†’ None





# Circular Linked List

Last node points back to the head (forms a circle).

10 â†’ 20 â†’ 30 â†’ back to 10




# ðŸ”¹ Arrays in Memory

**Example:**

| Index  | 0  | 1  | 2  | 3  |
|--------|----|----|----|----|
| Value  | 10 | 20 | 30 | 40 |

# ðŸ”¹ Linked List in Memory

In a **Linked List**, elements are stored as **nodes**.  
Each node has two parts:

- **Data** â†’ the actual value (like 10, 20, 30 â€¦)  
- **Pointer (next)** â†’ the memory address of the next node  

ðŸ‘‰ A node looks like:


[ Data | Next ]


**Example Linked List:**
Head â†’ [1|] â†’ [2|] â†’ [4|*] â†’ [5|None]


- **Head** points to the first node.  
- Each node points to the **next node**.  
- The **last node** points to `None` (Tail).  

---

# ðŸ”¹ Important Points

- Nodes are **not stored contiguously**.  
- When you create a node, it gets memory from **any free space** in RAM.  

**Example (scattered nodes):**


Node1 (data=1) â†’ Address 200

Node2 (data=2) â†’ Address 540

Node3 (data=4) â†’ Address 120

Node4 (data=5) â†’ Address 999

(They are scattered but linked using the **Next pointer**)

- **Extra memory** is needed for storing the pointer.  
- **Dynamic sizing** â†’ Unlike arrays, we donâ€™t fix the size at the start.  
- **Access is slower** â†’ Must traverse from **Head â†’ Node1 â†’ Node2 â†’ â€¦** (O(n)).  

---

# ðŸ”¹ Difference from Arrays

- **Arrays** â†’ Fast access (**O(1)**), fixed size.  
- **Linked List** â†’ Slow access (**O(n)**), dynamic size.

```python
def insert(self, index, value):
    # Boundary check
    if index < 0 or index > self.length:
        raise Exception("Index out of bounds")
    
    new_node = Node(value)
    
    # Insert in empty list
    if self.head is None:
        self.head = new_node
        self.tail = new_node
    
    # Insert at beginning
    elif index == 0:
        new_node.next = self.head
        self.head = new_node
    
    # Insert at end (or when index == length)
    elif index == self.length:
        self.tail.next = new_node
        self.tail = new_node
    
    # Insert in middle
    else:
        temp_node = self.head
        for _ in range(index - 1):#0
            temp_node = temp_node.next #20
        new_node.next = temp_node.next #30-20
        temp_node.next = new_node
        
        # If inserting at the end, update tail
        if new_node.next is None:
            self.tail = new_node
    
    self.length += 1
```

```python
def search(self, target):#30
        current = self.head #10
        index = 0
        while current is not None:
            if current.value == target: #10==30 F,20==30F, 30 == 30 T
                return index #2
            current = current.next #20, 30
            index += 1
        return -1
```

```python
# Node and LinkedList classes
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1

    def search(self, target):
        current = self.head
        index = 0
        while current is not None:
            if current.value == target:
                return index
            current = current.next
            index += 1
        return -1

# -----------------------
# Example usage
ll = LinkedList()
ll.append(10)
ll.append(20)
ll.append(30)

print("search(20) â†’", ll.search(20))   # Output: 1
print("search(30) â†’", ll.search(30))   # Output: 2
print("search(40) â†’", ll.search(40))   # Output: -1
```

```python
# Example continuation with LinkedList

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def append(self, value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1

    def get(self, index):
        if index == -1:
            return self.tail
        elif index < -1 or index >= self.length:
            return None
        current = self.head
        for _ in range(index):
            current = current.next
        return current

# -----------------------
# Example usage
ll = LinkedList()
ll.append(10)
ll.append(20)
ll.append(30)

print("get(0) â†’", ll.get(0).value)   # Output: 10
print("get(1) â†’", ll.get(1).value)   # Output: 20
print("get(2) â†’", ll.get(2).value)   # Output: 30
print("get(-1) â†’", ll.get(-1).value) # Output: 30 (tail)
print("get(3) â†’", ll.get(3))         # Output: None
```

# 11. pop_first Method

Removes and returns the first node (head).

Handles empty lists and single-node lists.

Updates head and length.

head â†’ [10] â†’ [20] â†’ [30]  
length = 3


| Step                   | head           | tail | length | Action             | Return      |
| ---------------------- | -------------- | ---- | ------ | ------------------ | ----------- |
| Save first node        | 10             | 30   | 3      | popped_node = head | â€”           |
| Move head              | 20             | 30   | â€”      | head = head.next   | â€”           |
| Disconnect popped node | 10.next = None | â€”    | â€”      | break link         | popped_node |
| Decrease length        | â€”              | â€”    | 2      | length -= 1        | â€”           |


```python
def pop_first(self):
    if self.length == 0:
        return 
    popped_node = self.head #10
    if self.length == 1:
        self.head = None
        self.tail = None
    else:
        self.head = self.head.next #20 
        popped_node.next = None # 10->
    self.length -= 1
    return popped_node    
```


# 12. pop Method
Removes and returns the last node (tail).

Handles empty lists and single-node lists.

Traverses to the second-last node to update tail.

#### ðŸ”¹ Dry Run of `pop()` Method (Remove Last Node)

## Initial List

head â†’ [10] â†’ [20] â†’ [30] â† tail

length = 3
| Step | Code                                | Action               | temp Points to | temp.next | Condition Check       | List State                        |
| ---- | ----------------------------------- | -------------------- | -------------- | --------- | --------------------- | --------------------------------- |
| 0    | `popped_node = self.tail`           | Save tail reference  | -              | -         | -                     | headâ†’[10]â†’[20]â†’[30]â†tail          |
| 1    | `temp = self.head`                  | Start traversal      | [10]           | [20]      | -                     | tempâ†’[10]â†’[20]â†’[30]â†tail          |
| 2    | `while temp.next is not self.tail:` | Check loop condition | [10]           | [20]      | [20] â‰  tail â†’ True âœ…  | Enter loop                        |
| 3    | `temp = temp.next`                  | Move forward         | [20]           | [30]      | -                     | tempâ†’[20]â†’[30]â†tail               |
| 4    | `while temp.next is not self.tail:` | Check loop condition | [20]           | [30]      | [30] = tail â†’ False âŒ | Exit loop                         |
| 5    | `temp.next = None`                  | Disconnect last node | [20]           | None      | -                     | [10]â†’[20]â†’None, [30] disconnected |
| 6    | `self.tail = temp`                  | Update tail pointer  | [20]           | None      | -                     | tailâ†’[20]                         |
| 7    | `self.length -= 1`                  | Decrease length      | [20]           | None      | -                     | Length: 3 â†’ 2                     |
| 8    | `return popped_node`                | Return removed node  | -              | -         | -                     | Return [30]                       |



```python
def pop(self):
    if self.length == 0:
        return None
    popped_node = self.tail #30
    if self.length == 1:
        self.head = None
        self.tail = None
    else: 
        temp =self.head #10
        while temp.next is not self.tail: #20 =30 T
            temp = temp.next #20
        temp.next = None   #10->20
        self.tail = temp
    self.length -=1
    return popped_node     

```

# ðŸ”¹ Dry Run of `remove(index)` Method

## Example List

head â†’ [10] â†’ [20] â†’ [30] â†’ [40] â† tail
length = 4


We will remove `index = 2` (value 30).

| Step | Code                                        | Action                  | prev_node | popped_node | List State                 |
| ---- | ------------------------------------------- | ----------------------- | --------- | ----------- | -------------------------- |
| 0    | `if index < -1 or index >= self.length:`    | Check index validity    | -         | -           | Valid index (2), continue  |
| 1    | `if index == 0:`                            | Check remove head       | -         | -           | Not index 0 â†’ skip         |
| 2    | `if index == -1 or index == self.length-1:` | Check remove tail       | -         | -           | Index 2 â‰  last â†’ skip      |
| 3    | `prev_node = self.get(index-1)`             | Find previous node      | [20]      | -           | prev_node â†’ [20]           |
| 4    | `popped_node = prev_node.next`              | Identify node to remove | [20]      | [30]        | popped_node â†’ [30]         |
| 5    | `prev_node.next = popped_node.next`         | Skip removed node       | [20]      | [30]        | List becomes: 10 â†’ 20 â†’ 40 |
| 6    | `popped_node.next = None`                   | Disconnect removed node | [20]      | None        | [30] fully isolated        |
| 7    | `self.length -= 1`                          | Reduce length           | -         | -           | length: 4 â†’ 3              |
| 8    | `return popped_node`                        | Return removed node     | -         | [30]        | Returns [30]               |


```python
def remove(self, index):
    if index < -1 or index >= self.length:
        return None
    
    if index == 0:
        return self.pop_first()
    
    if index == -1 or index == self.length-1:
        return self.pop()
    
    prev_node = self.get(index-1) #get(1) =20
    popped_node = prev_node.next #30
    prev_node.next = popped_node.next # -> 30->
    popped_node.next = None #30
    self.length -= 1
    return popped_node
```
