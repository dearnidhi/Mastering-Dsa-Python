# Leetcode 35. Search Insert Position

### Step-by-Step Dry Run

**Input**
- `nums = [1, 3, 5, 6]`
- `target = 5`

| Step | Left | Right | Mid | nums[mid] | Condition Checked | Action |
|-----:|:----:|:-----:|:---:|:---------:|-------------------|--------|
| 0 | 0 | 3 | ‚Äì | ‚Äì | ‚Äì | Initialize |
| 1 | 0 | 3 | (0+3)//2 = 1 | 3 | 3 == 5 ‚ùå<br>3 < 5 ‚úÖ | `left = mid + 1 = 2` |
| 2 | 2 | 3 | (2+3)//2 = 2 | 5 | 5 == 5 ‚úÖ | **Return 2** |

**Output:** `2`




```python
from typing import List
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        left, right = 0 , len(nums) - 1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1 
            else: 
                right = mid - 1  
        return left  
print(Solution().searchInsert([1,3,5,6], 5))            
```

# Leetcode - 34. Find First and Last Position of Element in Sorted Array

üîπ Idea

Use binary search twice:

Once to find the first occurrence (move left when nums[mid] == target).

Once to find the last occurrence (move right when nums[mid] == target).

If the target is not found, return [-1, -1].

nums = [5, 7, 7, 8, 8, 10]   
target = 8   
Table 1: findFirst(nums, target) ‚Äî Find First Occurrence  

| Step | Left | Right |      Mid     | nums[mid] | Condition    | Action                   | First |
| ---: | :--: | :---: | :----------: | :-------: | ------------ | ------------------------ | :---: |
|    0 |   0  |   5   |       ‚Äì      |     ‚Äì     | ‚Äì            | ‚Äì                        |   -1  |
|    1 |   0  |   5   | (0+5)//2 = 2 |     7     | 7 < 8        | `left = 3`               |   -1  |
|    2 |   3  |   5   | (3+5)//2 = 4 |     8     | 8 == 8       | `first = 4`, `right = 3` |   4   |
|    3 |   3  |   3   | (3+3)//2 = 3 |     8     | 8 == 8       | `first = 3`, `right = 2` |   3   |
|    4 |   3  |   2   |       ‚Äì      |     ‚Äì     | left > right | Exit loop                |   3   |

Result: findFirst returns 3 ‚úÖ  

Table 2: findLast(nums, target) ‚Äî Find Last Occurrence

| Step | Left | Right |      Mid     | nums[mid] | Condition    | Action                 | Last |
| ---: | :--: | :---: | :----------: | :-------: | ------------ | ---------------------- | :--: |
|    0 |   0  |   5   |       ‚Äì      |     ‚Äì     | ‚Äì            | ‚Äì                      |  -1  |
|    1 |   0  |   5   | (0+5)//2 = 2 |     7     | 7 < 8        | `left = 3`             |  -1  |
|    2 |   3  |   5   | (3+5)//2 = 4 |     8     | 8 == 8       | `last = 4`, `left = 5` |   4  |
|    3 |   5  |   5   | (5+5)//2 = 5 |     10    | 10 > 8       | `right = 4`            |   4  |
|    4 |   5  |   4   |       ‚Äì      |     ‚Äì     | left > right | Exit loop              |   4  |

Result: findLast returns 4 ‚úÖ     
output :- [3, 4]


```python
from typing import List

class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        def findFirst(nums, target):
            left, right = 0, len(nums) - 1
            first = -1
            while left <= right:
                mid = (left + right) // 2
                if nums[mid] == target:
                    first = mid
                    right = mid -1
                elif nums[mid] < target:    
                    left = mid + 1

                else: 
                    right = mid -1
            return first  
                
        def findLast(nums, target):
            left, right = 0, len(nums) - 1
            last = -1
            while left <= right:
                mid = (left + right) // 2
                if nums[mid] == target:
                    last = mid
                    left = mid + 1
                elif nums[mid]< target:
                    left = mid + 1
                else:
                    right = mid -1
            return last
        return [findFirst(nums, target), findLast(nums, target)]
```

# 33. Search in Rotated Sorted Array

nums = [4, 5, 6, 7, 0, 1, 2]  
target = 0

| Step | Left | Right | Mid `(left+right)//2` | nums[mid] | Comparison & Check                                                                 | Action                | Next Range      |
| ---: | :--: | :---: | :-------------------: | :-------: | ---------------------------------------------------------------------------------- | --------------------- | --------------- |
|    0 |   0  |   6   |           ‚Äì           |     ‚Äì     | Initialization                                                                     | Begin loop            | `[0, 6]`        |
|    1 |   0  |   6   |      (0+6)//2 = 3     |     7     | 1. `7 == 0` ‚ùå <br> 2. Left sorted? `4 ‚â§ 7` ‚úÖ <br> 3. Target in left? `4 ‚â§ 0 < 7` ‚ùå | `left = mid + 1 = 4`  | `[4, 6]`        |
|    2 |   4  |   6   |      (4+6)//2 = 5     |     1     | 1. `1 == 0` ‚ùå <br> 2. Left sorted? `0 ‚â§ 1` ‚úÖ <br> 3. Target in left? `0 ‚â§ 0 < 1` ‚úÖ | `right = mid - 1 = 4` | `[4, 4]`        |
|    3 |   4  |   4   |      (4+4)//2 = 4     |     0     | `0 == 0` ‚úÖ                                                                         | **Return `mid = 4`**  | Target found üéØ |


```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            # Step 1: Found target
            if nums[mid] == target:
                return mid
            # Step 2: Left half is sorted
            if nums[left] <= nums[mid]:
                if nums[left] <= target < nums[mid] :
                    right = mid - 1
                else: 
                    left = mid + 1    

            # Step 3: Right half is sorted
            else:
                if nums[mid] < target <= nums[right]:
                    left = mid + 1
                else:
                    right = mid - 1
            
print(Solution().search([4, 5, 6, 7, 0, 1, 2],0))    
```

# 153 - Find Minimum in Rotated Sorted Array

nums = [3, 4, 5, 1, 2]

| Step | Left | Right | `left < right` | Mid | nums[mid] | nums[right] | Condition | Action           | Next           |
| ---: | :--: | :---: | :------------: | :-: | :-------: | :---------: | :-------: | :--------------- | :------------- |
|    0 |   0  |   4   |        ‚úÖ       |  ‚Äì  |     ‚Äì     |      ‚Äì      |     ‚Äì     | enter loop       | L=0 R=4        |
|    1 |   0  |   4   |        ‚úÖ       |  2  |     5     |      2      |  5 > 2 ‚úÖ  | `left = mid + 1` | L=3 R=4        |
|    2 |   3  |   4   |        ‚úÖ       |  3  |     1     |      2      |  1 > 2 ‚ùå  | `right = mid`    | L=3 R=3        |
|    3 |   3  |   3   |        ‚ùå       |  ‚Äì  |     ‚Äì     |      ‚Äì      |     ‚Äì     | exit loop        | return nums[3] |

Return nums[3] = 1

```python
from typing import List

class Solution:
    def findMin(self, nums: List[int]) -> int:
        left, right = 0, len(nums) - 1
        
        while left < right:
            mid = (left + right) // 2
            
            if nums[mid] > nums[right]:
                # minimum right half me hai
                left = mid + 1
            else:
                # minimum left half (mid included) me hai
                right = mid  
        
        return nums[left]

print(Solution().findMin([3, 4, 5, 1, 2]))
```

# 278. First Bad Version

We need to find the first bad version using binary search to minimize API calls.  
All versions after a bad version are also bad  
We have an API: isBadVersion(version) returns True/False  

Example: n=5, bad=4 ‚Üí Output: 4

| Step | Left | Right | `left < right` | Mid `left + (right-left)//2` | `isBadVersion(mid)` | Action               | Explanation                                        |
| ---: | :--: | :---: | :------------: | :--------------------------: | :-----------------: | -------------------- | -------------------------------------------------- |
|    0 |   1  |   5   |     1 < 5 ‚úÖ    |               ‚Äì              |          ‚Äì          | Enter loop           | Initialize search range                            |
|    1 |   1  |   5   |        ‚úÖ       |      `1 + (5-1)//2 = 3`      |        False        | `left = mid + 1 = 4` | Version 3 is good ‚Üí bad version must be after 3    |
|    2 |   4  |   5   |     4 < 5 ‚úÖ    |      `4 + (5-4)//2 = 4`      |         True        | `right = mid = 4`    | Version 4 is bad ‚Üí first bad could be 4 or earlier |
|    3 |   4  |   4   |     4 < 4 ‚ùå    |               ‚Äì              |          ‚Äì          | Exit loop            | Found first bad version                            |


```python
def isBadVersion(version):
    return version >= 4   # maan lo 4 first bad version hai

class Solution:
    def firstBadVersion(self, n: int) -> int:
        left, right = 1, n
        
        while left < right:
            mid = left + (right - left) // 2
            if isBadVersion(mid):
                right = mid
            else:
                left = mid + 1
        
        print("First bad version is:", left)
        return left

sol = Solution()
sol.firstBadVersion(5)
```

# 5. 162 - Find Peak Element

nums = [1, 2, 3, 1]
| Step |  i  | nums[i] | nums[i+1] | Condition `nums[i] > nums[i+1]` | Action                           |
| ---: | :-: | :-----: | :-------: | :-----------------------------: | -------------------------------- |
|    0 |  ‚Äì  |    ‚Äì    |     ‚Äì     |                ‚Äì                | Loop through `range(3)` (0 to 2) |
|    1 |  0  |    1    |     2     |         `1 > 2` ‚ùå False         | Continue                         |
|    2 |  1  |    2    |     3     |         `2 > 3` ‚ùå False         | Continue                         |
|    3 |  2  |    3    |     1     |          `3 > 1` ‚úÖ True         | **Return `i = 2`**               |

output - > 2


```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        for i in range(len(nums) - 1):
            if nums[i] > nums[i + 1]:
                return i
        return len(nums) - 1
```

# 540. Single Element in a Sorted Array

nums = [1, 1, 2, 3, 3, 4, 4, 8, 8]  

| Step |  i  |   nums[i]   |  nums[i+1]  | Condition `nums[i] != nums[i+1]` | Action                                            |
| ---: | :-: | :---------: | :---------: | :------------------------------: | ------------------------------------------------- |
|    0 |  ‚Äì  |      ‚Äì      |      ‚Äì      |                 ‚Äì                | `n = 9`, loop `range(0, 8, 2)` ‚Üí `i = 0, 2, 4, 6` |
|    1 |  0  | nums[0] = 1 | nums[1] = 1 |         `1 != 1` ‚ùå False         | Continue                                          |
|    2 |  2  | nums[2] = 2 | nums[3] = 3 |          `2 != 3` ‚úÖ True         | **Return `nums[2] = 2`**                          |


```python
class Solution:
    def singleNonDuplicate(self, nums: List[int]) -> int:
        n = len(nums) #9    
        # Check pairs
        for i in range(0, n-1, 2):
            if nums[i] != nums[i+1]:       
                return nums[i]    
        # If not found, last element is the single one
        return nums[-1]
print(Solution().singleNonDuplicate([1, 1, 2, 3, 3, 4, 4, 8, 8]))    
```
