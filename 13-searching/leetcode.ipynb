# Leetcode 35. Search Insert Position

### Step-by-Step Dry Run

**Input**
- `nums = [1, 3, 5, 6]`
- `target = 5`

| Step | Left | Right | Mid | nums[mid] | Condition Checked | Action |
|-----:|:----:|:-----:|:---:|:---------:|-------------------|--------|
| 0 | 0 | 3 | – | – | – | Initialize |
| 1 | 0 | 3 | (0+3)//2 = 1 | 3 | 3 == 5 ❌<br>3 < 5 ✅ | `left = mid + 1 = 2` |
| 2 | 2 | 3 | (2+3)//2 = 2 | 5 | 5 == 5 ✅ | **Return 2** |

**Output:** `2`




```python
from typing import List
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        left , right = 0, len(nums) - 1 #4-1=3
        while left <= right: #0<=3
            mid = (left + right)//2 # 0+3//2=1
            #arr[mid=1] =3
            if nums[mid] == target:
                # 3 == 5
                return mid 
            elif  nums[mid] < target:
                #3<5
                left  = mid+ 1 # 1+1 =2
                #arr[2]=5
            else:
                right = mid -1
        return left          
print(Solution().searchInsert([1,3,5,6], 5))            
```

# Leetcode - 34. Find First and Last Position of Element in Sorted Array

🔹 Idea

Use binary search twice:

Once to find the first occurrence (move left when nums[mid] == target).

Once to find the last occurrence (move right when nums[mid] == target).

If the target is not found, return [-1, -1].

nums = [5, 7, 7, 8, 8, 10]   
target = 8   
Table 1: findFirst(nums, target) — Find First Occurrence  

| Step | Left | Right |      Mid     | nums[mid] | Condition    | Action                   | First |
| ---: | :--: | :---: | :----------: | :-------: | ------------ | ------------------------ | :---: |
|    0 |   0  |   5   |       –      |     –     | –            | –                        |   -1  |
|    1 |   0  |   5   | (0+5)//2 = 2 |     7     | 7 < 8        | `left = 3`               |   -1  |
|    2 |   3  |   5   | (3+5)//2 = 4 |     8     | 8 == 8       | `first = 4`, `right = 3` |   4   |
|    3 |   3  |   3   | (3+3)//2 = 3 |     8     | 8 == 8       | `first = 3`, `right = 2` |   3   |
|    4 |   3  |   2   |       –      |     –     | left > right | Exit loop                |   3   |

Result: findFirst returns 3 ✅  

Table 2: findLast(nums, target) — Find Last Occurrence

| Step | Left | Right |      Mid     | nums[mid] | Condition    | Action                 | Last |
| ---: | :--: | :---: | :----------: | :-------: | ------------ | ---------------------- | :--: |
|    0 |   0  |   5   |       –      |     –     | –            | –                      |  -1  |
|    1 |   0  |   5   | (0+5)//2 = 2 |     7     | 7 < 8        | `left = 3`             |  -1  |
|    2 |   3  |   5   | (3+5)//2 = 4 |     8     | 8 == 8       | `last = 4`, `left = 5` |   4  |
|    3 |   5  |   5   | (5+5)//2 = 5 |     10    | 10 > 8       | `right = 4`            |   4  |
|    4 |   5  |   4   |       –      |     –     | left > right | Exit loop              |   4  |

Result: findLast returns 4 ✅     
output :- [3, 4]


```python
from typing import List

class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        def findFirst(nums, target):
            left , right = 0, len(nums)-1#6-1=5
            first = -1
            while left<= right:# 0<5
                mid = (left+right)//2 #mid[2]=7
                if nums[mid] == target:#7==8, 8==8
                    first = mid #4
                    right = mid -1
                elif nums[mid] < target:#7<8
                    left = mid +1 # 2+ 1 =3
                else:
                    right = mid -1
            return first        
     #last            
        def findLast(nums, target):
            left, right = 0, len(nums) - 1
            last = -1
            while left<= right:# 0<5
                mid = (left+right)//2 #mid[2]=7
                if nums[mid] == target:#7==8
                    last = mid
                    left = mid + 1
                elif nums[mid] < target:#7<8
                    left = mid + 1 # 2+ 1 =3    
                else:
                    right = mid -1
            return last    
        return [findFirst(nums, target),findLast(nums, target) ]                    
```

# 33. Search in Rotated Sorted Array

nums = [4, 5, 6, 7, 0, 1, 2]  
target = 0

| Step | Left | Right | Mid `(left+right)//2` | nums[mid] | Comparison & Check                                                                 | Action                | Next Range      |
| ---: | :--: | :---: | :-------------------: | :-------: | ---------------------------------------------------------------------------------- | --------------------- | --------------- |
|    0 |   0  |   6   |           –           |     –     | Initialization                                                                     | Begin loop            | `[0, 6]`        |
|    1 |   0  |   6   |      (0+6)//2 = 3     |     7     | 1. `7 == 0` ❌ <br> 2. Left sorted? `4 ≤ 7` ✅ <br> 3. Target in left? `4 ≤ 0 < 7` ❌ | `left = mid + 1 = 4`  | `[4, 6]`        |
|    2 |   4  |   6   |      (4+6)//2 = 5     |     1     | 1. `1 == 0` ❌ <br> 2. Left sorted? `0 ≤ 1` ✅ <br> 3. Target in left? `0 ≤ 0 < 1` ✅ | `right = mid - 1 = 4` | `[4, 4]`        |
|    3 |   4  |   4   |      (4+4)//2 = 4     |     0     | `0 == 0` ✅                                                                         | **Return `mid = 4`**  | Target found 🎯 |


```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums) - 1 #7-1=6
        while left <= right:#0<=6, 4<=6
            mid =(left + right)//2 #0+6//2=3=>mid[3] =7, 5

            if nums[mid] == target:#7==0, 1==0
                return mid
            if nums[left] <= nums[mid]:#4<=7, 0<= 1
                if nums[left] <= target < nums[mid]: #4<=0<7, 0<=0<7
                    right = mid -1
                else:
                    left = mid + 1 #4

            else:
                if nums[mid] < target <=nums[right]:
                    left = mid+1
                else:
                    right = mid - 1
            
print(Solution().search([4, 5, 6, 7, 0, 1, 2],0))    
```

# 153 - Find Minimum in Rotated Sorted Array

nums = [3, 4, 5, 1, 2]

| Step | Left | Right | `left < right` | Mid | nums[mid] | nums[right] | Condition | Action           | Next           |
| ---: | :--: | :---: | :------------: | :-: | :-------: | :---------: | :-------: | :--------------- | :------------- |
|    0 |   0  |   4   |        ✅       |  –  |     –     |      –      |     –     | enter loop       | L=0 R=4        |
|    1 |   0  |   4   |        ✅       |  2  |     5     |      2      |  5 > 2 ✅  | `left = mid + 1` | L=3 R=4        |
|    2 |   3  |   4   |        ✅       |  3  |     1     |      2      |  1 > 2 ❌  | `right = mid`    | L=3 R=3        |
|    3 |   3  |   3   |        ❌       |  –  |     –     |      –      |     –     | exit loop        | return nums[3] |

Return nums[3] = 1

```python
from typing import List
class Solution:
    def findMin(self, nums:List[int])->int:
        left , right = 0, len(nums)-1
        while left < right: #0< 4, 3<4, 3<3
            mid = (left + right)//2#0+4//2 =2, 3
            #nums[mid=2,3] =5, 1
            if nums[mid] > nums[right]:#5>2, 1>2
                left = mid + 1 # 3
            else:
                right = mid   #3  
        return nums[left]        
print(Solution().findMin([3, 4, 5, 1, 2]))
```

# 278. First Bad Version

We need to find the first bad version using binary search to minimize API calls.  
All versions after a bad version are also bad  
We have an API: isBadVersion(version) returns True/False  

Example: n=5, bad=4 → Output: 4

| Step | Left | Right | `left < right` | Mid `left + (right-left)//2` | `isBadVersion(mid)` | Action               | Explanation                                        |
| ---: | :--: | :---: | :------------: | :--------------------------: | :-----------------: | -------------------- | -------------------------------------------------- |
|    0 |   1  |   5   |     1 < 5 ✅    |               –              |          –          | Enter loop           | Initialize search range                            |
|    1 |   1  |   5   |        ✅       |      `1 + (5-1)//2 = 3`      |        False        | `left = mid + 1 = 4` | Version 3 is good → bad version must be after 3    |
|    2 |   4  |   5   |     4 < 5 ✅    |      `4 + (5-4)//2 = 4`      |         True        | `right = mid = 4`    | Version 4 is bad → first bad could be 4 or earlier |
|    3 |   4  |   4   |     4 < 4 ❌    |               –              |          –          | Exit loop            | Found first bad version                            |


```python
def isBadVersion(version):
    return version >= 4 # man lo 4 first bad version

    
class Solution(object):
    def firstBadVersion(self, n):
        left , right = 1, n
        while left < right: #1<5
            mid = left+(right-left)//2 #3,4
            if isBadVersion(mid):
                right = mid #4
            else:
                left = mid +1  #4  
        print("first bad version is:", left)        
sol = Solution()
sol.firstBadVersion(5)
```

# 5. 162 - Find Peak Element

nums = [1, 2, 3, 1]
| Step |  i  | nums[i] | nums[i+1] | Condition `nums[i] > nums[i+1]` | Action                           |
| ---: | :-: | :-----: | :-------: | :-----------------------------: | -------------------------------- |
|    0 |  –  |    –    |     –     |                –                | Loop through `range(3)` (0 to 2) |
|    1 |  0  |    1    |     2     |         `1 > 2` ❌ False         | Continue                         |
|    2 |  1  |    2    |     3     |         `2 > 3` ❌ False         | Continue                         |
|    3 |  2  |    3    |     1     |          `3 > 1` ✅ True         | **Return `i = 2`**               |

output - > 2


```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        for i in range(len(nums) - 1):
            if nums[i] > nums[i+1]:#1>2, 2>3, 3>1
                return i #2
        return len(nums)-1  
print(Solution().findPeakElement([1,2,3,1]))                    
```

# 540. Single Element in a Sorted Array

nums = [1, 1, 2, 3, 3, 4, 4, 8, 8]  

| Step |  i  |   nums[i]   |  nums[i+1]  | Condition `nums[i] != nums[i+1]` | Action                                            |
| ---: | :-: | :---------: | :---------: | :------------------------------: | ------------------------------------------------- |
|    0 |  –  |      –      |      –      |                 –                | `n = 9`, loop `range(0, 8, 2)` → `i = 0, 2, 4, 6` |
|    1 |  0  | nums[0] = 1 | nums[1] = 1 |         `1 != 1` ❌ False         | Continue                                          |
|    2 |  2  | nums[2] = 2 | nums[3] = 3 |          `2 != 3` ✅ True         | **Return `nums[2] = 2`**                          |


```python
class Solution:
    def singleNonDuplicate(self, nums: List[int]) -> int:
        n = len(nums)
        for i in range(0, n-1, 2):# start, stop, step
            if nums[i] != nums[i+1]: #1 != 1, 2!=3
                return nums[i]    
        return -1         
print(Solution().singleNonDuplicate([1, 2, 3, 4, 4, 8, 8]))    
```
