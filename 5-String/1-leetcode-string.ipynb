### 1  Custom â€“ Count vowels and consonants (JSON output)


```python
import json
def count_vowels_cosonants(s):
  vowels = "aioueAOIUE"
  vowels_count = 0
  consonants_count =0
  for ch in s:
    if ch.isalpha():
      if ch in vowels:
        vowels_count += 1
      else:
        consonants_count += 1
  result ={"vowels_count":vowels_count, "consonants_count":consonants_count}
  return json.dumps(result, indent=4)
print(count_vowels_cosonants("Hello"))
```

### 2. #5 Longest Palindromic Substring

```python
def longestPalindrome(s):
    n = len(s)
    ans = ""

    for i in range(n):
        # Odd length palindrome (like "aba")
        l, r = i, i
        while l >= 0 and r < n and s[l] == s[r]:
            if (r - l + 1) > len(ans):
                ans = s[l:r+1]
            l -= 1
            r += 1

        # Even length palindrome (like "abba")
        l, r = i, i+1
        while l >= 0 and r < n and s[l] == s[r]:
            if (r - l + 1) > len(ans):
                ans = s[l:r+1]
            l -= 1
            r += 1

    return ans
```

### 3 #6. Zigzag Conversion

```python
class Solution(object):
    def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """
        # Edge case: single row means no zigzag
        if numRows == 1 or numRows >= len(s):
            return s

        # Buckets for each row
        rows = [""] * numRows

        curRow = 0
        goingDown = False

        for c in s:
            rows[curRow] += c
            # Top or bottom row â†’ change direction
            if curRow == 0 or curRow == numRows - 1:
                goingDown = not goingDown
            # Move up or down
            curRow += 1 if goingDown else -1

        # Join all rows
        return "".join(rows)
```

### 4- #14. Longest Common Prefix


```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs:
            return ""

        # Take the first string as prefix
        prefix = strs[0]

        # Compare with the rest of the strings
        for s in strs[1:]:
            # Keep reducing prefix until it matches the beginning of s
            while not s.startswith(prefix):
                prefix = prefix[:-1]  # shrink from end
                if not prefix:       # if prefix becomes empty, no common prefix
                    return ""

        return prefix
```
