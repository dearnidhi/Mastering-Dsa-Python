# Recursion ‚Äî The Concept
### üß† What is Recursion?

Recursion means:
üëâ A function calls itself until it reaches a base condition.

It breaks a big problem into smaller subproblems

### üîë 3 Parts of Every Recursive Function:

Base Case ‚Üí When to stop calling itself

Work / Processing ‚Üí What to do in the current step

Recursive Call ‚Üí Call the same function with a smaller input

### üîÅ Basic Structure of a Recursive Function

```python
def func():
    # 1Ô∏è‚É£ Base case ‚Äì when to stop
    if condition:
        return something
    
    # 2Ô∏è‚É£ Recursive case ‚Äì call itself
    else:
        return func(smaller_input)
```

### 1: Print Numbers from 1 to N

```python
def print_nums(n):
    if n == 0:        # base case
        return
    print_nums(n - 1) # recursive call
    print(n)          # after recursion (ascending order)

print_nums(5)
```

### 2 Simple Recursion Example

Explanation:

This is classic recursion with counting up after recursion.

recursiveMethod(4) first goes down to 0 (base case), then prints numbers while returning back.

```python
def RecursiveMethod(n):
    if n < 1:
        print("n is less than 1")
    else:
        RecursiveMethod(n-1)
        print(n)
RecursiveMethod(4)            
```

### 3 Factorial Function

```python
def fac(n):
    if n in [0,1]:
        return 1
    else:
        return n *fac(n-1)
print(fac(4))    
```

Recursive factorial.

Base case: 0! = 1 or 1! = 1

Recursive case: n! = n * (n-1)!

### 4: Fibonacci Sequence

Calculates nth Fibonacci number.

Base cases: fibonacci(0)=0, fibonacci(1)=1

Recursive case: fibonacci(n) = fibonacci(n-1) + fibonacci(n-2)

Output: 13

Sequence: 0, 1, 1, 2, 3, 5, 8, 13

```python
def fib(n):
    if n in [0,1]:
        return 1
    else:
        return fib(n-1)+fib(n-2)
print(fib(7))    
 
```

### 5: Sum of Digits

```python
def sum_of_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_of_digits(n // 10)
```

### 6: Reverse a String

```python
def reverse_str(s):
    if len(s) == 0:
        return ""
    return reverse_str(s[1:]) + s[0]
```
