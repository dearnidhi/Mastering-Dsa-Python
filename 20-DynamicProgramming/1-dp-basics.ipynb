## 1. What is Dynamic Programming?

DP = **Recursion + Caching (remembering answers)**

That's it. Nothing more.

### The Problem with Plain Recursion

Calculate Fibonacci: `fib(5)`

```
fib(5)
├── fib(4)
│   ├── fib(3)
│   │   ├── fib(2)   <-- calculated
│   │   └── fib(1)
│   └── fib(2)       <-- calculated AGAIN (wasted!)
└── fib(3)           <-- calculated AGAIN (wasted!)
    ├── fib(2)       <-- calculated AGAIN (wasted!)
    └── fib(1)
```

Same subproblems computed multiple times -> **exponential time O(2^n)**

### The DP Fix: Remember what you already solved

If `fib(3)` is already computed, just **look it up** instead of recomputing.
Now each subproblem is solved only **once** -> **O(n) time**

### Two Conditions for DP
1. **Overlapping Subproblems** — same subproblem appears multiple times
2. **Optimal Substructure** — optimal answer can be built from optimal answers of subproblems

---
## 2. Two Ways to Write DP

### Way 1 — Memoization (Top-Down)
- Write recursion normally
- Add a cache (dictionary) to store results
- Before computing, check if answer is already in cache
- Feels natural, start from big problem, break it down

### Way 2 — Tabulation (Bottom-Up)
- Build answer from smallest subproblem up to the big one
- Use a dp array/table
- No recursion, uses a for loop
- Usually more space efficient

**Both give the same answer. Choose whichever feels easier for the problem.**

## 3. First Example — Fibonacci

`fib(n) = fib(n-1) + fib(n-2)`,  base: `fib(0)=0, fib(1)=1`

````python
# Plain recursion — SLOW O(2^n)
def fib_slow(n):
    if n <= 1:
        return n
    return fib_slow(n-1) + fib_slow(n-2)

print(fib_slow(10))  # 55  (but fib_slow(50) will hang!)
````

````python
# Way 1: Memoization (Top-Down) — O(n)
def fib_memo(n, cache={}):
    if n <= 1:
        return n
    if n in cache:          # already solved? return it
        return cache[n]
    cache[n] = fib_memo(n-1, cache) + fib_memo(n-2, cache)
    return cache[n]

print(fib_memo(50))   # 12586269025  (instant!)
````

````python
# Cleaner memoization using Python's lru_cache
from functools import lru_cache

@lru_cache(maxsize=None)
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(50))   # 12586269025
````

````python
# Way 2: Tabulation (Bottom-Up) — O(n)
def fib_tab(n):
    if n <= 1:
        return n
    dp = [0] * (n + 1)
    dp[0] = 0   # base case
    dp[1] = 1   # base case
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]   # recurrence
    return dp[n]

print(fib_tab(50))   # 12586269025
print("dp table for n=7:", [fib_tab(i) for i in range(8)])
````

## 4. The 5-Step Framework to Solve ANY DP Problem

Follow these steps every time:

```
Step 1: Define what dp[i] means (the state)
Step 2: Find the recurrence relation (how dp[i] depends on smaller values)
Step 3: Set the base cases
Step 4: Decide iteration order (left to right? right to left?)
Step 5: Return the answer (dp[n] or dp[n][m] etc.)
```

**The hardest part is Step 1 — defining the state correctly.**
Everything else follows from it.

## 5. Classic 1D DP — Climbing Stairs (LeetCode 70)

**Problem:** You can climb 1 or 2 steps at a time. How many ways to reach step `n`?

**Step 1 — State:** `dp[i]` = number of ways to reach step `i`

**Step 2 — Recurrence:**
To reach step `i`, you came from:
- step `i-1` (took 1 step), OR
- step `i-2` (took 2 steps)
```
dp[i] = dp[i-1] + dp[i-2]
```
(Same as Fibonacci!)

**Step 3 — Base cases:** `dp[1] = 1`,  `dp[2] = 2`

**Step 4 — Order:** left to right

**Step 5 — Answer:** `dp[n]`

````python
def climbStairs(n):
    if n <= 2:
        return n
    dp = [0] * (n + 1)
    dp[1] = 1
    dp[2] = 2
    for i in range(3, n + 1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

for i in range(1, 8):
    print(f"n={i} -> {climbStairs(i)} ways")
````

## 6. Classic 1D DP — House Robber (LeetCode 198)

**Problem:** Houses in a row with money. Can't rob two adjacent houses. Max money?

```
houses = [2, 7, 9, 3, 1]
Rob:          2     9     1  = 12  (skip 7 and 3)
```

**Step 1 — State:** `dp[i]` = max money robbing from house 0 to house i

**Step 2 — Recurrence:** At house `i`, two choices:
- Skip house `i` -> `dp[i-1]`
- Rob house `i` -> `nums[i] + dp[i-2]`
```
dp[i] = max(dp[i-1], nums[i] + dp[i-2])
```

**Step 3 — Base cases:** `dp[0] = nums[0]`,  `dp[1] = max(nums[0], nums[1])`

**Step 5 — Answer:** `dp[n-1]`

````python
def rob(nums):
    n = len(nums)
    if n == 1:
        return nums[0]

    dp = [0] * n
    dp[0] = nums[0]
    dp[1] = max(nums[0], nums[1])

    for i in range(2, n):
        dp[i] = max(dp[i-1],          # skip house i
                    nums[i] + dp[i-2]) # rob house i
    return dp[n-1]

print(rob([2, 7, 9, 3, 1]))   # 12
print(rob([1, 2, 3, 1]))      # 4

# Visualize dp table
nums = [2, 7, 9, 3, 1]
n = len(nums)
dp = [0]*n
dp[0] = nums[0]; dp[1] = max(nums[0], nums[1])
for i in range(2, n):
    dp[i] = max(dp[i-1], nums[i] + dp[i-2])
print("nums:", nums)
print("dp:  ", dp)
````

## 7. Classic 1D DP — Coin Change (LeetCode 322)

**Problem:** Given coins and a target amount, find minimum coins needed.

```
coins = [1, 5, 6, 9],  amount = 11
Answer: 2  (5 + 6)
```

**Step 1 — State:** `dp[i]` = min coins to make amount `i`

**Step 2 — Recurrence:** For each coin, if we use it:
```
dp[i] = min(dp[i], 1 + dp[i - coin])   for each coin <= i
```

**Step 3 — Base cases:** `dp[0] = 0` (0 coins to make amount 0)

**Step 4 — Order:** left to right (amount 0 to target)

**Step 5 — Answer:** `dp[amount]`

````python
def coinChange(coins, amount):
    dp = [float('inf')] * (amount + 1)  # inf = impossible
    dp[0] = 0                           # base case

    for i in range(1, amount + 1):      # for every amount
        for coin in coins:              # try every coin
            if coin <= i:
                dp[i] = min(dp[i], 1 + dp[i - coin])

    return dp[amount] if dp[amount] != float('inf') else -1


print(coinChange([1,5,6,9], 11))   # 2  (5+6)
print(coinChange([1,2,5], 11))     # 3  (5+5+1)
print(coinChange([2], 3))          # -1 (impossible)

# Visualize
coins, amount = [1,5,6,9], 11
dp = [float('inf')] * (amount + 1)
dp[0] = 0
for i in range(1, amount+1):
    for c in coins:
        if c <= i: dp[i] = min(dp[i], 1 + dp[i-c])
print("dp:", dp)
````

## 8. Classic 1D DP — Longest Increasing Subsequence (LeetCode 300)

**Problem:** Find length of longest strictly increasing subsequence.

```
nums = [10, 9, 2, 5, 3, 7, 101, 18]
LIS  =  [2, 5, 7, 101]  ->  length 4
```

**Step 1 — State:** `dp[i]` = length of LIS ending at index `i`

**Step 2 — Recurrence:**
For each `j < i`, if `nums[j] < nums[i]`, we can extend:
```
dp[i] = max(dp[i], dp[j] + 1)  for all j < i where nums[j] < nums[i]
```

**Step 3 — Base cases:** Every element alone is LIS of length 1 -> `dp[i] = 1`

**Step 5 — Answer:** `max(dp)`

````python
def lengthOfLIS(nums):
    n = len(nums)
    dp = [1] * n     # every element is LIS of length 1 by itself

    for i in range(1, n):
        for j in range(i):             # look at all previous elements
            if nums[j] < nums[i]:      # can extend
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)


nums = [10, 9, 2, 5, 3, 7, 101, 18]
print(lengthOfLIS(nums))   # 4

# Visualize
dp = [1] * len(nums)
for i in range(1, len(nums)):
    for j in range(i):
        if nums[j] < nums[i]: dp[i] = max(dp[i], dp[j]+1)
print("nums:", nums)
print("dp:  ", dp, "-> answer:", max(dp))
````

## 9. 2D DP — Introduction

When the problem depends on **two variables** (like two strings, or row+col in a grid), you need a 2D dp table.

```
dp[i][j] = answer considering first i chars of string1 and first j chars of string2
```

Common 2D DP problems:
- Longest Common Subsequence
- Edit Distance
- 0/1 Knapsack
- Unique Paths

---
## 10. 2D DP — Unique Paths (LeetCode 62)

**Problem:** Grid of m x n. Start at top-left, reach bottom-right. Only move right or down. How many paths?

```
3x3 grid:
S . .
. . .
. . E

Answer: 6 paths
```

**Step 1 — State:** `dp[i][j]` = number of paths to reach cell (i, j)

**Step 2 — Recurrence:** Can only come from top or left:
```
dp[i][j] = dp[i-1][j] + dp[i][j-1]
```

**Step 3 — Base cases:** First row and first column = 1 (only one way to reach them)

````python
def uniquePaths(m, n):
    dp = [[1] * n for _ in range(m)]   # fill with 1 (base case for edges)

    for i in range(1, m):
        for j in range(1, n):
            dp[i][j] = dp[i-1][j] + dp[i][j-1]

    return dp[m-1][n-1]


print(uniquePaths(3, 3))   # 6
print(uniquePaths(3, 7))   # 28

# Visualize the table
m, n = 3, 3
dp = [[1]*n for _ in range(m)]
for i in range(1, m):
    for j in range(1, n):
        dp[i][j] = dp[i-1][j] + dp[i][j-1]
print("dp table:")
for row in dp:
    print(row)
````

## 11. 2D DP — Longest Common Subsequence (LeetCode 1143)

**Problem:** Given two strings, find length of their longest common subsequence.

```
s1 = "abcde"
s2 = "ace"
LCS = "ace"  ->  length 3
```

**Step 1 — State:** `dp[i][j]` = LCS of `s1[:i]` and `s2[:j]`

**Step 2 — Recurrence:**
```
if s1[i-1] == s2[j-1]:   # chars match -> extend LCS
    dp[i][j] = dp[i-1][j-1] + 1
else:                     # skip one char from either string
    dp[i][j] = max(dp[i-1][j], dp[i][j-1])
```

**Step 3 — Base cases:** `dp[0][j] = 0`,  `dp[i][0] = 0`  (empty string -> LCS = 0)

````python
def longestCommonSubsequence(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]   # extra row/col for empty string

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:                  # chars match
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]


print(longestCommonSubsequence("abcde", "ace"))    # 3
print(longestCommonSubsequence("abc", "abc"))      # 3
print(longestCommonSubsequence("abc", "def"))      # 0

# Visualize the table
s1, s2 = "abcde", "ace"
m, n = len(s1), len(s2)
dp = [[0]*(n+1) for _ in range(m+1)]
for i in range(1, m+1):
    for j in range(1, n+1):
        if s1[i-1] == s2[j-1]: dp[i][j] = dp[i-1][j-1]+1
        else: dp[i][j] = max(dp[i-1][j], dp[i][j-1])
print(f"\nDP table for s1={s1!r}, s2={s2!r}")
print("    '' " + " ".join(s2))
labels = ["''"] + list(s1)
for i, row in enumerate(dp):
    print(f"  {labels[i]}  {row}")
````

## 12. 0/1 Knapsack — The Foundation of Many DP Problems

**Problem:** Items with weights and values. Bag has capacity W. Max value you can carry?
Each item can be taken at most once (0 or 1 time).

```
weights = [1, 3, 4, 5]
values  = [1, 4, 5, 7]
W = 7
Answer: 9  (take items with weight 3 and 4, values 4+5)
```

**Step 1 — State:** `dp[i][w]` = max value using first `i` items with capacity `w`

**Step 2 — Recurrence:**
```
For item i (0-indexed), weight=wt[i], value=val[i]:

Option A: Don't take item i  ->  dp[i-1][w]
Option B: Take item i (only if wt[i] <= w)  ->  val[i] + dp[i-1][w - wt[i]]

dp[i][w] = max(Option A, Option B)
```

````python
def knapsack(weights, values, W):
    n = len(weights)
    dp = [[0] * (W + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(W + 1):
            # Option A: skip item i
            dp[i][w] = dp[i-1][w]
            # Option B: take item i (if it fits)
            if weights[i-1] <= w:
                take = values[i-1] + dp[i-1][w - weights[i-1]]
                dp[i][w] = max(dp[i][w], take)

    return dp[n][W]


weights = [1, 3, 4, 5]
values  = [1, 4, 5, 7]
print(knapsack(weights, values, 7))   # 9
````

## 13. Space Optimization (Important for Interviews)

Often `dp[i]` only depends on `dp[i-1]`.
In that case you don't need the full table — just two variables or one rolling array.

Example — Fibonacci with O(1) space:

````python
# O(n) space -> O(1) space
def fib_optimized(n):
    if n <= 1:
        return n
    prev2, prev1 = 0, 1
    for _ in range(2, n + 1):
        curr = prev1 + prev2
        prev2 = prev1
        prev1 = curr
    return prev1

print(fib_optimized(50))   # 12586269025


# House robber with O(1) space
def rob_optimized(nums):
    prev2, prev1 = 0, 0
    for num in nums:
        curr = max(prev1, num + prev2)
        prev2 = prev1
        prev1 = curr
    return prev1

print(rob_optimized([2, 7, 9, 3, 1]))   # 12
````

## 14. Memoization Template (use this in interviews)

When you don't know the recurrence clearly, start with memoization — it's easier to think recursively.

````python
from functools import lru_cache

# Generic memoization template
# Replace the body with your problem's logic

def solve(input_data):
    @lru_cache(maxsize=None)
    def dp(state1, state2):         # define your state variables
        # base case
        if state1 == 0:
            return 0
        # recurrence
        option_a = dp(state1 - 1, state2)        # one choice
        option_b = dp(state1 - 1, state2 - 1)    # another choice
        return max(option_a, option_b)

    return dp(len(input_data), 0)

print("Template ready — replace state and recurrence with your problem!")
````

## 15. LeetCode DP Patterns Cheatsheet

| Pattern | Example Problems | Key State |
|---|---|---|
| Linear DP | Climbing Stairs, House Robber | `dp[i]` |
| Kadane's (max subarray) | Maximum Subarray (LC 53) | `dp[i]` = max ending here |
| Unbounded Knapsack | Coin Change, Coin Change II | `dp[amount]` |
| 0/1 Knapsack | Partition Equal Subset Sum | `dp[i][w]` |
| LCS pattern | LCS, Edit Distance, Distinct Subsequences | `dp[i][j]` |
| LIS pattern | LIS, Russian Doll Envelopes | `dp[i]` |
| Grid DP | Unique Paths, Min Path Sum | `dp[i][j]` |
| Interval DP | Burst Balloons, Matrix Chain | `dp[l][r]` |
| Tree DP | House Robber III | DFS return value |

## 16. How to Recognize a DP Problem on LeetCode

Look for these signals:

```
"minimum / maximum"         -> optimize something
"how many ways / paths"     -> count something
"can you / is it possible"  -> yes/no with optimal choices
"at most k times"           -> constrained choices
```

And the input usually is:
- An array/string (1D DP)
- Two arrays/strings (2D DP)
- A grid (2D DP)
- A number target (Knapsack-style)

## Summary

```
WHAT IS DP?
  Recursion + Caching = DP
  Same subproblem appears multiple times -> store result -> O(n) instead of O(2^n)

TWO WAYS:
  Memoization  = Top-Down  = Recursion + cache dict / @lru_cache
  Tabulation   = Bottom-Up = for loop + dp array

5-STEP FRAMEWORK:
  1. Define dp[i] (or dp[i][j]) clearly
  2. Write recurrence relation
  3. Set base cases
  4. Decide iteration order
  5. Return dp[n] or dp[n][m]

PATTERNS:
  1D:  Fibonacci, Climbing Stairs, House Robber, Coin Change, LIS
  2D:  Unique Paths, LCS, Edit Distance, Knapsack

SPACE TRICK:
  If dp[i] only needs dp[i-1], use two variables instead of full array

SPOT DP IN PROBLEMS:
  "minimum", "maximum", "how many ways", "is it possible" + array/string/grid input
```
