 # 5ï¸âƒ£ 20 â€“ Valid Parentheses


### Dry Run: Valid Parentheses

**Example:** `s = "{[]}"`

---

**Initial:**  
`s = "{[]}"`

**Step 1:**  
`'[]' in "{[]}"` â†’ True  
`s = "{[]}".replace('[]', '')` = `"{}"`

**Step 2:**  
`'{}' in "{}"` â†’ True  
`s = "{}".replace('{}', '')` = `""`

**Return:**  
`"" == ""` â†’ True âœ…


```python
class Solution:
    def isValid(self, s: str) -> bool:
        while '()' in s or '{}' in s or '[]' in s:
            s = s.replace('()', '').replace('{}', '').replace('[]', '')
        return s == '' #Agar string poori khaali ho gayi toh True, nahi toh False
sol = Solution()
print(sol.isValid("()"))         # True
print(sol.isValid("()[]{}"))     # True
print(sol.isValid("(]"))         # False
print(sol.isValid("([)]"))       # False
print(sol.isValid("{[]}"))       # True    
```

# 6ï¸âƒ£ 28 â€“ Implement strStr() (Find Substring)

Explanation:

If needle is empty, return 0.

Loop through haystack only until the remaining substring can fit needle (len_h - len_n + 1).

Compare substring haystack[i:i+len_n] with needle.

Return i if matched, else continue.

If no match found, return -1.

Dry Run Example

Input: haystack = "sadbutsad", needle = "sad"

| i | haystack[i:i+len(needle)] | Compare with needle | Match? | Action   |
| - | ------------------------- | ------------------- | ------ | -------- |
| 0 | "sad"                     | "sad"               | âœ… Yes  | Return 0 |
| 1 | "adb"                     | "sad"               | âŒ No   | Continue |
| 2 | "dbu"                     | "sad"               | âŒ No   | Continue |
| 3 | "but"                     | "sad"               | âŒ No   | Continue |
| 4 | "uts"                     | "sad"               | âŒ No   | Continue |
| 5 | "tsa"                     | "sad"               | âŒ No   | Continue |
| 6 | "sad"                     | "sad"               | âœ… Yes  | Return 6 |

Result: 0 (first occurrence at index 0)

```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if not needle:  # If needle is empty, return 0
            return 0
        len_h = len(haystack) # 9
        len_n = len(needle) # 3
        for i in range(len_h-len_n+1): #9-3+1=7
            # Check substring starting at i
            if haystack[i:i+len_n]==needle: #haystack[0:0+3]=sad ,needle=sad
                return i
        return -1  
sol = Solution()
print(sol.strStr("sadbutsad", "sad"))    # 0
print(sol.strStr("leetcode", "leeto"))   # -1
print(sol.strStr("hello", "ll"))         # 2
```



 # 7ï¸âƒ£ 58 â€“ Length of Last Word

# Dry Run: Length of Last Word
**Example:** `"Hello World"`

---

## Step 0: Initial Setup
- `length = 0`
- `i = len(s) - 1 = 10` (start from last character)
- `s = "Hello World"`

---

## Step 1: Skip Trailing Spaces
**Goal:** Ignore any spaces at the end of the string.  

| i  | s[i] | Condition        | Action                  |
|----|------|-----------------|------------------------|
| 10 | 'd'  | Not a space      | Skip loop, i stays 10  |

**Result after Step 1:**  
- `i = 10`  
- `length = 0`

---

## Step 2: Count Length of Last Word
**Goal:** Count characters of the last word moving backward.  

| i  | s[i] | Condition          | length | i after decrement |
|----|------|------------------|--------|-----------------|
| 10 | 'd'  | Not a space        | 0 + 1 = 1 | 9 |
| 9  | 'l'  | Not a space        | 1 + 1 = 2 | 8 |
| 8  | 'r'  | Not a space        | 2 + 1 = 3 | 7 |
| 7  | 'o'  | Not a space        | 3 + 1 = 4 | 6 |
| 6  | 'W'  | Not a space        | 4 + 1 = 5 | 5 |
| 5  | ' '  | Is a space         | STOP       | length = 5 |

---

## Step 3: Return Result
- `length = 5` âœ…
- Last word = `"World"`  
- Output = `5`


```python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        length = 0
        i = len(s) - 1
        
        # Skip trailing spaces
        while i >= 0 and s[i] == " ":#space
            i -= 1
        
        # Count length of last word
        while i >= 0 and s[i] != " ": #no space
            length += 1
            i -= 1
        
        return length
sol = Solution()
print(sol.lengthOfLastWord("Hello World"))           # 5
print(sol.lengthOfLastWord("   fly me   to   the moon  "))  # 4
print(sol.lengthOfLastWord("luffy is still joyboy")) # 6
```

```python
class Solution:
    def lengthOfLastWord(self, s: str) -> int:
        return len(s.split()[-1])
sol = Solution()
print(sol.lengthOfLastWord("Hello World"))           # 5
print(sol.lengthOfLastWord("   fly me   to   the moon  "))  # 4
print(sol.lengthOfLastWord("luffy is still joyboy")) # 6    
```

# 9ï¸âƒ£ 151 â€“ Reverse Words in a String



Given a string s, reverse the order of words.

Words are separated by spaces. Remove any extra spaces.

Example: " the sky is blue " â†’ "blue is sky the"

Dry Run Example

Input: s = " the sky is blue "

| Step | Action            | Value                                  |
| ---- | ----------------- | -------------------------------------- |
| 1    | `split()`         | `words = ["the", "sky", "is", "blue"]` |
| 2    | `reverse()`       | `words = ["blue", "is", "sky", "the"]` |
| 3    | `" ".join(words)` | `"blue is sky the"`                    |


```python
class Solution:
    def reverseWords(self, s: str) -> str:
        # Step 1: Split string into words (split automatically removes extra spaces)
        words = s.split()
        
        # Step 2: Reverse the list of words
        words.reverse()
        
        # Step 3: Join words with single space
        return " ".join(words)
sol = Solution()
print(sol.reverseWords("the sky is blue"))    # "blue is sky the"
print(sol.reverseWords("  hello world  "))    # "world hello"
print(sol.reverseWords("a good   example"))   # "example good a"
```

```python
class Solution:
    def reverseWords(self, s: str) -> str:
        # Remove extra spaces and split
        words = s.strip().split()
        
        # Reverse the list of words
        reversed_words = words[::-1]
        
        # Join back with single space
        return ' '.join(reversed_words)
```

# ðŸ”Ÿ 242 â€“ Valid Anagram âœ… (new â€” fits string type)


Input: s = "anagram", t = "nagaram"

| Step | Action  | s_sorted                      | t_sorted                      | Result |
| ---- | ------- | ----------------------------- | ----------------------------- | ------ |
| 1    | Sort s  | ['a','a','a','g','m','n','r'] | -                             | -      |
| 2    | Sort t  | -                             | ['a','a','a','g','m','n','r'] | -      |
| 3    | Compare | ['a','a','a','g','m','n','r'] | ['a','a','a','g','m','n','r'] | âœ… True |


```python
class Solution:
    def isAnagram(self, s:str, t:str)->bool:
        return sorted(s) == sorted(t)

sol = Solution()
print(sol.isAnagram("anagram", "nagaram"))  # True
print(sol.isAnagram("rat", "car"))          # False    
```
