# üß† Memory Representation
In Python, strings are stored as arrays of characters in contiguous memory locations, but being immutable, modifications require creating a new string.

üîç Common String Operations (with Examples)

```python
# 1Ô∏è‚É£ Length of String
s = "Nidhi"
print(len(s))
```

```python
# 2Ô∏è‚É£ Search a Character
s ="Chouhan"
print(s.find('h'))
print(s.rfind('h'))
```

```python
# 3Ô∏è‚É£ Check for Substring
s = "bhopal"
print(s.startswith("bho"))
print(s.endswith("pal"))
print("bho" in s)


```

```python
#4Ô∏è‚É£ Insert a Character
#Since strings are immutable, use slicing:
s = "Nidhi"
s= s[:2]+ "X" +s[2:] # Ni+X+dhi=NiXdhi
print(s)
```

```python
# 5Ô∏è‚É£ Delete a Character
s = "gungun"
s= s[:1]+s[2:] #g+ngun=gngun
print(s)
```

```python
# 6Ô∏è‚É£ Compare Strings
a = "hello"
b= "hello"
print(a == b)
```

```python
# 7Ô∏è‚É£ String Concatenation
a = "hello"
b= "world"
print(a +" "+b)
```

```python
# 8Ô∏è‚É£ Reverse a String
a = "hello"
print(a[::-1])
```

```python
# 9Ô∏è‚É£ Rotate a String


s = "abcde"
rotated = s[2:] + s[:2] #cde+ab=cdeab
print(rotated)  # Output: cdeab


# Right Rotation by 2:
rotated = s[-2:] + s[:-2] #de+abc=deabc
print(rotated)  # Output: deabc

```

```python
# üîü Check for Palindrome
def is_Palindrome(s):
    return s == s[::-1]
print(is_Palindrome("madam"))
```
