| 4 | 4 | "d"       | "d"     | âœ… Yes     | 1 > 3 â†’ False| "bab"   |


````python
class Solution:
    def longestPalindrome(self, s:str)->str:
        if len(s) <= 1:
            return s
        longest = " "
        # Check every possible substring

        for i in range(len(s)): #5
            for j in range(i, len(s)):# i =0 ,j=0, i=0,j=1, i=0,j=2
                substring =s[i:j+1] #s[0:0+1]=S[0:1]=b,  s[0:1+1]=s[0:2]=0,1=ba , s[0:2+1]=s[0:3]=0,1,2=bab
                # Check if palindrome

                if substring == substring[::-1]:
                    if len(substring) > len(longest): # 1>0, 3>1
                        longest = substring # b, baba
        return longest   #b ,bab   
sol= Solution()
s = "babad"
print(sol.longestPalindrome(s))             
````

# 3 -6. Zigzag Conversion

### Explanation of Columns

- **curRow Before** â†’ The row we are at **before placing the character**.
- **goingDown** â†’ The direction **before** adding the character.
- **Action** â†’ Adding the character to the current row.
- **Direction Change?** â†’ Flips only if we hit **top** (`curRow = 0`) or **bottom** (`curRow = numRows - 1`).
- **curRow After** â†’ Move **up or down** depending on `goingDown`:
  - `goingDown = True` â†’ `curRow` increases by 1
  - `goingDown = False` â†’ `curRow` decreases by 1

---

### How `goingDown` Works

- Starts `False` â†’ immediately flipped `True` because the first row is top.
- When `curRow = top (0)` â†’ flip to `True` â†’ start going down.
- When `curRow = bottom (numRows - 1)` â†’ flip to `False` â†’ start going up.
- In middle rows â†’ no flip â†’ keep going in the same direction.

---

### Visual Zigzag



**Read row by row:** `"PA" + "APL" + "Y" = "PAAPLY"`


### Initial Setup

- **rows** = `["", "", ""]` â†’ one string per row
- **curRow** = `0` â†’ we start at the **top row**
- **goingDown** = `False` â†’ initially the â€œdirectionâ€ is **up** (but top row flips it immediately)


Dry Run: "PAYPAL", numRows = 3
| Character | curRow Before | goingDown | Action                    | Direction Change?                          | curRow After | Rows State         |
| --------- | ------------- | --------- | ------------------------- | ------------------------------------------ | ------------ | ------------------ |
| **P**     | 0             | False     | Add 'P' â†’ rows[0] = "P"   | curRow=0 â†’ top â†’ flip goingDown = True     | 1            | ["P", "", ""]      |
| **A**     | 1             | True      | Add 'A' â†’ rows[1] = "A"   | Middle â†’ no change                         | 2            | ["P", "A", ""]     |
| **Y**     | 2             | True      | Add 'Y' â†’ rows[2] = "Y"   | curRow=2 â†’ bottom â†’ flip goingDown = False | 1            | ["P", "A", "Y"]    |
| **P**     | 1             | False     | Add 'P' â†’ rows[1] = "AP"  | Middle â†’ no change                         | 0            | ["P", "AP", "Y"]   |
| **A**     | 0             | False     | Add 'A' â†’ rows[0] = "PA"  | curRow=0 â†’ top â†’ flip goingDown = True     | 1            | ["PA", "AP", "Y"]  |
| **L**     | 1             | True      | Add 'L' â†’ rows[1] = "APL" | Middle â†’ no change                         | 2            | ["PA", "APL", "Y"] |


````python
class Solution(object):
    def convert(self, s, numRows):
      
        # Edge case: single row means no zigzag
        if numRows == 1 or numRows >= len(s):
            return s

        # Buckets for each row
        rows = [""] * numRows

        curRow = 0
        goingDown = False

        for c in s:
            rows[curRow] += c
            # Top or bottom row â†’ change direction
            if curRow == 0 or curRow == numRows - 1:
                goingDown = not goingDown
            # Move up or down
            curRow += 1 if goingDown else -1

        # Join all rows
        return "".join(rows)
print(Solution().convert("PAYPAL",3))    
````

# 4- #14. Longest Common Prefix


Example: strs = ["flower", "flow", "flight"]

| Step | Current String (s) | Current Prefix | Action / Check | Updated Prefix |
|------|-------------------|----------------|----------------|----------------|
| 1    | "flow"            | "flower"       | Does "flow" start with "flower"? âŒ | Shrink prefix â†’ "flowe" |
| 2    | "flow"            | "flowe"        | Does "flow" start with "flowe"? âŒ  | Shrink prefix â†’ "flow" |
| 3    | "flow"            | "flow"         | Does "flow" start with "flow"? âœ…   | Prefix stays "flow" |
| 4    | "flight"          | "flow"         | Does "flight" start with "flow"? âŒ | Shrink prefix â†’ "flo" |
| 5    | "flight"          | "flo"          | Does "flight" start with "flo"? âŒ  | Shrink prefix â†’ "fl" |
| 6    | "flight"          | "fl"           | Does "flight" start with "fl"? âœ…   | Prefix stays "fl" |

âœ… Final Result: `"fl"`


````python
class Solution(object):
    def longestCommonPrefix(self, strs):
        
        if not strs:
            return ""

        # Take the first string as prefix
        prefix = strs[0]

        # Compare with the rest of the strings
        for s in strs[1:]:
            # Keep reducing prefix until it matches the beginning of s
            while not s.startswith(prefix):
                prefix = prefix[:-1]  # shrink from end
                if not prefix:       # if prefix becomes empty, no common prefix
                    return ""

        return prefix
print(Solution().longestCommonPrefix(["flower", "flow", "flight"]))    
    
````

# #125 â€” Valid Palindrome

s = "m., aD aM"
| Step | Character (`ch`) | `ch.isalnum()` | `ch.lower()` | Added to final string? | Final String So Far |
| ---- | ---------------- | -------------- | ------------ | ---------------------- | ------------------- |
| 1ï¸âƒ£  | `'m'`            | âœ… True         | `'m'`        | âœ… Yes                  | `m`                 |
| 2ï¸âƒ£  | `'.'`            | âŒ False        | â€”            | âŒ No                   | `m`                 |
| 3ï¸âƒ£  | `','`            | âŒ False        | â€”            | âŒ No                   | `m`                 |
| 4ï¸âƒ£  | `' '` (space)    | âŒ False        | â€”            | âŒ No                   | `m`                 |
| 5ï¸âƒ£  | `'a'`            | âœ… True         | `'a'`        | âœ… Yes                  | `ma`                |
| 6ï¸âƒ£  | `'D'`            | âœ… True         | `'d'`        | âœ… Yes                  | `mad`               |
| 7ï¸âƒ£  | `' '` (space)    | âŒ False        | â€”            | âŒ No                   | `mad`               |
| 8ï¸âƒ£  | `'a'`            | âœ… True         | `'a'`        | âœ… Yes                  | `mada`              |
| 9ï¸âƒ£  | `'M'`            | âœ… True         | `'m'`        | âœ… Yes                  | `madam`             |

output :-s = "madam"

.isalnum() removed spaces (kept only letters/numbers)

.lower() converted everything to lowercase

''.join() stitched all cleaned characters together into one continuous string

````python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = ''.join(ch.lower() for ch in s if ch.isalnum())
        return s == s[::-1]
str = "m., aD aM"  
print(Solution().isPalindrome(str))
````
